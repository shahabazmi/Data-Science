## 1️⃣ Comments — Notes for humans, ignored by Python

A **comment** is a line of text that you write inside your code to explain what the code is doing.

🧠 **Think of it like a sticky note.** Imagine you paste a small yellow sticky note on a page of a book. *You* can read the note, but the book's story does not change because of it. A comment works the same way — *you* read it, but the program does not run it.

**Key points about comments:**
- In Python, a comment starts with the **hash symbol `#`**.
- The **interpreter** (the program that reads and runs your Python code) completely **ignores** everything written after the `#`.
- Comments do **not** appear in the output. They exist only to help *you* and *other people* understand the code.

**Where can you write a comment?**
- On its **own line**, usually *above* the code it describes, **or**
- On the **same line**, to the *right* of the code.

> 📌 **Why comments matter:** When you open your own code after 6 months, you will forget why you wrote it. Comments remind you. In real jobs, many people read the same code, so comments help the whole team understand it quickly.

````python
# This whole line is a comment — Python completely ignores it.
print("Hello")   # This part, after the #, is also a comment.
````

## 2️⃣ The `print()` Function — Showing Output

The `print()` function is how we **display things on the screen**. Whatever we put inside the round brackets `( )` gets shown to the user.

🔑 **Two important rules to remember:**
1. `print` is a **function**. A function is a ready-made tool that does a job for us. We *call* (use) a function by writing its name followed by round brackets `()`.
2. Python is a **case-sensitive** language. This means **capital and small letters are different**. So `print()` works, but `Print()` does **not** — Python sees them as two completely different words.

````python
# print() is a function — the () means "call this function".
# Python is CASE-SENSITIVE, so these two are DIFFERENT:
#   print()  -> correct, this is the real function
#   Print()  -> wrong, Python does not know this one
print("hello world")
````

### ✍️ Text must be inside quotes

Any **text** (called a **string** in Python) must be wrapped inside quotation marks — either double quotes `" "` or single quotes `' '`. The quotes are how you *tell Python* "this is text, not code."

Look at the two cells below. The first one works. The second one is **wrong on purpose** so you can see the mistake.

````python
print("salman khan")
````

````python
# ❌ This line is WRONG on purpose (it will give an error).
# We forgot the quotation marks around salman khan.
# Python gets confused — it does not know what "salman" and "khan" mean.
# Rule: any text (a string) MUST be inside quotes " " or ' '.
print(salman khan)
````

### 🔢 `print()` can also show numbers

For numbers, you do **not** need quotes. Numbers are understood by Python directly.

````python
# We can print numbers too (no quotes needed for numbers)
print(7)
````

### 📦 `print()` can show many things at once

The `print()` function is very **flexible and powerful**. You can put many items inside it, separated by commas. Python will print them all on **one line**, automatically separated by a **single space**.

````python
# print() can take many values at once, separated by commas.
# By default, Python puts a SPACE between each value.
print("hello", 1, 2, 3, True)

# It does not matter if there are 4 items or 400 — it still works.
print("hello world", 1, 2, 3, 4.5, True, False)
````

### 🔧 The `sep` parameter — change the separator

By default the separator between items is a **space**. But we can change it to anything we want using `sep` (short for *separator*). For example, `sep="/"` puts a slash between the items instead of a space.

````python
# 'sep' decides what goes BETWEEN the items.
# Here we use "/" instead of the default space.
print("hello world", 1, 2, 3, 4.5, True, False, sep="/")
````

### ↩️ Two `print()` statements = two lines

Normally, each `print()` finishes by moving to a **new line**. So if you write two `print()` statements, the output appears on two separate lines.

````python
# Each print() ends by moving to a new line by default.
print("hello")
print("world")
````

### ➡️ The `end` parameter — print on the same line

If we do **not** want to move to a new line after printing, we use `end`. It decides what to add at the **end** of the printed text. By default `end="\n"` (a new line). Here we change it to `"-"`, so the two prints stay on the same line, joined by a dash.

````python
# 'end' decides what comes AFTER the printed text.
# Here end="-" keeps both prints on the SAME line, joined by a dash.
print("hello", end="-")
print("world")
````

## 3️⃣ Data Types — The Different Kinds of Data

Data comes in different **types**, just like in real life we have different kinds of things: numbers, words, yes/no answers, lists of items, and so on. Python has a specific type for each kind of data.

Here is a quick summary. We will look at each one below.

| Type | Meaning | Example | Written with |
|------|---------|---------|--------------|
| **int** | Whole number | `8`, `100`, `-5` | just the number |
| **float** | Decimal number | `8.55`, `1.5` | number with a dot |
| **bool** | True / False | `True`, `False` | no quotes |
| **str** | Text (string) | `"hello"` | quotes `" "` |
| **complex** | Complex number | `5 + 6j` | uses `j` |
| **list** | Ordered collection (changeable) | `[1, 2, 3]` | square brackets `[ ]` |
| **tuple** | Ordered collection (fixed) | `(1, 2, 3)` | round brackets `( )` |
| **set** | Collection of unique items | `{1, 2, 3}` | curly braces `{ }` |
| **dict** | Key–value pairs | `{"name": "Shahab"}` | curly braces with `:` |

### ➊ Integer (`int`) — whole numbers

An **integer** is a whole number (no decimal point), like `8`, `100`, or `-5`.

**Interesting fact:** Python can handle *very, very large* numbers. `1e308` means `1 × 10³⁰⁸` (the number 1 followed by 308 zeros!). But there is a limit — beyond about `1e309`, Python simply calls it `inf`, which means **infinity** (a number too big to store).

````python
# A simple integer
print(8)

# Python can handle huge numbers.
# 1e308 means 1 x 10^308 (1 followed by 308 zeros)
print(1e308)

# But beyond the limit, Python shows 'inf' (infinity — too big to store)
print(1e309)
````

### ➋ Float — decimal numbers

A **float** is a number with a decimal point, like `8.55`. The word "float" comes from *floating-point number*. Just like integers, floats also have a maximum limit, after which Python shows `inf` (infinity).

````python
# A float is a number with a decimal point
print(8.55)

# Floats also have a maximum limit
print(1.7e308)   # this still works
print(1.7e309)   # too big -> shows 'inf' (infinity)
````

### ➌ Boolean (`bool`) — True or False

A **boolean** can only be one of two values: `True` or `False`. It is used to answer yes/no questions in code (Is the user logged in? Is 5 greater than 3?).

> ⚠️ **Very important:** Write `True` and `False` with a **capital first letter** and **without quotes**. `"True"` (with quotes) would be text, not a boolean.

````python
# Boolean = True or False (capital letter, NO quotes)
print(True)
print(False)
````

### ➍ String (`str`) — text

A **string** is any text: a word, a sentence, or even a single letter. It must always be written **inside quotes** `" "` or `' '`.

````python
# A string is text inside quotes
print("hello world")
````

### ➎ Complex number

A **complex number** has two parts: a *real* part and an *imaginary* part. In Python the imaginary part is written with the letter `j` (in maths it is usually `i`). You will mostly meet these in advanced maths, so just know that they exist.

````python
# Complex number: real part 5, imaginary part 6 (written as 6j)
print(5 + 6j)
````

### ➏ List — an ordered collection you can change

A **list** stores many items together in a fixed order. In languages like C, this is called an **array**.

🧠 **Analogy:** A list is like a **shopping bag** — you can put many items inside, keep them in order, and add or remove items whenever you want.

A list is written inside **square brackets `[ ]`**, with items separated by commas.

````python
# A list holds many items, written inside square brackets [ ]
# (In the C language, this is called an "array".)
print([1, 2, 3, 4, 5])
````

### ➐ Tuple — an ordered collection you *cannot* change

A **tuple** is very similar to a list, but with one big difference: **once created, you cannot change it** (you cannot add or remove items). Because of this, a tuple is used for data that should stay fixed.

A tuple is written inside **round brackets `( )`**.

> ✅ **Correction to remember:** A tuple uses **round brackets `( )`**, *not* square brackets. (Square brackets `[ ]` are for lists.)

````python
# A tuple holds many items, written inside round brackets ( )
# A tuple cannot be changed after it is created.
print((1, 2, 3, 4, 5))
````

### ➑ Set — a collection of *unique* items

A **set** stores many items, but with two special features:
1. It **does not allow duplicates** (every item must be unique).
2. It has **no fixed order**.

A set is written inside **curly braces `{ }`**.

````python
# A set holds unique items, written inside curly braces { }
# Sets do not allow duplicate values.
print({1, 2, 3, 4, 5})
````

### ➒ Dictionary (`dict`) — key–value pairs

A **dictionary** stores data in **pairs**: a **key** and its **value**, written as `key: value`.

🧠 **Analogy:** Think of a real dictionary — you look up a *word* (the key) to find its *meaning* (the value). Or think of a student ID card: the label `"name"` (key) points to `"Shahab"` (value).

A dictionary is written inside **curly braces `{ }`**, and each pair uses a colon `:`.

````python
# A dictionary stores key: value pairs, inside curly braces { }
# Left of ':' is the KEY, right of ':' is the VALUE.
print({"name": "Shahab", "gender": "Male", "Weight": "70"})
````

## 4️⃣ Variables — Boxes That Store Information

A **variable** is a **container (a box)** where we store information so we can use it later.

### 🤔 Why do we need variables?

Imagine you are building a website. On the homepage, you want to show **"Hello, [user's name]"** to whoever logs in.

But here is the problem: **you do not know who will log in!** Today it might be *Rahul*, tomorrow *Varun*, next week someone else. You cannot fix one name in the code.

**The solution:** You keep a variable — let's call it `name` — right after the word "Hello". When a user logs in, their real name is taken from the database and placed inside that `name` variable automatically. So the same code shows "Hello, Rahul" for one user and "Hello, Varun" for another.

👉 **In short: a variable is a container that holds a value for future use.**

### 🐍 Making a variable in Python is easy

In languages like C or C++, you must **declare the type** first, like this:
```c
int name = "shahab";   // you must say it is an int, char, etc.
```

But in Python, you simply write:
```python
name = "shahab"
print(name)
```

Python is **smart enough** to understand the data type on its own. You do not have to tell it. 🙌

````python
# Create two variables and store numbers in them
a = 5
b = 6

# Use the variables to do a calculation
print(a + b)
````

## 5️⃣ Dynamic vs Static Typing ⭐ (Popular Interview Question)

This is about **who decides the data type** of a variable — you, or the language.

### 🟢 Dynamic Typing (Python uses this)

In **dynamic typing**, you do **not** tell the data type when creating a variable. The Python **interpreter figures it out automatically**.

```python
a = 5        # Python understands on its own that this is an integer
```

> 📌 **Definition:** Python is a **dynamically typed** language. This means Python's interpreter is intelligent enough to *automatically understand* the data type of a variable, based on the value you give it.

### 🔵 Static Typing (C, C++, Java use this)

In **static typing**, you **must** clearly mention the data type while creating the variable.

```c
int a = 5;   // here we clearly say 'a' is an int
```

| | Dynamic Typing | Static Typing |
|--|----------------|---------------|
| **Tell the type?** | No, language decides | Yes, you must declare |
| **Example languages** | Python | C, C++, Java |
| **Example code** | `a = 5` | `int a = 5;` |
| **Flexibility** | More flexible | More strict |

## 6️⃣ Keywords and Identifiers

### 🔒 Keywords — words reserved by Python

**Keywords** are special words that Python has **reserved for itself**. Python uses these words to understand and run your program (for example: `if`, `else`, `print`, `for`, `True`, `False`).

Because Python already uses these words for its own purpose, **you are not allowed to use them as your own variable names**. If you did, Python would get confused.

> 🧠 **Simple idea:** Keywords are like the *official words* of the language. You cannot rename them or use them for something else.

**How many keywords are there?** Older notes often say **32**. In current Python (version 3.10 and above) there are **35** keywords. You do not need to memorise them — you can ask Python to list them, as shown below.

### 🖥️ Bonus: Compiler vs Interpreter (quick note)
- A **compiler** translates the *whole* program into machine code *at once*, then runs it — this is usually **faster** overall.
- An **interpreter** translates and runs the code *line by line* — a little slower, but easier to test.
- Python uses an **interpreter**, and it supports both **single-thread and multi-thread** execution (doing one task at a time, or many tasks together).

````python
# You can ask Python to show all its keywords using the 'keyword' module.
import keyword

print(len(keyword.kwlist))   # how many keywords exist
print(keyword.kwlist)        # the full list
````

### 🏷️ Identifiers — the names *you* give

An **identifier** is any **name that you create** in your program. When you make a variable, a function, or a class, the name you give it is called an *identifier*.

For example, in `age = 20`, the name `age` is an identifier.

**📋 Rules for naming identifiers in Python:**

| Rule | ❌ Wrong | ✅ Correct |
|------|---------|-----------|
| 1. Cannot **start with a digit** | `1name` | `name1` |
| 2. Can use the special character **underscore `_`** | `name@` | `name_your`, `_` |
| 3. Cannot be a **keyword** (`print`, `if`, `else`, `True`, `False`, ...) | `if = 5` | `my_if = 5` |

> ⚠️ Only the underscore `_` is allowed as a special character. Symbols like `@`, `#`, `$` are **not** allowed in names.

Let's test each rule with real code. 👇

**Rule 1 test:** A name cannot start with a digit. The cell below is **wrong on purpose** — it will give an error.

````python
# ❌ WRONG: a variable name cannot start with a number.
1name = "shahab"
print(1name)
````

**Rule 1 fixed:** Put the digit *after* the letters. `name1` is perfectly valid.

````python
# ✅ CORRECT: the name starts with a letter, digit comes after.
name1 = "shahab"
print(name1)
````

**Rule 2 test:** The underscore `_` is allowed inside a name.

````python
# ✅ CORRECT: underscore _ is allowed in names
name_your = "shahab"
print(name_your)
````

**Rule 2 (special case):** You can even use a single underscore `_` as a full variable name. But symbols like `@`, `#`, `$` are **not** allowed.

````python
# A single underscore _ can even be used as a variable name.
# But @, #, $ etc. are NOT allowed in variable names.
_ = "shahab"
print(_)
````

## 8️⃣ Taking Input From the User

To make dynamic software, we need to **ask the user for information**. In Python we do this with the **`input()`** function.

- `input()` **pauses** the program and waits for the user to type something and press Enter.
- We can also show a **message (prompt)** inside the brackets, like `input("Enter your name: ")`, so the user knows what to type.
- Whatever the user types is usually **stored in a variable** so we can use it later.

> ⚠️ **Extremely important rule:** `input()` **always** gives back the value as a **string (text)** — even if the user types a number! This causes a common beginner mistake, which we will see below.

````python
# The simplest form: input() just waits for the user to type something.
input()
````

````python
# We can show a message (prompt) so the user knows what to type.
input("enter your name")
````

### 🐞 The classic beginner trap

Below, we ask the user for two numbers and try to add them. But something surprising happens: instead of `32 + 32 = 64`, we get `3232`!

**Why?** Because `input()` gives back **text**, not numbers. And when you use `+` between two pieces of text, Python **joins them together** (this is called *string concatenation*) instead of adding. So `"32" + "32"` becomes `"3232"`.

We can prove the values are text by using `type()`, which tells us the data type. It shows `<class 'str'>`, meaning **string**.

````python
# Take two inputs from the user and store them in variables
fnum = input("enter first number")
snum = input("enter second number")
print(fnum, snum)

# Try to add them
result = fnum + snum
print(result)   # Surprise! We get 3232, not 64

# Let's check WHY — what data type are they?
print(type(fnum))
print(type(snum))
# Both are 'str' (string). input() ALWAYS returns text by default,
# so '+' joined the text instead of adding the numbers.
````

## 9️⃣ Type Conversion — Changing One Data Type Into Another

**Type conversion** means **changing data from one type to another** — for example, turning the text `"32"` into the number `32`.

> 📌 **Important:** Type conversion does **not** change your original data. Python does *not* edit the old value — it **creates a brand-new value** in memory. The original stays exactly the same.

There are **two types** of type conversion:

### 🔵 1. Implicit Conversion (Python does it automatically)
Here the **interpreter changes the type by itself**. The programmer does not have to do anything.

For example, when you add an integer and a float, Python automatically turns the integer into a float so the maths works.

````python
# IMPLICIT conversion: Python handles the type change on its own.
print(5 + 5.6)              # int + float -> Python makes the result a float
print(type(5), type(5.6))   # one is int, the other is float
````

**But implicit conversion does not always work.** Python can only do the *safe, obvious* conversions automatically. The cell below tries to mix things in a way Python cannot fix by itself, so it gives an **error** (it is wrong on purpose). When implicit conversion fails, we must do it ourselves — that is **explicit** conversion.

````python
# ❌ WRONG on purpose: Python cannot fix this automatically -> error.
# When implicit conversion does not work, we must convert manually (explicit).
print(4 + , "4")
````

### 🟢 2. Explicit Conversion (the programmer does it manually)

Here **you** clearly tell Python to change the data type, using built-in functions:

| Function | Converts to | Example | Result |
|----------|-------------|---------|--------|
| `int()` | integer | `int("4")` | `4` |
| `str()` | string (text) | `str(5)` | `"5"` |
| `float()` | float (decimal) | `float(2.3)` | `2.3` |

> ⚠️ Note: You **cannot** convert a *complex* number into an integer.

````python
# EXPLICIT conversion: we tell Python to change the type ourselves.

# string "4" -> integer 4
int("4")
type(int("4"))   # confirms it is now an int

# integer -> string
str(5)

# make a float
float(2.3)
````

### ✅ Fixing the input problem with type conversion

Remember the `3232` problem? Now we can fix it. We convert each input into an integer using `int()` **before** adding. This time we get the correct answer: `12 + 12 = 24`.

Notice that `type(fnum)` is still `str` in the first version below — because we only converted the values *inside* the calculation, not the original `fnum` variable.

````python
# Method 1: convert the values while adding.
fnum = input("enter first number:")
snum = input("enter second number:")
print(fnum, snum)

result = int(fnum) + int(snum)   # convert to int, THEN add
print(result)                    # correct answer: 24

print(type(fnum))  # still 'str' — we only converted inside the calculation
````

**A cleaner way:** wrap the `input()` directly inside `int()`. This converts the value *the moment* it is typed, so `fnum` itself becomes an integer. Now `type(fnum)` shows `int`.

````python
# Method 2 (cleaner): convert the input immediately using int(input(...)).
fnum = int(input("enter first number:"))
snum = int(input("enter second number:"))
print(fnum, snum)

result = fnum + snum
print(result)          # correct answer: 24

print(type(fnum))      # now it is 'int' from the start
````

## 🔟 Literals — The Actual Values We Store

A **literal** is the **actual value** that we store inside a variable.

Let's break down a simple line: `a = 2`
- `a` → the **variable** (the box)
- `=` → the **operator** (the assignment sign that puts the value into the box)
- `2` → the **literal** (the actual value)

So, in short: **the value on the right side of `=` is the literal.**

Python has several ways of writing number literals. Let's explore them.

### 🔢 Integer literals in different number systems

Computers can understand numbers written in different *number systems*. Python lets us write the same value in four ways using special prefixes:

| System | Prefix | Example | Normal value |
|--------|--------|---------|--------------|
| **Binary** (base 2) | `0b` | `0b1010` | `10` |
| **Decimal** (base 10) | *(none)* | `100` | `100` |
| **Octal** (base 8) | `0o` | `0o310` | `200` |
| **Hexadecimal** (base 16) | `0x` | `0x12c` | `300` |

Whatever prefix you use, Python stores and prints the **normal decimal value**.

````python
# Four ways to write whole-number literals:
a = 0b1010   # binary literal      (prefix 0b)
b = 100      # decimal literal     (no prefix)
c = 0o310    # octal literal       (prefix 0o)
d = 0x12c    # hexadecimal literal (prefix 0x)
````

````python
# Binary literal -> Python prints the normal decimal value
a = 0b1010
print(a)
````

````python
# Decimal literal
b = 100
print(b)
````

````python
# Octal literal
c = 0o310
print(c)
````

````python
# Hexadecimal literal
d = 0x12c
print(d)
````

### 🌊 Float literals

Floats (decimal numbers) can also be written using scientific notation with `e`:
- `1.5e2` means `1.5 × 10²` = `150.0`
- `1.5e-3` means `1.5 × 10⁻³` = `0.0015`

````python
# Different ways to write float (decimal) literals
float_1 = 10.5        # normal decimal
float_2 = 1.5e2       # 1.5 x 10^2  = 150.0
float_3 = 1.5e-3      # 1.5 x 10^-3 = 0.0015
````

### 🔷 Complex literals

A complex number has a **real** part and an **imaginary** part. We can read each part separately using `.real` and `.imag`. In `x = 3.14j`, the imaginary part is `3.14` and the real part is `0.0`.

````python
# A complex literal: 3.14j has only an imaginary part
x = 3.14j
print(x.imag)   # imaginary part -> 3.14
print(x.real)   # real part      -> 0.0
````

### 🔤 String literals

Text values are **string literals**. A few things to know:
- You can use **double quotes** `" "` or **single quotes** `' '` — both work the same way.
- A single character (like `"c"`) is also just a string in Python.
- There are also special string types (shown as comments): **multiline strings** (triple quotes `""" """`), **raw strings** (`r"..."` — keeps backslashes as they are), and **unicode strings** (`u"..."`).

````python
# Strings can use double or single quotes — both are the same.
strings = ("this is python")
string  = ('this is python')
char    = "c"    # a single character is still a string

# Extra string types (kept as notes for later):
# multiline_str = """this is a multiline python program with more than one line"""
# unicode_str   = u"\u0000 some text"
# raw_str       = r"raw \n string"   # r means 'raw' — \n stays as \n, not a new line

print(strings)
print(string)
print(char)
````

### ✔️ Booleans behave like numbers

A neat fact: Python treats **`True` as `1`** and **`False` as `0`**. So you can even use them in maths!
- `True + 4` → `1 + 4` → `5`
- `False + 10` → `0 + 10` → `10`

````python
# Python treats booleans as numbers: True = 1, False = 0
a = True + 4     # 1 + 4  = 5
b = False + 10   # 0 + 10 = 10
print("a", a)
print("b", b)
````

### 🚫 The `None` literal — "nothing"

`None` is a special literal that means **"no value" or "nothing here yet."** It is used when you want a variable to exist but not hold any real value for now.

> 📌 `None` is written with a **capital N** and **no quotes**. It is not `0`, not `False`, and not an empty text — it literally means *nothing*.

````python
# None means 'no value' / 'nothing'
a = None
print(a)
````

| Topic | Key idea to remember |
|-------|----------------------|
| **Comments** | Start with `#`; Python ignores them; they are notes for humans |
| **`print()`** | Shows output; case-sensitive; use `sep` and `end` to control formatting |
| **Data types** | int, float, bool, str, complex, list `[]`, tuple `()`, set `{}`, dict `{k:v}` |
| **Variables** | Containers that store values for later use |
| **Dynamic typing** | Python decides the data type automatically (no need to declare it) |
| **Keywords** | Reserved words (35 in modern Python); cannot be used as names |
| **Identifiers** | Names you create; can't start with a digit; only `_` is allowed as a symbol |
| **Software types** | Static (no user interaction) vs Dynamic (talks to the user) |
| **`input()`** | Asks the user for data; **always returns a string** |
| **Type conversion** | Implicit (automatic) vs Explicit (`int()`, `str()`, `float()`) |
| **Literals** | The actual value stored (binary `0b`, octal `0o`, hex `0x`, `None`, etc.) |

### 📝 Try It Yourself (Practice Questions)

1. Print your **name**, your **age**, and `True`, all on **one line**, separated by a **comma and a space**. *(Hint: use `sep`.)*
2. Ask the user for their **birth year**, then calculate and print their **age**. *(Hint: `input()` returns a string — convert it!)*
3. Create a variable using an **underscore** in its name and print it.
4. What will `False + True + 5` print? First **guess**, then run it to check.
5. Use type conversion: turn the string `"42"` into a number with `int("42")`, and `"3.5"` into a decimal with `float("3.5")`, then print both results.
