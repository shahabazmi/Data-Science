## 1️⃣ Problem 1 — Length of a String *without* `len()`

**Question:** Find the number of characters in a string, but you are **not allowed** to use the `len()` function.

**The idea:** We go through the string **one character at a time** using a `for` loop, and keep a **counter** that increases by 1 for each character. At the end, the counter holds the total length.

🧠 *Think of counting people entering a room — you add 1 each time someone walks in.*

```python
# Find the length of a string WITHOUT using len()
s = input("enter the strings: ")
counter = 0
for i in s:          # go through every character
    counter += 1     # add 1 for each character
print("length of strings is", counter)
```

## 2️⃣ Problem 2 — Extract the Username from an Email

**Question:** Given an email, print the **username** (the part *before* the `@` sign).
Example: for `shahabazmi321@gmail.com`, the username is `shahabazmi321`.

**The idea:**
1. Find the **position** of the `@` sign using `index("@")`.
2. Use **slicing** to take everything from the start up to (but not including) that position.

```python
# Extract the username (the part before @) from an email
s = input("enter the email")
pos = s.index("@")     # find the position of the @ sign
print(s[0:pos])        # slice from the start up to the @
```

## 3️⃣ Problem 3 — Count the Frequency of a Character

**Question:** Count how many times a particular character appears in a string.
Example: in `"hello how are you"`, the letter `h` appears **2** times.

**The idea:** Loop through each character. Every time it **matches** the character we are searching for, we add 1 to our counter.

```python
# Count how many times a particular character appears
s = input("enter the strings: ")
term = input("what would like to search for: ")
counter = 0
for i in s:
    if i == term:      # does this character match what we are searching for?
        counter += 1   # if yes, count it
print("frequency", counter)
```

## 4️⃣ Problem 4 — Remove a Particular Character

**Question:** Write a program that removes **every occurrence** of a chosen character from a string.

**The idea:** Since strings are **immutable** (cannot be changed), we build a **new string**. We loop through the original and add each character to the new string — **except** the one we want to remove.

```python
# Remove a particular character from a string
s = input("enter the strings")
term = input("what would i like to remove")
result = ""             # start with an empty new string
for i in s:
    if i != term:       # keep every character EXCEPT the one to remove
        result = result + i
print(result)
```

## 5️⃣ Problem 5 — Check for a Palindrome

**Question:** A **palindrome** is a word that reads the **same forwards and backwards** (for example: *madam*, *malayalam*, *level*). Check whether a given string is a palindrome.

We will solve this in **three different ways** — from the detailed method to the shortest.

### Way 1 — Compare characters from both ends (using a flag)

**The idea:** Compare the **first** character with the **last**, the **second** with the **second-last**, and so on. If any pair does **not** match, it is not a palindrome.

We use a **flag** variable (`True`/`False`) to remember the result. We only need to check **half** the string (`len(s)//2`), because the other half is just the mirror.

```python
# Palindrome check — Way 1: compare characters from both ends
s = input("enter the strings")
flag = True   # this flag tracks whether the string is a palindrome
for i in range(0, len(s) // 2):        # only need to check the first half
    if s[i] != s[len(s) - i - 1]:      # compare front character with back character
        flag = False
        print("not a palindrome")
        break                          # mismatch found, no need to check further
if flag:
    print("palindrome")
```

### Way 2 — The short trick using slicing

**The idea:** `s[::-1]` reverses the string. If the reversed string is **equal** to the original, it is a palindrome. This is the cleanest one-line comparison.

```python
# Palindrome check — Way 2: compare the string with its reverse
s = input("Enter a string: ")
if s == s[::-1]:        # s[::-1] is the reversed string
    print("palindrome")
else:
    print("not a palindrome")
```

### Way 3 — Using a function

**The idea:** Wrap the logic inside a **function** so we can reuse it any time. A function is like a machine: we give it a string, and it gives back the answer.

```python
# Palindrome check — Way 3: using a function
# Define a function
def isPalindrome(string):
    if string == string[::-1]:
        return "The string is a palindrome."
    else:
        return "The string is not a palindrome."

# Enter input string
string = input("Enter string: ")

print(isPalindrome(string))
```

**A cleaner function** that returns `True`/`False` instead of a full sentence. The part that actually *runs* the function is called the **driver code**.

```python
# Palindrome function that returns True or False
def isPalindrome(s):
    return s == s[::-1]

# Driver code (the part that runs the function)
s = "malayalam"
ans = isPalindrome(s)

if ans:
    print("Yes")
else:
    print("No")
```

## 6️⃣ Problem 6 — Count Words *without* `split()`

**Question:** Count the number of words in a sentence, but you cannot use `split()`.

**The idea:** A **space** separates words. So we build each word letter by letter into a temporary variable `temp`. Every time we hit a **space**, it means one word has ended — we save it into a list and reset `temp`. After the loop, we add the **last** word (because there is no space after it).

```python
# Count the number of words WITHOUT using split()
s = input("enter the strings")
l = []             # a list to store the words
temp = ""          # builds the current word letter by letter
for i in s:
    if i != " ":
        temp = temp + i      # not a space -> keep building the word
    else:
        l.append(temp)       # hit a space -> the word is complete, save it
        temp = ""            # reset for the next word
l.append(temp)               # add the LAST word (no space comes after it)
print(l)
```

## 7️⃣ Problem 7 — Title Case *without* `title()`

**Question:** Convert a sentence to **title case** (the first letter of every word capital), but without using the `title()` function.
Example: `hi my name is shahab` → `Hi My Name Is Shahab`.

**The idea:**
1. Break the sentence into words using `split()`.
2. For each word, make the **first letter uppercase** (`i[0].upper()`) and the **rest lowercase** (`i[1:].lower()`), then join them.
3. Finally, `join` the words back with a space.

```python
# Convert a string to title case WITHOUT using title()
s = input("enter the strings")
l = []
for i in s.split():                          # go through each word
    l.append(i[0].upper() + i[1:].lower())   # first letter capital + rest small
print(" ".join(l))                           # join the words back into a sentence
```

## 8️⃣ Problem 8 — Convert an Integer to a String (manually)

**Question:** Convert a number like `2345` into the **text** `"2345"`, without simply using `str()`.

**The idea:** We already know two tricks from before:
- `number % 10` gives the **last digit**.
- `number // 10` **removes** the last digit.

We take digits off from the right one by one, look up each digit in the text `"0123456789"`, and **add it to the front** of our result string (so the order stays correct). We repeat until the number becomes 0.

```python
# Convert an integer to a string manually (without str())
number = int(input("enter the number"))
digits = "0123456789"        # position 0 has "0", position 5 has "5", etc.
result = ""
while number != 0:
    result = digits[number % 10] + result   # take last digit as text, add it to the FRONT
    number = number // 10                    # remove the last digit
print(result)
print(type(result))          # proof: it is now a string (str)
```
