## 2. Output — the `print()` Function

`print()` ek **function** hai jo screen par output display karta hai.

**Remember:** Python ek **case-sensitive** language hai. `print()` aur `Print()` alag hai —
sirf lowercase `print()` chalega, `Print()` error dega.

````python
# print() is a function — the () means "function"
# Python is a CASE-SENSITIVE language:
#   print()  -> works
#   Print()  -> error
# So print() and Print() are NOT the same.

print("hello world")
````

````python
print("salman khan")
````

````python
# This line will give an ERROR (intentionally) 👇
print(salman khan)

# Why the error?
# We did NOT put salman khan inside quotes "".
# So Python gets confused — it thinks salman and khan are variables (which don't exist).
# 👉 Inverted commas ("") tell Python that the thing inside is a STRING (text).
# Correct way:  print("salman khan")
````

````python
# We can also print numbers (no quotes needed for numbers)
print(7)
````

````python
# Python's print() is flexible and powerful.
# We can print many things in one print() — Python separates them with a SPACE by default.

print("hello", 1, 2, 3, True)
print("hello world", 1, 2, 3, 4.5, True, False)
````

````python
# Question: why is there a space between the outputs above?
# Answer: print() uses a SPACE as the default separator.
# We can change it using sep="..."

print("hello world", 1, 2, 3, 4.5, True, False, sep="/")
````

````python
# What if we want two print() outputs?
# By default, each print() prints on a NEW line.

print("hello")
print("world")
````

````python
# If we want both prints on the SAME line, use end="..."
# end controls what comes AFTER the print (default is a newline "\n").

print("hello", end="-")
print("world")
````

## 3. Data Types

Ek **data type** batata hai ki hum kis kind ki value ke saath kaam kar rahe hai. Python me
kai built-in data types hai:

1. **Integer (`int`)** — whole numbers, jaise `8`
2. **Float (`float`)** — decimal wale numbers, jaise `8.55`
3. **Boolean (`bool`)** — sirf `True` ya `False` (bina quotes ke)
4. **String (`str`)** — text, quotes ke andar jaise `"hello"`
5. **Complex (`complex`)** — real + imaginary part, jaise `5 + 6j`
6. **List** — square brackets me, jaise `[1, 2, 3]`
7. **Tuple** — round brackets me, jaise `(1, 2, 3)`
8. **Set** — curly braces me, jaise `{1, 2, 3}`
9. **Dictionary** — `key: value` pairs, jaise `{"name": "Shahab"}`

````python
# 1. Integer (int) — whole numbers, positive or negative
print(8)

# 👉 Python integers can be VERY large (practically unlimited size):
print(2 ** 200)   # 2 to the power 200 — still an exact integer!

# Note: 1e308 below is actually a FLOAT (decimal), not an int.
# Floats DO have a limit (~1.8e308). Beyond that, Python shows 'inf' (infinity):
print(1e308)   # 1 * 10^308  -> works
print(1e309)   # too big for a float -> inf
````

````python
# 2. Float — decimal numbers
print(8.55)

# Floats also have a maximum limit (~1.8e308):
print(1.7e308)   # works
print(1.7e309)   # too big -> inf
````

````python
# 3. Boolean — only two values: True or False
# 👉 Remember: do NOT put quotes around them (True, not "True")
print(True)
print(False)
````

````python
# 4. String (str) — text, always inside quotes
print("hello world")
````

````python
# 5. Complex number — has a real part and an imaginary part (j)
print(5 + 6j)
````

````python
# 6. List — written inside SQUARE brackets [ ]
# (In C language, the closest thing is an array.)
print([1, 2, 3, 4, 5])
````

````python
# 7. Tuple — written inside ROUND brackets ( )   ✅  (NOT square brackets!)
print((1, 2, 3, 4, 5))
````

````python
# 8. Set — written inside CURLY braces { }
# (A set automatically removes duplicate values.)
print({1, 2, 3, 4, 5})
````

````python
# 9. Dictionary — key:value pairs inside CURLY braces { }
print({"name": "Shahab", "gender": "Male", "weight": "70"})
````

## 4. Variables

Ek **variable** ek container (dabba) hai jisme hum information (data) store karte hai.

**Variables kyun use karte hai?**

Software banate waqt aksar hume pehle se nahi pata hota ki user kya value dega. For example,
maan lo ek app hai jo login karne wale ke homepage par "Hello [name]" dikhati hai. Ab kal
kaun login karega — Rahul, Varun, ya koi aur — ye pehle se pata nahi. Is case me hum "Hello"
ke aage ek **variable** rakh dete hai, aur jo bhi user login karega uska naam database se
uthaake us variable me daal dete hai.

Short me: variable ek container hai jo aage use ke liye value store karke rakhta hai.

### Variable kaise banate hai?

**C / C++** me data type declare karna padta hai:

```c
int name = "shahab";   // type (int) khud batana padta hai
```

**Python** me bas itna likhte hai:

```python
name = "shahab"
print(name)
```

Python khud smart hai — value dekh ke data type samajh leta hai, hume declare karne ki
zaroorat nahi hoti.

````python
a = 5
b = 6
print(a + b)
````

## Interview Question — Dynamic vs Static Typing

**1. Dynamic Typing**

C/C++ me variable banate waqt type batana padta hai:
```c

```
Python me seedha:
```python

```
Yaani **dynamic typing** me hum type nahi batate — Python interpreter khud value se samajh
leta hai. Python dynamic typing follow karta hai.

**2. Static Typing**

Jahan variable banate waqt type **batana zaroori** hota hai — jaise C / C++.

## 5. Keywords & Identifiers

### Keywords

**Keywords** wo special reserved words hai jo Python ne pehle se rakh rakhe hai (jaise `if`,
`else`, `for`, `True`, `import`). Python inhi ki madad se program ko samajhta aur run karta hai.

- Inhe hum variable name ke roop me use nahi kar sakte. Jaise `print` ko variable bana denge
  to Python confuse ho jaayega.
- Python 3 me **35 keywords** hai (version ke hisaab se ye number thoda change ho sakta hai;
  purane Python me 33 the).

Quick note (Compiler vs Interpreter): **Compiler** poore program ko ek saath translate karta
hai (generally fast), jabki **Interpreter** line-by-line translate karta hai (thoda slow).
Python interpreter use karta hai.

### Identifiers

Ek **identifier** koi bhi naam hai jo hum program me khud dete hai — variable, function, ya
class ka naam.

**Naming ke rules:**

1. Identifier digit se **start nahi** ho sakta.
   ```python
   1name = "shahab"   # error
   name1 = "shahab"   # chalega
   ```
2. Underscore `_` use kar sakte hai (par `@`, `#`, `$` jaise symbols nahi).
3. Koi bhi **keyword** identifier nahi ban sakta (`if`, `else`, `True`, ...).

````python
# Rule 1: an identifier cannot START with a digit — this will give an error 👇
1name = "shahab"
print(1name)
````

````python
# But a digit AFTER the first character is fine
name1 = "shahab"
print(name1)
````

````python
# Underscore _ is allowed in identifiers
name_your = "shahab"
print(name_your)
````

````python
# A single underscore _ can even be used as a variable name
# (But symbols like @, #, $ are NOT allowed.)
_ = "shahab"
print(_)
````

## 6. Taking Input from the User

**`input()`** function program ko user se input (text) lene deta hai. Ek important baat yaad
rakho — `input()` hamesha ek **string** return karta hai, chahe user number hi kyun na type kare.

````python
# A basic input() — it waits for the user to type something
input()
````

````python
# We can show a message (prompt) to guide the user
input("enter your name: ")
````

````python
# Taking input from the user and storing it in variables
fnum = input("enter first number: ")
snum = input("enter second number: ")
print(fnum, snum)

result = fnum + snum
print(result)
# ⚠️ The output is NOT what we expect — it joins them (like "23") instead of adding.

# Why? Let's check the types:
print(type(fnum))
print(type(snum))
# 👉 input() ALWAYS returns a STRING by default.
# So "2" + "3" becomes "23" (text joining), not 2 + 3 = 5.
# To do maths, we must convert the input into an integer first (see Type Conversion below).
````

## 7. Type Conversion

**Type conversion** ka matlab hai ek data type ko doosre me convert karna.

**Important:** type conversion original data ko change nahi karta — ye memory me ek bilkul
nayi value banata hai, aur original waisa ka waisa rehta hai.

Do tarah ke conversion hote hai:
1. **Implicit** — Python khud automatically karta hai.
2. **Explicit** — programmer khud manually karta hai.

````python
# 1. IMPLICIT type conversion
# The interpreter does it AUTOMATICALLY — the programmer doesn't need to do anything.

print(5 + 5.6)              # int + float -> Python automatically makes the result a float
print(type(5), type(5.6))   # <class 'int'> <class 'float'>

# But implicit conversion does NOT always work:
print(4 + "4")              # ERROR: Python can't add an int and a str automatically
# 👉 When Python can't convert automatically, WE must do it -> EXPLICIT conversion.
````

````python
# 2. EXPLICIT type conversion
# The programmer explicitly tells Python to change the data type.
# Built-in conversion functions: int(), str(), float(), ...

# str -> int
print(int("4"), type(int("4")))     # 4 <class 'int'>

# int -> str
print(str(5), type(str(5)))         # '5' <class 'str'>

# -> float
print(float(2.3), type(float(2.3))) # 2.3 <class 'float'>

# ⚠️ Note: a complex number CANNOT be converted to int.
````

````python
# Now let's fix the earlier addition problem using type conversion.
fnum = input("enter first number: ")
snum = input("enter second number: ")
print(fnum, snum)

result = int(fnum) + int(snum)   # convert to int ONLY for the calculation
print(result)

print(type(fnum))   # still <class 'str'> — we changed only 'result', not fnum itself
````

````python
# Even cleaner: convert the input to int right away.
fnum = int(input("enter first number: "))
snum = int(input("enter second number: "))
print(fnum, snum)

result = fnum + snum
print(result)

print(type(fnum))   # now <class 'int'> — we converted the user's input immediately
````

## 8. Literals

Ek **literal** wo actual value hai jo hum variable me store karte hai.

```python
a = 2
```
Yahan:
- `a` — **variable**
- `=` — **assignment operator**
- `2` — **literal** (the value)

Aage integer (binary/octal/hex), float, complex, string, boolean aur `None` literals dekhte hai.

````python
# Different ways to write INTEGER literals:
a = 0b1010   # binary literal       (base 2)
b = 100      # decimal literal      (base 10)
c = 0o310    # octal literal        (base 8)
d = 0x12c    # hexadecimal literal  (base 16)
````

````python
# Binary literal -> Python prints its decimal value
a = 0b1010
print(a)   # 10
````

````python
# Decimal literal
b = 100
print(b)   # 100
````

````python
# Octal literal
c = 0o310
print(c)   # 200
````

````python
# Hexadecimal literal
d = 0x12c
print(d)   # 300
````

````python
# FLOAT literals
float_1 = 10.5
float_2 = 1.5e2    # 1.5 * 10^2  = 150.0
float_3 = 1.5e-3   # 1.5 * 10^-3 = 0.0015

print(float_1, float_2, float_3)
````

````python
# COMPLEX literals — written as  real + imaginary j
x = 3.14j
print(x.imag)   # 3.14  -> imaginary part
print(x.real)   # 0.0   -> real part
````

````python
# STRING literals — text inside quotes
strings = "this is python"     # double quotes
string  = 'this is python'     # single quotes (same thing)
char    = "c"                  # a single character is also a string

# Other useful string types:
multiline_str = """this is a
multiline string"""            # triple quotes -> can span multiple lines
raw_str = r"raw \n string"     # r"" -> raw string, so \n is NOT treated as a newline

print(strings)
print(string)
print(char)
print(multiline_str)
print(raw_str)
````

````python
# Python treats Boolean as a number:  True = 1,  False = 0
a = True + 4     # 1 + 4  = 5
b = False + 10   # 0 + 10 = 10
print("a", a)
print("b", b)
````

````python
# Special literal: None
# None means "nothing" / "no value" (the absence of a value).
a = None
print(a)
````
