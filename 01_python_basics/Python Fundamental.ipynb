## 1. Comments 💬

A **comment** is a piece of text that the Python interpreter **ignores** while running the code. Comments are only for humans — to explain *what* the code does.

- We write a comment using the **`#`** symbol.
- A comment can be on its **own line** *(above the code)* or **next to the code** *(code ke bagal me)*.

```python
# This is a full-line comment
print("Hello")   # This is an inline comment (code ke bagal me)
```

## 2. Output — the `print()` Function 🖨️

`print()` is a **function** that displays output on the screen.

> 🔑 **Remember:** Python is a **case-sensitive** language.
> `print()` and `Print()` are **different** — only lowercase `print()` works.

````python
# print() is a function — the () means "function"
# Python is a CASE-SENSITIVE language:
#   print()  -> works
#   Print()  -> error
# So print() and Print() are NOT the same.

print("hello world")
````

````python
print("salman khan")
````

````python
# This line will give an ERROR (intentionally) 👇
print(salman khan)

# Why the error?
# We did NOT put salman khan inside quotes "".
# So Python gets confused — it thinks salman and khan are variables (which don't exist).
# 👉 Inverted commas ("") tell Python that the thing inside is a STRING (text).
# Correct way:  print("salman khan")
````

````python
# We can also print numbers (no quotes needed for numbers)
print(7)
````

````python
# Python's print() is flexible and powerful.
# We can print many things in one print() — Python separates them with a SPACE by default.

print("hello", 1, 2, 3, True)
print("hello world", 1, 2, 3, 4.5, True, False)
````

````python
# Question: why is there a space between the outputs above?
# Answer: print() uses a SPACE as the default separator.
# We can change it using sep="..."

print("hello world", 1, 2, 3, 4.5, True, False, sep="/")
````

````python
# What if we want two print() outputs?
# By default, each print() prints on a NEW line.

print("hello")
print("world")
````

````python
# If we want both prints on the SAME line, use end="..."
# end controls what comes AFTER the print (default is a newline "\n").

print("hello", end="-")
print("world")
````

## 3. Data Types 🧩

A **data type** tells Python *what kind of value* we are working with. Python has many built-in data types:

| # | Data Type | Example | Written with |
|---|-----------|---------|--------------|
| 1 | Integer (`int`) | `8` | plain whole number |
| 2 | Float (`float`) | `8.55` | number with a decimal |
| 3 | Boolean (`bool`) | `True`, `False` | no quotes |
| 4 | String (`str`) | `"hello"` | inside `" "` or `' '` |
| 5 | Complex (`complex`) | `5 + 6j` | real + imaginary `j` |
| 6 | List | `[1, 2, 3]` | square brackets `[ ]` |
| 7 | Tuple | `(1, 2, 3)` | round brackets `( )` |
| 8 | Set | `{1, 2, 3}` | curly braces `{ }` |
| 9 | Dictionary | `{"name": "Shahab"}` | curly braces with `key: value` |

````python
# 1. Integer (int) — whole numbers, positive or negative
print(8)

# 👉 Python integers can be VERY large (practically unlimited size):
print(2 ** 200)   # 2 to the power 200 — still an exact integer!

# Note: 1e308 below is actually a FLOAT (decimal), not an int.
# Floats DO have a limit (~1.8e308). Beyond that, Python shows 'inf' (infinity):
print(1e308)   # 1 * 10^308  -> works
print(1e309)   # too big for a float -> inf
````

````python
# 2. Float — decimal numbers
print(8.55)

# Floats also have a maximum limit (~1.8e308):
print(1.7e308)   # works
print(1.7e309)   # too big -> inf
````

````python
# 3. Boolean — only two values: True or False
# 👉 Remember: do NOT put quotes around them (True, not "True")
print(True)
print(False)
````

````python
# 4. String (str) — text, always inside quotes
print("hello world")
````

````python
# 5. Complex number — has a real part and an imaginary part (j)
print(5 + 6j)
````

````python
# 6. List — written inside SQUARE brackets [ ]
# (In C language, the closest thing is an array.)
print([1, 2, 3, 4, 5])
````

````python
# 7. Tuple — written inside ROUND brackets ( )   ✅  (NOT square brackets!)
print((1, 2, 3, 4, 5))
````

````python
# 8. Set — written inside CURLY braces { }
# (A set automatically removes duplicate values.)
print({1, 2, 3, 4, 5})
````

````python
# 9. Dictionary — key:value pairs inside CURLY braces { }
print({"name": "Shahab", "gender": "Male", "weight": "70"})
````

## 4. Variables 📦

A **variable** is a **container** where we store information (data).

**Why do we use variables?**

As a programmer, when we build software we often **don't know in advance** what value the user will give.

> **Example:** Suppose I'm building an app that shows *"Hello &lt;name&gt;"* on the homepage of whoever logs in. But I don't know who will log in tomorrow — Rahul, Varun, ya koi aur?
>
> Toh is case me main **"Hello"** ke aage ek **variable** rakh dunga. Jo bhi user login karega, uska naam **database se pick karke** us variable me fill kar denge. 🎯

✅ **In short:** variables are containers that hold a value for future use.

### How to create a variable in Python

In **C / C++** you must declare the data type:

```c
int name = "shahab";   // you tell the type (int) yourself
```

In **Python**, you simply write:

```python
name = "shahab"
print(name)
```

👉 Python is **smart enough** to figure out the data type on its own — you don't need to declare it.

````python
a = 5
b = 6
print(a + b)
````

## ⭐ Important Interview Question — Dynamic vs Static Typing

**1. Dynamic Typing**

In C/C++ we mention the data type while creating a variable:

```c
int a = 5;   // we tell the type (int)
```

In Python we simply write:

```python
a = 5        # no type mentioned
```

> **Dynamic typing** = we do **not** tell the data type; the Python interpreter **automatically figures it out**.

📝 **Note:** Python supports **dynamic typing** — the interpreter is intelligent enough to automatically understand the data type of the value given by the user.

**2. Static Typing**

Where we **must** tell the data type while creating a variable — e.g. in **C / C++**.

## 5. Keywords & Identifiers 🔑

### Keywords

**Keywords** are special words that are **reserved** by Python. The interpreter uses them to understand and run your program (e.g. `if`, `else`, `for`, `True`, `import`, ...).

- 👉 We are **not allowed** to use keywords as variable names. For example, using `print` as a variable name would confuse Python.
- In **Python 3 there are 35 keywords** *(in this notebook's version)*. The exact number can change slightly between Python versions — older Python had 33.

**Compiler vs Interpreter (quick note):**
- **Compiler** → translates the whole program at once → generally faster.
- **Interpreter** → translates line by line → a bit slower.
- 📝 Python supports both **single-threading and multi-threading**.

---

### Identifiers

An **identifier** is **any name we give** inside a program — a variable name, a function name, a class name, etc. *(Jo bhi naam hum khud dete hain, woh identifier hai.)*

**Rules for naming identifiers:**

1. ❌ An identifier **cannot start with a digit**.
   ```python
   1name = "shahab"   # error
   name1 = "shahab"   # works
   ```
2. ✅ You **can** use the underscore `_` (but not symbols like `@`, `#`, `$`).
   ```python
   name_ = "shahab"   # works
   _     = "shahab"   # even a single underscore works
   ```
3. ❌ An identifier **cannot be a keyword** (`print`, `if`, `else`, `True`, `False`, ...).

````python
# Rule 1: an identifier cannot START with a digit — this will give an error 👇
1name = "shahab"
print(1name)
````

````python
# But a digit AFTER the first character is fine
name1 = "shahab"
print(name1)
````

````python
# Underscore _ is allowed in identifiers
name_your = "shahab"
print(name_your)
````

````python
# A single underscore _ can even be used as a variable name
# (But symbols like @, #, $ are NOT allowed.)
_ = "shahab"
print(_)
````

## 6. Taking Input from the User ⌨️

The **`input()`** function lets the program take input (text) from the user.

````python
# A basic input() — it waits for the user to type something
input()
````

````python
# We can show a message (prompt) to guide the user
input("enter your name: ")
````

````python
# Taking input from the user and storing it in variables
fnum = input("enter first number: ")
snum = input("enter second number: ")
print(fnum, snum)

result = fnum + snum
print(result)
# ⚠️ The output is NOT what we expect — it joins them (like "23") instead of adding.

# Why? Let's check the types:
print(type(fnum))
print(type(snum))
# 👉 input() ALWAYS returns a STRING by default.
# So "2" + "3" becomes "23" (text joining), not 2 + 3 = 5.
# To do maths, we must convert the input into an integer first (see Type Conversion below).
````

## 7. Type Conversion 🔄

**Type conversion** means converting one data type into another.

> 🔒 **Important:** Python's type conversion does **not** change the original data. It creates a **completely new value** in memory — the original stays exactly the same.

There are **two types** of type conversion:
1. **Implicit** — done automatically by Python.
2. **Explicit** — done manually by the programmer.

````python
# 1. IMPLICIT type conversion
# The interpreter does it AUTOMATICALLY — the programmer doesn't need to do anything.

print(5 + 5.6)              # int + float -> Python automatically makes the result a float
print(type(5), type(5.6))   # <class 'int'> <class 'float'>

# But implicit conversion does NOT always work:
print(4 + "4")              # ERROR: Python can't add an int and a str automatically
# 👉 When Python can't convert automatically, WE must do it -> EXPLICIT conversion.
````

````python
# 2. EXPLICIT type conversion
# The programmer explicitly tells Python to change the data type.
# Built-in conversion functions: int(), str(), float(), ...

# str -> int
print(int("4"), type(int("4")))     # 4 <class 'int'>

# int -> str
print(str(5), type(str(5)))         # '5' <class 'str'>

# -> float
print(float(2.3), type(float(2.3))) # 2.3 <class 'float'>

# ⚠️ Note: a complex number CANNOT be converted to int.
````

````python
# Now let's fix the earlier addition problem using type conversion.
fnum = input("enter first number: ")
snum = input("enter second number: ")
print(fnum, snum)

result = int(fnum) + int(snum)   # convert to int ONLY for the calculation
print(result)

print(type(fnum))   # still <class 'str'> — we changed only 'result', not fnum itself
````

````python
# Even cleaner: convert the input to int right away.
fnum = int(input("enter first number: "))
snum = int(input("enter second number: "))
print(fnum, snum)

result = fnum + snum
print(result)

print(type(fnum))   # now <class 'int'> — we converted the user's input immediately
````

## 8. Literals 🔢

A **literal** is the actual **value** that we store inside a variable.

```python
a = 2
```
Here:
- `a` → **variable**
- `=` → **assignment operator**
- `2` → **literal** (the value)

````python
# Different ways to write INTEGER literals:
a = 0b1010   # binary literal       (base 2)
b = 100      # decimal literal      (base 10)
c = 0o310    # octal literal        (base 8)
d = 0x12c    # hexadecimal literal  (base 16)
````

````python
# Binary literal -> Python prints its decimal value
a = 0b1010
print(a)   # 10
````

````python
# Decimal literal
b = 100
print(b)   # 100
````

````python
# Octal literal
c = 0o310
print(c)   # 200
````

````python
# Hexadecimal literal
d = 0x12c
print(d)   # 300
````

````python
# FLOAT literals
float_1 = 10.5
float_2 = 1.5e2    # 1.5 * 10^2  = 150.0
float_3 = 1.5e-3   # 1.5 * 10^-3 = 0.0015

print(float_1, float_2, float_3)
````

````python
# COMPLEX literals — written as  real + imaginary j
x = 3.14j
print(x.imag)   # 3.14  -> imaginary part
print(x.real)   # 0.0   -> real part
````

````python
# STRING literals — text inside quotes
strings = "this is python"     # double quotes
string  = 'this is python'     # single quotes (same thing)
char    = "c"                  # a single character is also a string

# Other useful string types:
multiline_str = """this is a
multiline string"""            # triple quotes -> can span multiple lines
raw_str = r"raw \n string"     # r"" -> raw string, so \n is NOT treated as a newline

print(strings)
print(string)
print(char)
print(multiline_str)
print(raw_str)
````

````python
# Python treats Boolean as a number:  True = 1,  False = 0
a = True + 4     # 1 + 4  = 5
b = False + 10   # 0 + 10 = 10
print("a", a)
print("b", b)
````

````python
# Special literal: None
# None means "nothing" / "no value" (the absence of a value).
a = None
print(a)
````
