## 1. Operators in Python

Operators wo symbols hote hai jo values (operands) par koi **operation** perform karte hai. Python me 6 main categories hai:

1. **Arithmetic** — `+  -  *  /  //  %  **`
2. **Relational (Comparison)** — `>  <  >=  <=  ==  !=`
3. **Logical** — `and  or  not`
4. **Bitwise** — `&  |  ^  ~  <<  >>`
5. **Assignment** — `=  +=  -=  *=  /=  %=` ...
6. **Membership** — `in  not in`

````python
# 1:- ARITHMETIC OPERATORS
print(5 + 5)   # Addition        -> 10
print(6 - 5)   # Subtraction     -> 1
print(5 * 5)   # Multiplication  -> 25
print(5 / 2)   # Division        -> 2.5  (hamesha float deta hai)
print(5 // 2)  # Floor division  -> 2    (quotient, decimal hata deta hai)
print(5 % 2)   # Modulus         -> 1    (remainder after dividing)
print(5 ** 2)  # Power           -> 25   (5 ka square)
````

````python
# 2:- RELATIONAL (COMPARISON) OPERATORS
# Ye hamesha True / False (Boolean) return karte hai.
print(4 > 5)    # False
print(4 < 5)    # True
print(4 <= 5)   # True
print(4 >= 5)   # False
print(4 == 4)   # True   (equal to)
print(4 != 4)   # False  (not equal to)
````

````python
# 3:- LOGICAL OPERATORS
# -> and : dono conditions True ho tabhi True. Ek bhi False -> False.
# -> or  : koi ek bhi True ho toh True. Dono False ho tabhi False.
# -> not : value ko ulta kar deta hai (True<->False).
print(1 and 0)   # 0      (ek operand 0/False hai)
print(1 or 0)    # 1      (ek operand True hai)
print(0 or 1)    # 1
print(not 1)     # False  (dhyan do: not 0 nahi, poora False deta hai)
````

````python
# 4:- BITWISE OPERATORS
# Ye operators number ki BINARY (0/1 bits) form par kaam karte hai.
print(2 & 3)    # bitwise AND   -> 2    (10 & 11 = 10)
print(2 | 3)    # bitwise OR    -> 3    (10 | 11 = 11)
print(2 ^ 3)    # bitwise XOR   -> 1    (bits ALAG ho -> 1, SAME ho -> 0)
print(2 ^ 2)    # bitwise XOR   -> 0    (dono bits same hai)
print(~3)       # bitwise NOT   -> -4   (formula: ~x = -(x + 1))
print(5 << 2)   # LEFT shift    -> 20   (bits left move; har shift = value * 2)
print(4 >> 2)   # RIGHT shift   -> 1    (bits right move; har shift = value // 2)
````

````python
# 5:- ASSIGNMENT OPERATORS
# '=' basic assignment hai, jaise a = 2.
# Baaki compound operators calculation + assignment ek saath karte hai.
a = 2
a += 2   # same as a = a + 2  -> 4
print(a)
a -= 2   # a = a - 2          -> 2
print(a)
a *= 2   # a = a * 2          -> 4
print(a)
a %= 2   # a = a % 2          -> 0
print(a)
````

````python
# 6:- MEMBERSHIP OPERATORS
# 'in' / 'not in' check karte hai ki koi value sequence ke andar hai ya nahi.
print("d" in "delhi")       # True
print("d" not in "delhi")   # False
# Ye har sequence/collection par chalte hai (string, list, tuple, set, dict).
print(1 in [2, 3, 4, 5])    # False
````

### 📝 Practice — Sum of digits of a 3-digit number

Ek 3-digit number ke teeno digits ka sum nikalna hai. Do tarike dekhte hai: **maths (modulus + floor division)** wala aur **loop** wala.

````python
# Way 1: Maths ka use (% se last digit, // se digit hatana)
number = int(input("enter the 3 digit number: "))
# Example: 345 % 10 -> 5
a = number % 10
number = number // 10
# 34 % 10 -> 4
b = number % 10
number = number // 10
# 3 % 10 -> 3
c = number % 10
print("sum of digits:", a + b + c)
````

````python
# Way 2: Loop ka use (string ke har character par ghoomo)
number = input("enter the 3 digit number: ")
digit_sum = 0                 # 'sum' Python ka built-in hai, isliye alag naam
for ch in number:
    digit_sum = digit_sum + int(ch)
print("sum of digits:", digit_sum)
````

## 2. Conditional Statements — `if` / `elif` / `else`

Program ko **decision** lena sikhata hai: agar condition `True` hai toh ye karo, warna wo karo.

```python
if condition1:
    # condition1 True hone par
elif condition2:
    # condition1 False, condition2 True hone par
else:
    # koi bhi condition True nahi hui toh
```

> ⚠️ **Indentation (spacing) bahut important hai.** Python me `if`/`else` ke andar ka code 4 spaces andar hota hai. Galat indentation = error ya galat logic.

````python
# Login program (nested if ka example)
# Correct email -> shahabazmi.com , correct password -> 1234
email = input("enter the email id: ")
password = input("enter the password: ")

if email == "shahabazmi.com" and password == "1234":
    print("welcome")
elif email == "shahabazmi.com" and password != "1234":
    print("incorrect password, try again")
    password = input("enter the correct password: ")
    if password == "1234":
        print("welcome, finally")
    else:
        print("beta tumse na ho payega")
else:
    print("email is not correct")
````

## 4. Loops in Python

Loop ek aisa code block hai jise hum **baar-baar (again and again)** chala sakte hai. Real apps (Flipkart, Amazon, Zomato, Ola...) me har product/order list par loop hi chalta hai.

Do tarah ke loops:
1. **`while` loop** — jab tak condition `True` hai tab tak chalta hai.
2. **`for` loop** — kisi sequence (range, string, list...) ke har element par chalta hai.

### While Loop

`while` tab use karte hai jab hamare paas ek **condition** ho ki loop kab rukna chahiye.

**Example — multiplication table print karna:**

````python
# Table of a number using while loop
number = int(input("enter the number: "))
i = 1
while i < 11:
    print(number, "*", i, "=", number * i)
    i += 1
````

````python
# WHILE with ELSE
# 'else' tab chalta hai jab loop ki condition False ho jaati hai (normally khatam).
x = 1
while x < 3:
    print(x)
    x += 1
else:
    print("limit cross")
````

````python
# Number guessing game (while-else ka acha example)
# 1 se 100 ke beech ek random number generate hoga, use guess karna hai.
import random
jackpot = random.randint(1, 100)
guess = int(input("guess karo (1-100): "))
counter = 1
while guess != jackpot:
    if guess < jackpot:
        print("galat! thoda higher try karo")
    else:
        print("galat! thoda lower try karo")
    guess = int(input("guess karo: "))
    counter += 1        # har attempt par badhega (pehle sirf ek branch me tha)
else:
    print("correct guess! 🎉")
    print("total attempts:", counter)
````

### For Loop

`for` loop kisi bhi sequence (range, string, list...) ke **har element** par iterate karta hai.

`range(start, stop, step)` ek series banata hai — `stop` **exclusive** hota hai (usse ek kam tak jaata hai).

````python
for i in range(1, 11):   # 1 se 10 tak
    print(i)
````

````python
for i in range(1, 10, 2):   # step 2 -> 1, 3, 5, 7, 9
    print(i)
````

````python
for i in range(10, 0, -1):   # ulta -> 10 se 1 tak
    print(i)
````

````python
for ch in "delhi":   # string ke har character par
    print(ch)
````

### 📝 Practice — Sequence sum

Series ka sum nikalo:  `1/1! + 2/2! + 3/3! + ...` (n terms tak).

````python
# 1/1! + 2/2! + 3/3! + ......
n = int(input("enter n: "))
result = 0
fact = 1
for i in range(1, n + 1):
    fact = fact * i          # i ka factorial (pichle fact ko reuse karke)
    result = result + i / fact
print(result)
````

## 5. Nested Loops

Ek loop ke andar doosra loop = **nested loop**. Outer loop ka har ek chakkar poore inner loop ko chalata hai. Ye patterns aur grids ke liye bahut kaam aata hai.

````python
# Nested loop example -> har i ke liye poora j chalega
for i in range(1, 5):
    for j in range(1, 5):
        print(i, j)
````

## 6. Pattern Programs

Patterns nested loops ki best practice hai. Idea simple hai:
- **Outer loop** -> kitni rows (lines).
- **Inner loop(s)** -> ek row me kya print hoga (stars, spaces, numbers).

Chaliye ek-ek karke banate hai. 👇

**Right-angled triangle:**
```
*
* *
* * *
```

````python
# Simple right triangle (stars inner loop se)
rows = int(input("enter number of rows: "))
for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print("* ", end="")
    print()   # row khatam -> agli line par jao
````

````python
# Wahi triangle, string multiplication ka short-cut use karke
rows = int(input("enter the rows: "))
for i in range(1, rows + 1):
    print("*" * i)
````

### Pattern 1.

````python
# Pyramid -> pehle spaces, phir stars
rows = int(input("enter the number: "))
for i in range(1, rows + 1):
    print(" " * (rows - i) + "*" * (i * 2 - 1))
````

### Pattern 2

````python
# Inverted pyramid -> rows ulta chalao
rows = int(input("enter the rows: "))
for i in range(rows, 0, -1):
    print(" " * (rows - i) + "*" * (i * 2 - 1))
````

### Pattern 3
3.jpeg

````python
# Diamond -> upar pyramid + neeche inverted pyramid
rows = int(input("enter the rows: "))
# upper part
for i in range(1, rows + 1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
# lower part
for i in range(rows - 1, 0, -1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
````

### Pattern 4 — Number pyramid
```
1
121
12321
1234321
```

````python
rows = int(input("enter rows: "))
for i in range(1, rows + 1):
    for j in range(1, i + 1):     # 1 se i tak (badhte hue)
        print(j, end="")
    for k in range(i - 1, 0, -1): # i-1 se 1 tak (ghatte hue)
        print(k, end="")
    print()
````

### Pattern 5
1.jpeg

````python
# Hollow square -> pehli/aakhri row full, beech me sirf kinare
rows = 5
for i in range(rows):
    if i == 0 or i == rows - 1:
        print("*" * rows)                    # top / bottom border
    else:
        print("*" + " " * (rows - 2) + "*")   # side borders only
````

### Pattern 6
2.jpeg

````python
# Butterfly pattern
rows = 5
# upper part
for i in range(1, rows + 1):
    print("*" * i + " " * (2 * (rows - i)) + "*" * i)
# lower part
for i in range(rows, 0, -1):
    print("*" * i + " " * (2 * (rows - i)) + "*" * i)
````

## 7. Loop Control Statements — `break`, `continue`, `pass`

Ye teen keywords loop ke normal flow ko control karte hai:

| Keyword | Kaam |
|---------|------|
| **`break`** | Loop ko turant **rok** deta hai aur bahar nikal jaata hai. |
| **`continue`** | Current chakkar ko **skip** karke seedha agle iteration par chala jaata hai. |
| **`pass`** | Kuch **nahi** karta — placeholder hai jahan syntactically code chahiye but abhi logic nahi likhna. |

````python
# BREAK
# Jab koi condition meet ho jaye toh loop ko wahin rok do (jaise searching me mil gaya).
for i in range(1, 10):
    if i == 5:
        break
    print(i)   # 1 2 3 4 print hoga, 5 par ruk jaayega
````

````python
# BREAK ka real use -> prime numbers nikalna (for-else ke saath)
lower = int(input("enter lower range: "))
upper = int(input("enter upper range: "))
for i in range(lower, upper + 1):
    for j in range(2, i):
        if i % j == 0:   # divisor mila -> prime nahi hai
            break
    else:                # inner loop bina break ke poora chala -> prime hai
        print(i)
````

````python
# CONTINUE
# Specific condition par current iteration skip karo, loop chalta rahe.
for i in range(1, 10):
    if i == 5:
        continue     # 5 skip ho jaayega
    print(i)
````

````python
# PASS
# Placeholder hai -> "abhi kuch mat karo". Empty block par error se bacha leta hai.
for i in range(1, 10):
    pass   # yahan baad me logic add kar sakte hai
print("loop chal gaya, koi error nahi aaya")
````
