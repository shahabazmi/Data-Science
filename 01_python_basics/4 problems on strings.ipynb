## Question 1 — Find the length of a string *without* using `len()`

```python
# Approach: loop over every character and count them one by one.
s = input("Enter the string: ")
counter = 0
for ch in s:
    counter += 1          # har character par counter badhao
print("Length of the string is", counter)
```

## Question 2 — Extract the username from an email
Example: for `shahabazmi321@gmail.com`, the username is **`shahabazmi321`**
(everything before the `@`).

```python
# The username is the part before '@'. Find the '@' position, then slice up to it.
s = input("Enter the email: ")
pos = s.index("@")        # position of '@'
print(s[0:pos])           # everything before '@'
```

## Question 3 — Count how often a character appears in a string
Example: in `"hello how are you"`, the character `h` appears **2** times.

```python
# Approach: loop through the string; whenever a character matches, count it.
s = input("Enter the string: ")
term = input("Which character to search for: ")
counter = 0
for ch in s:
    if ch == term:
        counter += 1
print("Frequency:", counter)
```

## Question 4 — Remove a particular character from a string

```python
# Approach: build a new string, skipping the character we want to remove.
s = input("Enter the string: ")
term = input("Which character to remove: ")
result = ""
for ch in s:
    if ch != term:
        result = result + ch      # keep everything except 'term'
print(result)
```

## Question 5 — Check whether a string is a palindrome
A **palindrome** reads the same forwards and backwards (e.g. *malayalam*, *madam*).

**Method 1 — compare characters from both ends.**

```python
# Compare first char with last, second with second-last, and so on.
s = input("Enter the string: ")
flag = True                        # assume palindrome until proven otherwise
for i in range(0, len(s) // 2):
    if s[i] != s[len(s) - i - 1]:  # mismatch -> not a palindrome
        flag = False
        break
if flag:
    print("palindrome")
else:
    print("not a palindrome")
```

**Method 2 — reverse the string with slicing (much shorter).**

```python
# s[::-1] reverses the string; if it equals the original, it's a palindrome.
s = input("Enter a string: ")
if s == s[::-1]:
    print("palindrome")
else:
    print("not a palindrome")
```

### Palindrome check using a function

```python
# Wrap the logic in a function so it can be reused with any string.
def isPalindrome(string):
    if string == string[::-1]:
        return "The string is a palindrome."
    else:
        return "The string is not a palindrome."

string = input("Enter string: ")
print(isPalindrome(string))
```

```python
# A version that RETURNS a boolean (True/False) instead of a message.
def isPalindrome(s):
    return s == s[::-1]

# Driver code (no input needed)
s = "malayalam"
ans = isPalindrome(s)
if ans:
    print("Yes")
else:
    print("No")
```

## Question 6 — Count the number of words *without* using `split()`

```python
# Approach: collect characters into 'temp'. On every space, the word is finished ->
# push it into the list and reset 'temp'. Append the last word after the loop ends.
s = input("Enter the string: ")
words = []
temp = ""
for ch in s:
    if ch != " ":
        temp = temp + ch
    else:
        words.append(temp)
        temp = ""
words.append(temp)                 # last word (loop ke baad bacha hua)
print(words)
print("Number of words:", len(words))
```

## Question 7 — Convert a string to title case *without* using `title()`

```python
# For each word: make the first letter uppercase and the rest lowercase.
s = input("Enter the string: ")
words = []
for word in s.split():
    words.append(word[0].upper() + word[1:].lower())
print(" ".join(words))
```

## Question 8 — Convert an integer to a string (manually)

```python
# Idea: pick digits from the right using '% 10', map each digit to its character,
# and build the result. '// 10' removes the last digit each time.
number = int(input("Enter the number: "))
digits = "0123456789"
result = ""
while number != 0:
    result = digits[number % 10] + result   # prepend the new digit
    number = number // 10
print(result)
print(type(result))          # <class 'str'> -> it is now a string
```
