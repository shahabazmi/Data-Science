## Question 1 — Length nikalo, bina `len()` ke

Ek string ki length find karni hai, but `len()` use nahi karna.

Idea: har character par ek loop chalao aur ek counter ko har baar `+1` karte jao.

```python
# Approach: loop over every character and count them one by one.
s = input("Enter the string: ")
counter = 0
for ch in s:
    counter += 1          # har character par counter badhao
print("Length of the string is", counter)
```

## Question 2 — Email se username nikalo

Example: `shahabazmi321@gmail.com` ka username hoga `shahabazmi321` (yaani jo bhi `@` se pehle hai).

Idea: `@` ki position find karo, phir uske pehle wala part slice kar lo.

```python
# The username is the part before '@'. Find the '@' position, then slice up to it.
s = input("Enter the email: ")
pos = s.index("@")        # position of '@'
print(s[0:pos])           # everything before '@'
```

## Question 3 — Ek character ki frequency count karo

Ek particular character string me kitni baar aaya, wo count karo. Example: `"hello how are you"`
me `h` **2** baar aata hai.

Idea: loop chalao, aur jab bhi character match ho to counter badha do.

```python
# Approach: loop through the string; whenever a character matches, count it.
s = input("Enter the string: ")
term = input("Which character to search for: ")
counter = 0
for ch in s:
    if ch == term:
        counter += 1
print("Frequency:", counter)
```

## Question 4 — Ek particular character remove karo

String me se koi ek character hataana hai.

Idea: ek nayi string banao, aur us character ko chhodkar baaki sab characters jodte jao.

```python
# Approach: build a new string, skipping the character we want to remove.
s = input("Enter the string: ")
term = input("Which character to remove: ")
result = ""
for ch in s:
    if ch != term:
        result = result + ch      # keep everything except 'term'
print(result)
```

## Question 5 — Check karo string palindrome hai ya nahi

Ek **palindrome** wo string hoti hai jo aage se aur peeche se same padhi jaaye (jaise
*malayalam*, *madam*).

**Method 1 — dono ends se characters compare karke.**

Idea: pehla character last se compare karo, doosra second-last se, aur aise hi aage. Kahin
mismatch mila to samajh lo palindrome nahi hai.

```python
# Compare first char with last, second with second-last, and so on.
s = input("Enter the string: ")
flag = True                        # assume palindrome until proven otherwise
for i in range(0, len(s) // 2):
    if s[i] != s[len(s) - i - 1]:  # mismatch -> not a palindrome
        flag = False
        break
if flag:
    print("palindrome")
else:
    print("not a palindrome")
```

**Method 2 — slicing se string ko ulta karke (bahut chhota tarika).**

`s[::-1]` string ko reverse kar deta hai. Agar reversed string original ke barabar hai, to
wo palindrome hai.

```python
# s[::-1] reverses the string; if it equals the original, it's a palindrome.
s = input("Enter a string: ")
if s == s[::-1]:
    print("palindrome")
else:
    print("not a palindrome")
```

### Palindrome check — function ki madad se

Upar wale logic ko ek **function** me daal dete hai, taaki kisi bhi string par dobara aasani
se use kar sake.

```python
# Wrap the logic in a function so it can be reused with any string.
def isPalindrome(string):
    if string == string[::-1]:
        return "The string is a palindrome."
    else:
        return "The string is not a palindrome."

string = input("Enter string: ")
print(isPalindrome(string))
```

```python
# A version that RETURNS a boolean (True/False) instead of a message.
def isPalindrome(s):
    return s == s[::-1]

# Driver code (no input needed)
s = "malayalam"
ans = isPalindrome(s)
if ans:
    print("Yes")
else:
    print("No")
```

## Question 6 — Words count karo, bina `split()` ke

Ek sentence me kitne words hai, wo count karna hai — par `split()` use nahi karna.

Idea: characters ko ek `temp` string me jodte jao. Jab space aaye to samajh lo ek word poora
hua — use list me daal do aur `temp` reset kar do. Loop khatam hone ke baad aakhri word bhi
list me daal do.

```python
# Approach: collect characters into 'temp'. On every space, the word is finished ->
# push it into the list and reset 'temp'. Append the last word after the loop ends.
s = input("Enter the string: ")
words = []
temp = ""
for ch in s:
    if ch != " ":
        temp = temp + ch
    else:
        words.append(temp)
        temp = ""
words.append(temp)                 # last word (loop ke baad bacha hua)
print(words)
print("Number of words:", len(words))
```

## Question 7 — Title case banao, bina `title()` ke

Har word ka pehla letter capital karna hai — par `title()` use nahi karna.

Idea: har word ka pehla character `upper()` karo aur baaki `lower()`, phir sab jod do.

```python
# For each word: make the first letter uppercase and the rest lowercase.
s = input("Enter the string: ")
words = []
for word in s.split():
    words.append(word[0].upper() + word[1:].lower())
print(" ".join(words))
```

## Question 8 — Integer ko string me convert karo (manually)

Ek number ko string me convert karna hai, bina `str()` use kiye.

Idea: `% 10` se aakhri digit nikalo, use uske character se map karo, aur `// 10` se wo digit
hata do — is tarah right se left string banti jaayegi.

```python
# Idea: pick digits from the right using '% 10', map each digit to its character,
# and build the result. '// 10' removes the last digit each time.
number = int(input("Enter the number: "))
digits = "0123456789"
result = ""
while number != 0:
    result = digits[number % 10] + result   # prepend the new digit
    number = number // 10
print(result)
print(type(result))          # <class 'str'> -> it is now a string
```
