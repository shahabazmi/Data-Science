## 1. Printing & Comments

```python
# This is a comment - Python ignores it
print("Hello, Data Science!")  # print() displays output
print("Python", "is", "fun")   # multiple values separated by commas
```

## 2. Variables

A variable stores a value. You don't need to declare the type — Python figures it out.

```python
name = "Shahab"
age = 22
height = 5.9
is_student = True

print(name, age, height, is_student)
```

## 3. Data Types

Use `type()` to check the type of any value.

```python
print(type("hello"))   # str  -> text
print(type(10))         # int  -> whole number
print(type(3.14))       # float -> decimal
print(type(True))       # bool -> True/False
print(type([1, 2, 3]))  # list
```

## 4. Operators

```python
a, b = 10, 3

print("Add:", a + b)
print("Subtract:", a - b)
print("Multiply:", a * b)
print("Divide:", a / b)
print("Floor divide:", a // b)
print("Modulus (remainder):", a % b)
print("Power:", a ** b)

# Comparison operators return True/False
print("a > b ?", a > b)
print("a == b ?", a == b)
```

## 5. Strings

Strings are sequences of characters. They support indexing, slicing, and many methods.

```python
text = "Data Science"

print(text.upper())       # DATA SCIENCE
print(text.lower())       # data science
print(len(text))          # length
print(text[0])            # first character
print(text[-1])           # last character
print(text[0:4])          # slicing -> 'Data'
print(text.replace("Data", "Computer"))

# f-strings: the modern way to format text
name = "Shahab"
print(f"Hello {name}, welcome to {text}!")
```

## 6. Collections: Lists, Tuples, Sets, Dictionaries

```python
# LIST - ordered, changeable, allows duplicates
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")     # add an item
fruits[0] = "mango"          # change an item
print("List:", fruits)
print("First fruit:", fruits[0])
```

```python
# TUPLE - ordered but UNchangeable (immutable)
coordinates = (10.5, 20.3)
print("Tuple:", coordinates)

# SET - unordered, no duplicates
numbers = {1, 2, 2, 3, 3, 3}
print("Set (duplicates removed):", numbers)
```

```python
# DICTIONARY - key:value pairs
student = {
    "name": "Shahab",
    "age": 22,
    "course": "Data Science"
}

print("Name:", student["name"])
student["age"] = 23          # update a value
student["city"] = "Delhi"    # add a new key
print("Updated dict:", student)
print("Keys:", student.keys())
print("Values:", student.values())
```

## 7. Conditional Statements (if / elif / else)

```python
marks = 75

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: F")
```

## 8. Loops

```python
# FOR loop - iterate over a sequence
for fruit in ["apple", "banana", "cherry"]:
    print("I like", fruit)

print("---")

# range() generates numbers
for i in range(1, 6):
    print("Number:", i)
```

```python
# WHILE loop - repeats while a condition is True
count = 1
while count <= 5:
    print("Count is", count)
    count += 1   # don't forget to update, or it loops forever!
```

## 9. Functions

Functions let you reuse a block of code. Define with `def`, call by name.

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Shahab"))
print(greet("World"))


def add(a, b=10):     # b has a default value
    return a + b

print("add(5, 3) =", add(5, 3))
print("add(5)    =", add(5))   # uses default b=10
```
