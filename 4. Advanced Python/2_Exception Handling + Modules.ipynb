## 1️⃣ Two Stages Where Errors Happen

| Stage | When | Error type |
|-------|------|------------|
| **While writing the program** (compilation) | before it runs | **Syntax Error** |
| **While running the program** (execution) | at runtime | **Exception** |

## 2️⃣ Syntax Errors

A **syntax error** means your code breaks the **grammar rules** of Python. The interpreter refuses to run it until you fix it.

**Common causes:**
- Missing symbols (colon `:`, brackets)
- Misspelling a keyword
- Wrong indentation
- An empty `if` / loop / class / function

These cells are **wrong on purpose** to show each mistake.

````python
# Old Python 2 style — missing parentheses
print "hello world"   # ❌ SyntaxError: Missing parentheses in call to 'print'.
````

````python
# Missing colon after the condition
a = 5
if a == 5          # ❌ SyntaxError: expected ':'
    print("hello")
````

````python
# Misspelled keyword ("iff" instead of "if")
a = 5
iff a == 3:        # ❌ SyntaxError: invalid syntax
    print("hello")
````

````python
# Wrong indentation — the line after 'if' must be indented
a = 5
if a == 5:
print(a)           # ❌ IndentationError: expected an indented block
````

## 3️⃣ Common Exceptions

An **exception** happens while the program is *running*. Here are the most common ones you'll meet. Each cell triggers one on purpose.

| Exception | Happens when... |
|-----------|-----------------|
| `IndexError` | you access a list index that doesn't exist |
| `ModuleNotFoundError` | you import a module that isn't installed |
| `KeyError` | you access a dictionary key that doesn't exist |
| `TypeError` | you use an operator on the wrong data types |
| `ValueError` | you pass a value of the wrong kind to a function |
| `NameError` | you use a variable that was never defined |
| `AttributeError` | you call a method that doesn't exist on that object |

````python
# IndexError — the list has 3 items, index 100 doesn't exist
l = [1, 2, 3]
l[100]   # ❌ IndexError: list index out of range
````

````python
# ModuleNotFoundError — "mati" is a typo, there's no such module
import mati   # ❌ ModuleNotFoundError: No module named 'mati' 
````

````python
# KeyError — the key "age" is not in the dictionary
d = {"name": "shahab"}
d["age"]   # ❌ KeyError: 'age' 
````

````python
# TypeError — you can't add a number and a string
1 + "a"   # ❌ TypeError: unsupported operand type(s) for +: 'int' and 'str' 
````

````python
# ValueError — "a" cannot be turned into an integer
int("a")   # ❌ ValueError: invalid literal for int() with base 10: 'a' 
````

````python
# NameError — 'k' was never defined
print(k)   # ❌ NameError: name 'k' is not defined
````

````python
# AttributeError — lists don't have an upper() method (only strings do)
l = [1, 2, 3]
l.upper()   # ❌ AttributeError: 'list' object has no attribute 'upper' 
````

## 4️⃣ What Are Exceptions & Why Handle Them?

An **exception** is an error the compiler **couldn't** catch — your syntax is perfect, but something goes wrong *while running* (a logical or unexpected situation).

**Examples:** dividing by zero, a database failing, running out of memory.

### 🤔 Why is it so important to handle exceptions?

**1. User experience.** When an unhandled error occurs, Python prints a scary, technical message called a **stack trace** (error type, line number, file name). That's useful for *you* the programmer — but showing it to a normal *user* looks broken and confusing. Instead, we want to show a friendly message.

**2. Security.** If a website shows its raw error, a **hacker** can learn your code, file names, and server details — a serious security leak. Handling exceptions hides that information.

## 5️⃣ `try` / `except` — Handling an Exception

The idea is simple:
- Put the **risky code** inside a **`try`** block.
- Put the **backup plan** (what to do if it fails) inside an **`except`** block.

> 💡 **When to use it:** any time you deal with an **external resource** — opening a file, connecting to a database or the internet, etc. That code is *risky*, so wrap it in `try`/`except`.

First, let's make a file to work with.

````python
# create a file we can use in the examples
with open("sample.txt", "w") as f:
    f.write("hello world")

with open("sample.txt", "r") as f:
    print(f.read())
````

Now imagine the file gets renamed by mistake, so we try to open a file that doesn't exist. Without `try`/`except`, the user sees a scary crash. **With** it, we show a friendly message.

````python
try:
    # oops — trying to open a file that doesn't exist
    with open("sample1.txt", "r") as f:
        print(f.read())
except:
    print("sorry, file not found")   # friendly message instead of a crash
````

## 6️⃣ Catching *Specific* Exceptions

A bare `except:` catches everything, but it can't tell you *which* error happened. Better: write a **separate `except` block for each type** of error, so you can respond to each one properly.

### First, how do we know which error we got?
Catch it as a variable and print `type(e)` — it shows the exact error class.

> 🛠️ **Improved from the original:** the original used `e.with_traceback` (which just prints a method's address). I replaced it with `type(e)`, which cleanly shows the error type.

````python
# See which error type occurred
try:
    f = open("sample.txt", "r")
    print(f.read())
    print(m)              # 'm' is not defined -> NameError
except Exception as e:
    print(type(e))        # shows the exact error class
    print(e)              # shows the error message
````

### One `except` block per error type

Now we handle each error differently. **Order matters** — put a general `except Exception` **last**, as a catch-all for anything you didn't specifically handle.

````python
# sample.txt exists, but 'm' is undefined -> the NameError block runs
try:
    f = open("sample.txt", "r")
    print(f.read())
    print(m)
except FileNotFoundError:
    print("file not found")
except NameError:
    print("variable not found")
````

````python
# now the file is missing -> the FileNotFoundError block runs
try:
    f = open("sample1.txt", "r")
    print(f.read())
    print(m)
except FileNotFoundError:
    print("file not found")
except NameError:
    print("variable not found")
````

````python
# adding more specific handlers, e.g. dividing by zero
try:
    m = 5
    f = open("sample.txt", "r")
    print(f.read())
    print(m)
    print(5 / 0)          # ZeroDivisionError
except FileNotFoundError:
    print("file not found")
except NameError:
    print("variable not found")
except ZeroDivisionError:
    print("can't divide by zero")
````

````python
# a generic 'except Exception' at the END catches anything else (here, IndexError)
try:
    m = 5
    f = open("sample.txt", "r")
    print(f.read())
    print(m)
    l = [1, 2, 3, 4]
    l[100]                # IndexError — not handled above, so the generic block catches it
except FileNotFoundError:
    print("file not found")
except NameError:
    print("variable not found")
except ZeroDivisionError:
    print("can't divide by zero")
except Exception as e:
    print(e)
````

## 7️⃣ `else` and `finally`

Two more blocks complete the picture:
- **`else`** → runs **only if `try` succeeded** (no error happened).
- **`finally`** → runs **always**, whether there was an error or not.

```
        try
       /    \
   (error)  (no error)
    except    else
       \      /
        finally   ← always runs
```

`finally` is perfect for **cleanup** — closing a file or a database connection no matter what happens.

````python
# No error -> the 'else' block runs
try:
    f = open("sample.txt", "r")
except FileNotFoundError:
    print("file not found")
else:
    print(f.read())      # runs only because try succeeded
````

````python
# 'finally' runs in BOTH cases
try:
    f = open("sample.txt", "r")
except FileNotFoundError:
    print("file not found")
else:
    print(f.read())
finally:
    print("this always runs")
````

````python
# even when there's an error, 'finally' still runs
try:
    f = open("sample1.txt", "r")   # missing file
except FileNotFoundError:
    print("file not found")
else:
    print(f.read())
finally:
    print("this always runs")
````

## 8️⃣ `raise` — Throwing Your Own Exception

Sometimes *you* want to signal an error on purpose — for example, if a user tries to withdraw a negative amount. The **`raise`** keyword lets you throw an exception yourself, with a custom message.

```
Python  ->  try / except / raise
Java    ->  try / catch  / throw   (same idea, different words)
```

````python
raise NameError("just testing")   # ❌ NameError: just testing
````

````python
raise FileNotFoundError("just testing")   # ❌ FileNotFoundError: just testing
````

### Why is `raise` useful? — a Bank example

We put validation inside `withdraw()`: raise an exception if the amount is negative or more than the balance. The caller wraps it in `try`/`except` to handle it gracefully.

````python
class Bank:
    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount < 0:
            raise Exception("amount cannot be negative")
        if self.balance < amount:
            raise Exception("insufficient balance")
        self.balance = self.balance - amount

# A valid withdrawal
obj = Bank(10000)
try:
    obj.withdraw(5000)
except Exception as e:
    print(e)
else:
    print(obj.balance)
````

````python
# A negative amount -> our raised exception is caught
obj = Bank(10000)
try:
    obj.withdraw(-5000)
except Exception as e:
    print(e)
else:
    print(obj.balance)
````

````python
# Withdrawing more than the balance
obj = Bank(10000)
try:
    obj.withdraw(15000)
except Exception as e:
    print(e)
else:
    print(obj.balance)
````

> 💡 **Behind the scenes:** every exception (`Exception`, `NameError`, ...) is actually a **class**. When you `raise Exception("...")`, you create an **object** of that class, and `except Exception as e` catches that same object as `e`.

## 9️⃣ Custom Exception Classes

You can create your **own** exception type by making a class that **inherits from `Exception`**. (Inheriting from `Exception` is required — that's what lets you `raise` it.)

````python
# A custom exception (must inherit from Exception)
class MyException(Exception):
    def __init__(self, message):
        print(message)

class Bank:
    def __init__(self, balance):
        self.balance = balance
    def withdraw(self, amount):
        if amount < 0:
            raise MyException("amount cannot be negative")
        if self.balance < amount:
            raise MyException("insufficient balance")
        self.balance = self.balance - amount

obj = Bank(10000)
try:
    obj.withdraw(5000)
except MyException as e:
    pass
else:
    print(obj.balance)
````

### Why bother with a custom exception? — a real scenario

Imagine you build **Google's** login system. At registration you save the user's **device signature** (which device they used). On login, you check not just email + password, but also **whether it's the same device**. If someone logs in from a **new device**, that could be a hacker — so you raise a **`SecurityError`** and log them out everywhere.

A custom exception lets you attach **extra behaviour** (like a `logout()` method) to the error itself.

> 🛠️ **Fixed from the original:** `obj = google(...)` used a lowercase `g` (would crash — Python is case-sensitive); corrected to `Google(...)`. Also cleaned the informal messages.

````python
# a custom exception that can also log the user out
class SecurityError(Exception):
    def __init__(self, message):
        print(message)
    def logout(self):
        print("logging out from all devices")

class Google:
    def __init__(self, name, email, password, device):
        self.name = name
        self.email = email
        self.password = password
        self.device = device      # the device saved at registration

    def login(self, email, password, device):
        # security check first: is it the same device?
        if device != self.device:
            raise SecurityError("login from an unrecognized device!")
        if email == self.email and password == self.password:
            print("welcome")
        else:
            print("login error")

# register from an "android" device
obj = Google("shahab", "azmi@gmail.com", "1234", "android")

# login from the SAME device -> all good
try:
    obj.login("azmi@gmail.com", "1234", "android")
except SecurityError as e:
    e.logout()
else:
    print(obj.name)
finally:
    print("database connection closed")
````

````python
# now login from a DIFFERENT device -> SecurityError is raised and logout() runs
obj = Google("shahab", "azmi@gmail.com", "1234", "android")
try:
    obj.login("azmi@gmail.com", "1234", "iphone")   # different device!
except SecurityError as e:
    e.logout()
else:
    print(obj.name)
finally:
    print("database connection closed")
````

### What is a Module?

A **module** is simply a **`.py` file full of code** (functions, classes, variables) that you can reuse. Python comes with many ready-made modules (the **standard library**), and you can make your own.

**Ways to import a module:**

````python
# 1. import the whole module, then use module.thing
import math
print(math.sqrt(25))     # 5.0
print(math.floor(4.7))   # 4
````

````python
# 2. import specific things from a module
from math import sqrt, pi
print(sqrt(49))   # 7.0
print(pi)         # 3.141592653589793
````

````python
# 3. import with a short nickname (alias) — very common in data science
import math as m
print(m.factorial(5))   # 120
````

### Making your own module
If you save a file called `mymath.py` with a function inside it, you can use it in another file:
```python
# in mymath.py
def add(a, b):
    return a + b

# in another file
import mymath
print(mymath.add(2, 3))   # 5
```

### What is a Package?

A **package** is a **folder of modules** grouped together (with a special `__init__.py` file). It's just a bigger, organized collection of code.

```
mypackage/            ← the package (folder)
├── __init__.py
├── module1.py        ← modules inside it
└── module2.py
```

### Installing external packages with `pip`

Python has thousands of free packages made by other people. You install them with **`pip`**:
```
pip install numpy
pip install pandas
```
Then use them like any module: `import numpy`. In Data Science, you'll use packages like **NumPy**, **Pandas**, and **Matplotlib** constantly.

| Term | Meaning |
|------|---------|
| **Module** | one `.py` file of reusable code |
| **Package** | a folder of related modules |
| **Standard library** | modules that come built-in (`math`, `random`, `json`, ...) |
| **`pip`** | the tool to install external packages |
