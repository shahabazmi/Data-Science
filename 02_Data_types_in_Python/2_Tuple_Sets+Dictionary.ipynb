# 🧾 Tuples

**In simple words:** a tuple is just like a list, **but you can't change it** after it's made.

🧾 **Analogy:** think of a **printed receipt** — once printed, the items on it are fixed.

> 🔑 A tuple is an **immutable list**. Once created, it cannot be changed in any way.

<a id="tuple-char"></a>
## ✨ Characteristics of a tuple

| # | Property | Meaning |
|---|---|---|
| 1 | **Ordered** | position matters → `(1, 2, 3) != (3, 2, 1)` |
| 2 | **Unchangeable** | immutable — cannot be edited once created |
| 3 | **Allows duplicates** | the same value can appear many times |

<a id="tuple-create"></a>
## 🛠️ Creating tuples

> ⚠️ **Gotcha:** `("hello")` is **not** a tuple — it's just a string in brackets!
> For a single-item tuple you **must** add a trailing comma: `("hello",)`.

````python
# empty tuple
t1 = ()
print(t1)

# single item WITHOUT comma -> this is NOT a tuple
t2 = ("hello")
print(type(t2))     # <class 'str'>

# single item WITH a trailing comma -> now it's a tuple
t3 = ("hello",)
print(t3, type(t3))

# homogeneous tuple (all same type)
t4 = (1, 2, 3, 4)
print(t4)

# heterogeneous tuple (mixed types, even a list inside)
t5 = (1, 2, 3, True, [1, 2, 3])
print(t5)

# nested (2D) tuple -> a tuple inside a tuple
t6 = (1, 2, 3, 4, (5, 6, 7))
print(t6)

# from another sequence using type conversion
t7 = tuple("hello")
print(t7)
````

<a id="tuple-access"></a>
## 🎯 Accessing items — Indexing & Slicing
Works exactly like lists/strings (positive & negative index, `start:end:step`).

````python
# indexing
t1 = (1, 2, 3, 4)
print(t1[0])     # 1
print(t1[-1])    # 4

# slicing
print(t1[0:4])       # (1, 2, 3, 4)
print(t1[0:4:2])     # (1, 3)
print(t1[-3:-1])     # (2, 3)
print(t1[::-1])      # (4, 3, 2, 1)  reversed

# reaching inside a nested tuple
t2 = (1, 2, 3, 4, (5, 6))
print(t2[-1])        # (5, 6)
print(t2[-1][0])     # 5
````

<a id="tuple-edit"></a>
## ✏️ Editing, ➕ Adding & 🗑️ Deleting

Because tuples are **immutable**:
- ❌ You **cannot edit** an item.
- ❌ You **cannot add** items (no `append` / `insert` / `extend`).
- ✅ You **can delete the whole tuple** with `del` (but not a single item).

````python
# trying to EDIT an item -> TypeError (tuple is immutable, just like a string)
t1 = (1, 2, 3, 4)
t1[0] = 100
````

````python
# deleting the WHOLE tuple works; deleting one item does not
t1 = (1, 2, 3, 4)
del t1
print(t1)     # NameError -> t1 ab exist nahi karta
````

<a id="tuple-ops"></a>
## 🔧 Operations on tuples
`+` (join), `*` (repeat), `in` (membership), and looping — same as lists.

````python
t1 = (1, 2, 3, 4)
t2 = (5, 6, 7, 8)
print(t1 + t2)      # concatenation
print(t1 * 2)       # repeat

print(1 in t1)      # membership -> True

for i in t1:        # iteration
    print(i)
````

<a id="tuple-func"></a>
## 📦 Tuple functions
`len`, `min`, `max`, `sum`, `sorted`, plus `count` and `index`.

````python
t = (1, 2, 3, 4)
print(len(t))                   # 4
print(min(t))                   # 1
print(max(t))                   # 4
print(sum(t))                   # 10
print(sorted(t))                # [1, 2, 3, 4]  (returns a list)
print(sorted(t, reverse=True))  # [4, 3, 2, 1]

print(t.count(3))   # 1  -> how many times 3 appears
print(t.index(3))   # 2  -> position of the first 3
````

````python
# index() raises an error if the value is missing (strings' find() returned -1 instead)
t = (1, 2, 3, 4)
print(t.index(6))   # ValueError -> 6 tuple me hai hi nahi
````

<a id="list-vs-tuple"></a>
## ⚖️ List vs Tuple

| Feature | List `[ ]` | Tuple `( )` |
|---|---|---|
| **Mutability** | mutable (changeable) | immutable (fixed) |
| **Speed** | slower | usually faster |
| **Memory** | heavier | lighter |
| **Built-in functions** | many | fewer |
| **Safety** | easy to change by mistake | safe from accidental edits |

Let's actually *measure* the speed and memory difference 👇

````python
# SPEED test (numbers vary per run/machine — tuple is usually a touch faster)
import time
n = 1_000_000
l = list(range(n))
t = tuple(range(n))

start = time.time()
for i in l:
    i * 5
print("time taken by list :", time.time() - start)

start = time.time()
for i in t:
    i * 5
print("time taken by tuple:", time.time() - start)
````

````python
# MEMORY test -> a tuple takes a little less space than the same list
import sys
n = 1_000_000
l = list(range(n))
t = tuple(range(n))
print("size of list :", sys.getsizeof(l), "bytes")
print("size of tuple:", sys.getsizeof(t), "bytes")
````

````python
# WHY tuples are 'safer' -> a mutability demo
# lists share the same memory, so changing one name changes the other
a = [1, 2, 3]
b = a
a.append(4)
print(a)      # [1, 2, 3, 4]
print(b)      # [1, 2, 3, 4]  <- b changed too!

# tuples are immutable, so 'a = (4,5,6)' makes a NEW tuple; b stays safe
a = (1, 2, 3)
b = a
a = (4, 5, 6)
print(a)      # (4, 5, 6)
print(b)      # (1, 2, 3)  <- unchanged
````

<a id="tuple-unpack"></a>
## 🎁 Special syntax — tuple unpacking

**Unpacking** means handing each item of a tuple to its own variable in one line.

````python
# basic unpacking -> variables must MATCH the number of items
a, b, c = (1, 2, 3)
print(a, b, c)
````

````python
# mismatch -> too many values, too few variables -> ValueError
c, d = (4, 5, 6)
````

````python
# swapping two values the Pythonic way (no temp variable needed!)
a = 1
b = 2
a, b = b, a
print(a, b)     # 2 1
````

````python
# '*' collects the leftover items into a list
a, b, *others = (1, 2, 3, 4)
print(a, b)     # 1 2
print(others)   # [3, 4]
````

````python
# zipping two tuples together
a = (1, 2, 3, 4)
b = (5, 6, 7, 8)
print(list(zip(a, b)))   # [(1, 5), (2, 6), (3, 7), (4, 8)]
````

<a id="what-set"></a>
# 🎱 Sets

**In simple words:** a set is a collection of **unique** items with **no order**.

🎱 **Analogy:** a **bag of marbles** — you can't have two identical marbles, and there's no
"first" or "second" marble.

> 🔑 Every element must be **unique** and **immutable**, but the set **itself is mutable**
> (you can add/remove items). Sets also do math like **union, intersection, difference**.

<a id="set-char"></a>
## ✨ Characteristics of a set

| # | Property | Meaning |
|---|---|---|
| 1 | **Unordered** | no positions → `{1, 2, 3} == {3, 2, 1}` |
| 2 | **Mutable** | you can add / remove items |
| 3 | **No duplicates** | repeated values are dropped automatically |
| 4 | **Only immutable items** | can hold numbers, strings, tuples — but **not** lists or sets |

<a id="set-create"></a>
## 🛠️ Creating a set
> ⚠️ **Gotcha:** `{}` creates an empty **dict**, not a set. Use `set()` for an empty set.

````python
# {} is an empty DICT, not a set!
s = {}
print(s, "->", type(s))     # {} -> <class 'dict'>

# use set() for an empty set
p = set()
print(p, "->", type(p))     # set() -> <class 'set'>

# a normal set
s1 = {1, 2, 3}
print(s1, "->", type(s1))

# heterogeneous -> mixed types allowed (True == 1, so it's a duplicate and dropped)
s3 = {1, 2, 3, True, "hello"}
print(s3)

# a tuple can live inside a set (immutable); a list cannot
s4 = {1, 2, 3, (4, 5)}
print(s4)

# from another sequence
s5 = set([1, 2, 3])
print(s5)

# duplicates are removed automatically
s6 = {1, 1, 2, 3, 4, 3}
print(s6)

# UNORDERED -> order doesn't matter, so these are equal
print({1, 2, 3} == {3, 2, 1})   # True
# note: {1, 2, 3, [4, 5]} would raise TypeError -> a list is not allowed inside a set
````

<a id="set-access"></a>
## 🎯 Accessing & ✏️ Editing items
Sets are **unordered**, so **indexing and slicing don't work** — and since you can't index,
you **can't edit** a specific item either. You can only add or remove whole items.

<a id="set-add"></a>
## ➕ Adding items

| Method | Adds |
|---|---|
| `add(x)` | one item (position is decided by Python, not you) |
| `update([...])` | many items at once |

````python
s1 = {1, 2, 3, 4}
s1.add(5)              # add a single item
print(s1)

s1.update([5, 6, 7])   # add multiple items (duplicates ignored)
print(s1)
````

<a id="set-delete"></a>
## 🗑️ Deleting items

| Way | Behaviour on a missing item |
|---|---|
| `discard(x)` | ✅ no error |
| `remove(x)` | ❌ raises `KeyError` |
| `pop()` | removes a **random** item (set is unordered) |
| `clear()` | empties the whole set |
| `del s` | deletes the entire set variable |

````python
# discard -> remove an item; missing item = NO error
s = {1, 2, 3, 4}
s.discard(4)
print(s)
s.discard(99)      # not present -> silently ignored
print(s)

# remove -> like discard, but a MISSING item raises KeyError
s.remove(3)
print(s)

# pop -> removes and returns a RANDOM item
s = {1, 2, 3, 4}
s.pop()
print(s)

# clear -> empties the set
s.clear()
print(s)           # set()

# del -> deletes the whole set variable
s1 = {1, 2, 3, 4}
del s1
# print(s1)  # <- would raise NameError, kyunki s1 delete ho chuka hai
````

<a id="set-ops"></a>
## 🔧 Set operations (the math part)

For `s1` and `s2`:

| Operation | Symbol | Meaning |
|---|---|---|
| **Union** | `\|` | everything in **either** set |
| **Intersection** | `&` | only the **common** items |
| **Difference** | `-` | in the first but **not** the second |
| **Symmetric difference** | `^` | in **either but not both** (drops the common ones) |

````python
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8}

print(s1 | s2)   # union            -> {1, 2, 3, 4, 5, 6, 7, 8}
print(s1 & s2)   # intersection     -> {4, 5}
print(s1 - s2)   # difference       -> {1, 2, 3}  (jo s2 me nahi hai)
print(s2 - s1)   # difference       -> {6, 7, 8}
print(s1 ^ s2)   # symmetric diff   -> {1, 2, 3, 6, 7, 8}  (common {4,5} hata diya)

print(1 in s1)       # membership
print(1 not in s2)

for i in s1:         # iteration
    print(i)
````

<a id="set-func"></a>
## 📦 Set functions
The math operations also have **method** forms — and each has an `_update` variant that
changes the set **in place** instead of returning a new one.

````python
# len / max / min / sorted
s = {1, 2, 3, 4, 5}
print(len(s))                   # 5
print(max(s))                   # 5
print(min(s))                   # 1
print(sorted(s))                # [1, 2, 3, 4, 5]
print(sorted(s, reverse=True))  # [5, 4, 3, 2, 1]
````

````python
# union  vs  update (in-place)
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8}
print(s1.union(s2))    # returns a NEW set (same as s1 | s2)

s1.update(s2)          # adds s2's items INTO s1
print(s1)
print(s2)              # s2 stays unchanged
````

````python
# intersection  vs  intersection_update (in-place)
s2 = {1, 2, 3, 4, 5}
s3 = {4, 5, 6, 7, 8}
print(s2.intersection(s3))     # new set -> {4, 5}

s2.intersection_update(s3)     # keep only the common items IN s2
print(s2)
print(s3)
````

````python
# difference  vs  difference_update (in-place)
s4 = {1, 2, 3, 4, 5}
s5 = {4, 5, 6, 7, 8}
print(s4.difference(s5))       # new set -> {1, 2, 3}

s4.difference_update(s5)       # remove s5's items FROM s4
print(s4)
print(s5)
````

````python
# symmetric_difference  vs  symmetric_difference_update (in-place)
s6 = {1, 2, 3, 4, 5}
s7 = {4, 5, 6, 7, 8}
print(s6.symmetric_difference(s7))   # new set -> {1, 2, 3, 6, 7, 8}

s6.symmetric_difference_update(s7)   # apply the change IN s6
print(s6)
print(s7)
````

````python
# isdisjoint -> True if the sets have NOTHING in common
print({1, 2, 3, 4}.isdisjoint({5, 6, 7, 8}))   # True

# issubset -> is every item of the first inside the second?
print({2, 3, 4}.issubset({1, 2, 3, 4, 5}))     # True

# issuperset -> does the first CONTAIN all items of the second?
print({1, 2, 3, 4, 5}.issuperset({2, 3, 4}))   # True

# copy -> a separate copy at a new memory address
s7 = {1, 2, 3}
s8 = s7.copy()
print(s7, s8)
````

<a id="frozenset"></a>
## 🧊 Frozenset
A **frozenset** is just an **immutable** version of a set — like a set frozen in ice 🧊.
It supports all the read-only operations, but you can't add/remove items.
Because it's immutable, a frozenset **can even be nested inside another set/frozenset**.

````python
# create a frozenset
fs = frozenset([1, 2, 3])
print(fs)
````

````python
# a frozenset CAN be placed inside another frozenset (a normal set could not be nested)
fs = frozenset([1, 2, frozenset([3, 4])])
print(fs)
````

<a id="set-comp"></a>
## ⚡ Set comprehension
Same idea as list comprehension, but with `{ }` → the result is a **set** (unique items).

````python
print({i for i in range(1, 11)})              # {1, 2, ..., 10}
print({i for i in range(1, 11) if i > 5})     # with a filter -> {6, 7, 8, 9, 10}
print({i ** 2 for i in range(1, 11)})         # squares
````

<a id="what-dict"></a>
# 📖 Dictionary

**In simple words:** a dictionary stores data as **key → value** pairs, so you can look up a
value by its key (instead of by a position).

📖 **Analogy:** a **phonebook** — you look up a *name* (key) to get a *number* (value).
In other languages it's called a **map** or **associative array**.

```python
d = {"name": "shahab", "age": 24, "gender": "male"}
```

<a id="dict-char"></a>
## ✨ Characteristics

| # | Property | Meaning |
|---|---|---|
| 1 | **Mutable** | you can add / change / remove pairs |
| 2 | **Indexing has no meaning** | you access by **key**, not position |
| 3 | **Keys are unique** | duplicate keys → the last one wins |
| 4 | **Keys must be immutable** | strings, numbers, tuples are OK; lists are **not** |

> 💡 Lists, tuples and sets store **single items**; a dictionary stores items in **pairs**.

<a id="dict-create"></a>
## 🛠️ Creating a dictionary
> ⚠️ Don't name a variable `dict` — it hides Python's built-in `dict()` function!

````python
# empty dict
d = {}
print(d)

# 1D dictionary -> key : value pairs
person = {"name": "shahab", "gender": "male"}
print(person)

# keys of different types (a tuple key + a string key)
d2 = {(1, 2, 3): 1, "hello": "world"}
print(d2)

# 2D dictionary -> a dict inside a dict (nested)
student = {
    "name": "shahab",
    "college": "snu",
    "semester": 4,
    "subject": {"maths": 50, "science": 60, "english": 70},
}
print(student)

# building a dict from a list of pairs using dict()
d4 = dict([(1, 1), (2, 2), (3, 3)])
print(d4)

# duplicate keys -> the LAST value wins
ds = {"name": "shahab", "name": "rahul"}
print(ds)        # {'name': 'rahul'}

# keys must be immutable -> a tuple key is fine, a list key is NOT
ms = {"name": "shahab", (1, 2, 3): 2}
print(ms)
# ss = {"name": "shahab", [1, 2, 3]: 2}   # <- would raise TypeError (list is unhashable)
````

<a id="dict-access"></a>
## 🎯 Accessing items
Use the **key** inside `[ ]`, or the safer `.get()` (which returns `None` instead of an error).

````python
my_dict = {"name": "shahab", "age": 24}

# access by KEY (positional indexing does not work)
print(my_dict["name"])
print(my_dict["age"])

# .get() -> safer: returns None if the key is missing (no error)
print(my_dict.get("name"))
print(my_dict.get("salary"))   # None
````

<a id="dict-add"></a>
## ➕ Adding key-value pairs
Assign to a **new key** → it gets added.

````python
person = {"name": "shahab", "age": 24}
person["gender"] = "male"     # new key -> added
person["weight"] = 70
print(person)
````

<a id="dict-remove"></a>
## 🗑️ Removing key-value pairs

| Way | Removes |
|---|---|
| `pop(key)` | the given key |
| `popitem()` | the **last inserted** pair |
| `del d[key]` | a specific key |
| `clear()` | everything |

````python
person = {"name": "shahab", "age": 24, "gender": "male", "weight": 70}

person.pop("name")     # remove a specific key
print(person)

person.popitem()       # remove the LAST inserted pair
print(person)

d = {"name": "shahab", "age": 24, "gender": "male", "weight": 70}
del d["name"]          # remove a specific key
print(d)

d.clear()              # empty the dict
print(d)
````

<a id="dict-2d"></a>
## 🎯 Accessing items in a 2D (nested) dictionary
Chain the keys: `student["subject"]["maths"]`.

````python
print(student)
print(student["subject"]["maths"])    # 50  -> go into 'subject', then 'maths'

# add / change a nested value
student["subject"]["maths"] = 75
print(student)

# delete a nested key
del student["subject"]["maths"]
print(student)
````

<a id="dict-edit"></a>
## ✏️ Editing key-value pairs
Assign to an **existing key** → its value is updated.

````python
print(student)
student["semester"] = 5              # change an existing value
print(student)
student["subject"]["science"] = 55   # change a nested value
print(student)
````

<a id="dict-ops"></a>
## 🔧 Dictionary operations
`in` checks **keys** (not values); looping also goes over the keys.

````python
# membership -> checks KEYS, not values
print("shahab" in student)   # False -> 'shahab' is a value, not a key
print("name" in student)     # True

# iteration -> loops over keys; use d[key] to reach the value
d = {"name": "shahab", "age": 24, "gender": "male"}
for key in d:
    print(key, ":", d[key])
````

<a id="dict-func"></a>
## 📦 Dictionary functions

| Method | Returns |
|---|---|
| `items()` | all key-value pairs (as tuples) |
| `keys()` | all keys |
| `values()` | all values |
| `update(d2)` | merges `d2` in (matching keys overwritten) |

`len`, `sorted`, `min`, `max` work on the **keys**.

````python
d = {"name": "shahab", "age": 24, "gender": "male"}
print(len(d))          # 3
print(sorted(d))       # sorts the KEYS -> ['age', 'gender', 'name']
print(min(d), max(d))  # min/max of the keys (alphabetical)

print(d.items())       # all pairs as tuples
print(d.keys())        # all keys
print(d.values())      # all values

# update -> merge another dict in (key 4 gets overwritten)
d1 = {1: 2, 3: 4, 4: 5}
d2 = {4: 7, 6: 8}
d1.update(d2)
print(d1)              # {1: 2, 3: 4, 4: 7, 6: 8}
````

<a id="dict-comp"></a>
## ⚡ Dictionary comprehension

```
{ key: value  for var in iterable  if condition }
```

````python
# first 10 numbers and their squares
print({i: i ** 2 for i in range(1, 11)})

# transform an existing dict -> km to miles (x 0.62)
distances = {"delhi": 1000, "mumbai": 2000, "banglore": 3000}
print({key: value * 0.62 for key, value in distances.items()})
````

````python
# build a dict from two lists using zip
days = ["sunday", "monday", "tuesday", "wednesday", "thursday", "friday", "saturday"]
temp = [30.5, 32.6, 31.8, 33.4, 29.8, 30.2, 29.9]
print({day: t for day, t in zip(days, temp)})
````

````python
# with an 'if' filter -> keep only in-stock products (value > 0)
products = {"phone": 10, "laptop": 0, "charger": 70, "tablet": 5}
print({key: value for key, value in products.items() if value > 0})
````

````python
# nested comprehension -> multiplication tables for 2, 3, 4 (a dict of dicts)
print({i: {j: i * j for j in range(1, 11)} for i in range(2, 5)})
````
