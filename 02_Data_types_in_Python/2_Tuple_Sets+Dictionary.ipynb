# Tuples

A **tuple** Python me list ke jaisa hi hota hai. Dono me difference sirf itna hai ki **tuple
ke elements ko ek baar assign karne ke baad hum change nahi kar sakte**, jabki list ke
elements ko hum change kar sakte hai. In short, a tuple is an **immutable list** — ek baar
create hone ke baad tuple ko kisi bhi tarah se change nahi kiya ja sakta.

## Characteristics of a tuple

- **Ordered** — for example `(1, 2, 3)` aur `(3, 2, 1)` equal nahi hai; content same hai par
  order matter karta hai.
- **Unchangeable** — tuple immutable hota hai.
- **Allows duplicates** — tuple me multiple duplicate values store kar sakte hai, koi
  restriction nahi.

## Creating Tuples

Tuple banane ke liye items ko round brackets `()` me, comma se separate karke likhte hai.

Note: single item ka tuple banate waqt end me comma (`,`) lagana zaroori hai — warna wo tuple
nahi banta. Jaise `("hello")` ek string hai, par `("hello",)` ek tuple hai.

````python
# empty tuple
t1 = ()
print(t1)

# single item WITHOUT comma -> this is NOT a tuple
t2 = ("hello")
print(type(t2))     # <class 'str'>

# single item WITH a trailing comma -> now it's a tuple
t3 = ("hello",)
print(t3, type(t3))

# homogeneous tuple (all same type)
t4 = (1, 2, 3, 4)
print(t4)

# heterogeneous tuple (mixed types, even a list inside)
t5 = (1, 2, 3, True, [1, 2, 3])
print(t5)

# nested (2D) tuple -> a tuple inside a tuple
t6 = (1, 2, 3, 4, (5, 6, 7))
print(t6)

# from another sequence using type conversion
t7 = tuple("hello")
print(t7)
````

## Accessing Items

Tuple me se items nikalne ke do tarike hai (bilkul list/string ki tarah):
- **Indexing** — ek single item ko uski position se nikalna (positive index `0` se start,
  negative index `-1` se).
- **Slicing** — ek saath multiple items nikalna, `t[start : end : step]` ke through (`end`
  wala index include nahi hota).

````python
# indexing
t1 = (1, 2, 3, 4)
print(t1[0])     # 1
print(t1[-1])    # 4

# slicing
print(t1[0:4])       # (1, 2, 3, 4)
print(t1[0:4:2])     # (1, 3)
print(t1[-3:-1])     # (2, 3)
print(t1[::-1])      # (4, 3, 2, 1)  reversed

# reaching inside a nested tuple
t2 = (1, 2, 3, 4, (5, 6))
print(t2[-1])        # (5, 6)
print(t2[-1][0])     # 5
````

## Editing, Adding & Deleting Items

Kyunki tuple **immutable** hai:
- Kisi item ko **edit nahi** kar sakte (error aayega, jaise string me hota hai).
- Naye items **add nahi** kar sakte.
- Poore tuple ko `del` se delete kar sakte hai, par uska ek portion delete nahi kar sakte.

````python
# trying to EDIT an item -> TypeError (tuple is immutable, just like a string)
t1 = (1, 2, 3, 4)
t1[0] = 100
````

````python
# deleting the WHOLE tuple works; deleting one item does not
t1 = (1, 2, 3, 4)
del t1
print(t1)     # NameError -> t1 ab exist nahi karta
````

## Operations on tuples

Tuples par `+` (jodna), `*` (repeat), `in` (membership), aur loop — sab kuch list ki tarah
chalta hai.

````python
t1 = (1, 2, 3, 4)
t2 = (5, 6, 7, 8)
print(t1 + t2)      # concatenation
print(t1 * 2)       # repeat

print(1 in t1)      # membership -> True

for i in t1:        # iteration
    print(i)
````

## Tuple Functions

`len`, `min`, `max`, `sum`, `sorted`, aur saath me `count` aur `index`. Dhyaan rahe — `index`
agar value na mile to error deta hai (string ke `find` ki tarah `-1` nahi).

````python
t = (1, 2, 3, 4)
print(len(t))                   # 4
print(min(t))                   # 1
print(max(t))                   # 4
print(sum(t))                   # 10
print(sorted(t))                # [1, 2, 3, 4]  (returns a list)
print(sorted(t, reverse=True))  # [4, 3, 2, 1]

print(t.count(3))   # 1  -> how many times 3 appears
print(t.index(3))   # 2  -> position of the first 3
````

````python
# index() raises an error if the value is missing (strings' find() returned -1 instead)
t = (1, 2, 3, 4)
print(t.index(6))   # ValueError -> 6 tuple me hai hi nahi
````

## Difference between Lists and Tuples

| Feature | List `[ ]` | Tuple `( )` |
|---|---|---|
| **Mutability** | mutable | immutable |
| **Speed** | slow | fast |
| **Memory** | zyada | kam |
| **Functions** | zyada | kam |
| **Error prone** | zyada | kam |

- **Syntax** — tuple banane ke liye `()` use karte hai, aur list ke liye `[]`.
- **Mutability** — tuples immutable hote hai, jabki lists mutable.
- **Speed** — tuples lists se faster hote hai, kyunki tuples immutable hai; generally immutable
  datatypes mutable datatypes se fast hote hai.
- **Memory** — tuples lists ke comparison me kam storage lete hai; immutable datatypes kam
  memory lete hai.
- **Built-in functions** — lists tuples se zyada functions deti hai.
- **Error prone** — lists zyada error-prone hoti hai, matlab debugging me thoda zyada dimaag
  lagana pad sakta hai.

Neeche speed aur memory ka difference actually measure karke dekhte hai.

````python
# SPEED test (numbers vary per run/machine — tuple is usually a touch faster)
import time
n = 1_000_000
l = list(range(n))
t = tuple(range(n))

start = time.time()
for i in l:
    i * 5
print("time taken by list :", time.time() - start)

start = time.time()
for i in t:
    i * 5
print("time taken by tuple:", time.time() - start)
````

````python
# MEMORY test -> a tuple takes a little less space than the same list
import sys
n = 1_000_000
l = list(range(n))
t = tuple(range(n))
print("size of list :", sys.getsizeof(l), "bytes")
print("size of tuple:", sys.getsizeof(t), "bytes")
````

````python
# WHY tuples are 'safer' -> a mutability demo
# lists share the same memory, so changing one name changes the other
a = [1, 2, 3]
b = a
a.append(4)
print(a)      # [1, 2, 3, 4]
print(b)      # [1, 2, 3, 4]  <- b changed too!

# tuples are immutable, so 'a = (4,5,6)' makes a NEW tuple; b stays safe
a = (1, 2, 3)
b = a
a = (4, 5, 6)
print(a)      # (4, 5, 6)
print(b)      # (1, 2, 3)  <- unchanged
````

## Special Syntax — Tuple unpacking

**Unpacking** matlab tuple ke har item ko alag-alag variables me ek saath assign karna. Dhyaan
rahe — variables ki count tuple ke items ke barabar honi chahiye.

````python
# basic unpacking -> variables must MATCH the number of items
a, b, c = (1, 2, 3)
print(a, b, c)
````

````python
# mismatch -> too many values, too few variables -> ValueError
c, d = (4, 5, 6)
````

````python
# swapping two values the Pythonic way (no temp variable needed!)
a = 1
b = 2
a, b = b, a
print(a, b)     # 2 1
````

````python
# '*' collects the leftover items into a list
a, b, *others = (1, 2, 3, 4)
print(a, b)     # 1 2
print(others)   # [3, 4]
````

````python
# zipping two tuples together
a = (1, 2, 3, 4)
b = (5, 6, 7, 8)
print(list(zip(a, b)))   # [(1, 5), (2, 6), (3, 7), (4, 8)]
````

# Sets

A **set** items ka ek collection hai. Set ka har element **unique** hota hai (no duplicates)
aur **immutable** hona chahiye (change nahi ho sakta). However, set khud **mutable** hai — hum
usme items add ya remove kar sakte hai.

Sets ko mathematical set operations ke liye bhi use kiya jaata hai — jaise union, intersection,
symmetric difference, difference, etc.

## Characteristics

- **Unordered** — `{1, 2, 3}` aur `{3, 2, 1}` same hai (order matter nahi karta).
- **Mutable** — set ke andar hum add ya delete kar sakte hai, matlab create karne ke baad set
  ko change kar sakte hai.
- **No duplicates** — duplicate values apne aap hat jaati hai.
- **Can't contain mutable data types** — set khud mutable hai (add/delete ho sakta hai), par
  hum uske andar kisi mutable data type ko nahi rakh sakte. For example, ek 2D set (set ke
  andar set) possible nahi hai.

## Creating a Set

Set banane ke liye items ko curly braces `{}` me, comma se separate karke rakhte hai (ya kisi
sequence par `set()` laga ke).

Note: `{}` se ek empty **dictionary** banti hai, set nahi. Empty set banane ke liye `set()`
use karte hai.

````python
# {} is an empty DICT, not a set!
s = {}
print(s, "->", type(s))     # {} -> <class 'dict'>

# use set() for an empty set
p = set()
print(p, "->", type(p))     # set() -> <class 'set'>

# a normal set
s1 = {1, 2, 3}
print(s1, "->", type(s1))

# heterogeneous -> mixed types allowed (True == 1, so it's a duplicate and dropped)
s3 = {1, 2, 3, True, "hello"}
print(s3)

# a tuple can live inside a set (immutable); a list cannot
s4 = {1, 2, 3, (4, 5)}
print(s4)

# from another sequence
s5 = set([1, 2, 3])
print(s5)

# duplicates are removed automatically
s6 = {1, 1, 2, 3, 4, 3}
print(s6)

# UNORDERED -> order doesn't matter, so these are equal
print({1, 2, 3} == {3, 2, 1})   # True
# note: {1, 2, 3, [4, 5]} would raise TypeError -> a list is not allowed inside a set
````

## Accessing & Editing Items

Set **unordered** hota hai, isliye ismein indexing aur slicing kaam nahi karti. Aur jab
indexing hi kaam nahi karti, to editing (jaise `s[0] = 100`) bhi kaam nahi karegi — kyunki set
unordered hota hai. Hum sirf poore items add ya remove kar sakte hai.

## Adding Items

Hum set me naye items add kar sakte hai:
- **add** — ek baar me ek single item add karta hai.
- **update** — ek saath multiple items add karta hai (list ke andar daal ke bhejna padta hai).

````python
s1 = {1, 2, 3, 4}
s1.add(5)              # add a single item
print(s1)

s1.update([5, 6, 7])   # add multiple items (duplicates ignored)
print(s1)
````

## Deleting Items

| Way | Missing item par |
|---|---|
| `discard(x)` | error nahi deta |
| `remove(x)` | error (KeyError) |
| `pop()` | random item delete |
| `clear()` | poora set khali |
| `del s` | poora set variable delete |

Delete karne ke multiple options hai:
- **discard** — jo item delete karna ho kar deta hai; agar item exist na kare to error nahi deta.
- **remove** — discard jaisa hi kaam, par agar item exist na kare to error (KeyError) deta hai.
- **pop** — set me se koi bhi random item delete karta hai (set unordered hai).
- **clear** — poore set ko khali kar deta hai.
- **del** — poore set variable ko delete kar deta hai (uske andar ke single item ko nahi).

````python
# discard -> remove an item; missing item = NO error
s = {1, 2, 3, 4}
s.discard(4)
print(s)
s.discard(99)      # not present -> silently ignored
print(s)

# remove -> like discard, but a MISSING item raises KeyError
s.remove(3)
print(s)

# pop -> removes and returns a RANDOM item
s = {1, 2, 3, 4}
s.pop()
print(s)

# clear -> empties the set
s.clear()
print(s)           # set()

# del -> deletes the whole set variable
s1 = {1, 2, 3, 4}
del s1
# print(s1)  # <- would raise NameError, kyunki s1 delete ho chuka hai
````

## Set Operations

Do sets `s1` aur `s2` ke liye:

| Operation | Symbol | Kya deta hai |
|---|---|---|
| **Union** | `\|` | dono me se saare items |
| **Intersection** | `&` | sirf common items |
| **Difference** | `-` | jo pehle me hai par doosre me nahi |
| **Symmetric Difference** | `^` | dono me se kisi ek me (common chhod ke) |

- **Union** (`|`) — dono me se saare items.
- **Intersection** (`&`) — sirf common items.
- **Difference** (`-`) — jo pehle set me hai par doosre me nahi.
- **Symmetric Difference** (`^`) — jo dono me se kisi ek me hai par common nahi (common items
  chhod ke sab).

````python
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8}

print(s1 | s2)   # union            -> {1, 2, 3, 4, 5, 6, 7, 8}
print(s1 & s2)   # intersection     -> {4, 5}
print(s1 - s2)   # difference       -> {1, 2, 3}  (jo s2 me nahi hai)
print(s2 - s1)   # difference       -> {6, 7, 8}
print(s1 ^ s2)   # symmetric diff   -> {1, 2, 3, 6, 7, 8}  (common {4,5} hata diya)

print(1 in s1)       # membership
print(1 not in s2)

for i in s1:         # iteration
    print(i)
````

## Set Functions

`len`, `max`, `min`, `sorted`, aur math operations ke method versions. Har math method ka ek
`_update` variant bhi hota hai jo naye set ki jagah usi set ko **in-place** change karta hai.

````python
# len / max / min / sorted
s = {1, 2, 3, 4, 5}
print(len(s))                   # 5
print(max(s))                   # 5
print(min(s))                   # 1
print(sorted(s))                # [1, 2, 3, 4, 5]
print(sorted(s, reverse=True))  # [5, 4, 3, 2, 1]
````

````python
# union  vs  update (in-place)
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8}
print(s1.union(s2))    # returns a NEW set (same as s1 | s2)

s1.update(s2)          # adds s2's items INTO s1
print(s1)
print(s2)              # s2 stays unchanged
````

````python
# intersection  vs  intersection_update (in-place)
s2 = {1, 2, 3, 4, 5}
s3 = {4, 5, 6, 7, 8}
print(s2.intersection(s3))     # new set -> {4, 5}

s2.intersection_update(s3)     # keep only the common items IN s2
print(s2)
print(s3)
````

````python
# difference  vs  difference_update (in-place)
s4 = {1, 2, 3, 4, 5}
s5 = {4, 5, 6, 7, 8}
print(s4.difference(s5))       # new set -> {1, 2, 3}

s4.difference_update(s5)       # remove s5's items FROM s4
print(s4)
print(s5)
````

````python
# symmetric_difference  vs  symmetric_difference_update (in-place)
s6 = {1, 2, 3, 4, 5}
s7 = {4, 5, 6, 7, 8}
print(s6.symmetric_difference(s7))   # new set -> {1, 2, 3, 6, 7, 8}

s6.symmetric_difference_update(s7)   # apply the change IN s6
print(s6)
print(s7)
````

````python
# isdisjoint -> True if the sets have NOTHING in common
print({1, 2, 3, 4}.isdisjoint({5, 6, 7, 8}))   # True

# issubset -> is every item of the first inside the second?
print({2, 3, 4}.issubset({1, 2, 3, 4, 5}))     # True

# issuperset -> does the first CONTAIN all items of the second?
print({1, 2, 3, 4, 5}.issuperset({2, 3, 4}))   # True

# copy -> a separate copy at a new memory address
s7 = {1, 2, 3}
s8 = s7.copy()
print(s7, s8)
````

## Frozenset

Ek **frozenset** Python set object ka bas ek **immutable** version hai. Ismein saare read-only
operations chalte hai, par items add/remove nahi kar sakte. Immutable hone ki wajah se ek
frozenset ko doosre set ya frozenset ke andar bhi rakh sakte hai.

````python
# create a frozenset
fs = frozenset([1, 2, 3])
print(fs)
````

### Frozenset saare operations par kaam karta hai

````python
# a frozenset CAN be placed inside another frozenset (a normal set could not be nested)
fs = frozenset([1, 2, frozenset([3, 4])])
print(fs)
````

## Set Comprehension

Set comprehension ek concise (chhota) tarika hai set banane ka — list comprehension jaisa hi,
bas `[]` ki jagah `{}` use karte hai. Result ek **set** hota hai, isliye duplicate values apne
aap hat jaati hai.

```
newset = {expression for item in iterable if condition}
```

````python
print({i for i in range(1, 11)})              # {1, 2, ..., 10}
print({i for i in range(1, 11) if i > 5})     # with a filter -> {6, 7, 8, 9, 10}
print({i ** 2 for i in range(1, 11)})         # squares
````

# Dictionary

Python me **dictionary** keys aur values ka ek collection hai, jo data ko ek **map** ki tarah
store karta hai. Baaki data types (jo ek element me sirf ek single value rakhte hai) ke ulat,
dictionary data ko **key: value** pairs me store karti hai. Kuch languages me isse **map** ya
**associative array** bhi kaha jaata hai.

```python
d = {"name": "shahab", "age": 24, "gender": "male"}
```

## Characteristics

- **Mutable** — pairs add / change / remove kar sakte hai.
- **Indexing ka koi matlab nahi** — value ko position se nahi, **key** se access karte hai.
- **Keys duplicate nahi ho sakti** — agar duplicate key ho to aakhri wali value chalti hai.
- **Keys mutable items nahi ho sakti** — string, number, tuple keys chalti hai; list key nahi.

Note: lists, tuple, aur set sirf single items store karte hai, jabki dictionary items ko
**pairs (key aur uski value)** ke form me store karti hai.

## Create Dictionary

Dictionary banane ke liye `key: value` pairs ko curly braces `{}` me, comma se separate karke
likhte hai.

Note: apne variable ka naam `dict` mat rakho — isse Python ka built-in `dict()` function chhup
jaata hai.

````python
# empty dict
d = {}
print(d)

# 1D dictionary -> key : value pairs
person = {"name": "shahab", "gender": "male"}
print(person)

# keys of different types (a tuple key + a string key)
d2 = {(1, 2, 3): 1, "hello": "world"}
print(d2)

# 2D dictionary -> a dict inside a dict (nested)
student = {
    "name": "shahab",
    "college": "snu",
    "semester": 4,
    "subject": {"maths": 50, "science": 60, "english": 70},
}
print(student)

# building a dict from a list of pairs using dict()
d4 = dict([(1, 1), (2, 2), (3, 3)])
print(d4)

# duplicate keys -> the LAST value wins
ds = {"name": "shahab", "name": "rahul"}
print(ds)        # {'name': 'rahul'}

# keys must be immutable -> a tuple key is fine, a list key is NOT
ms = {"name": "shahab", (1, 2, 3): 2}
print(ms)
# ss = {"name": "shahab", [1, 2, 3]: 2}   # <- would raise TypeError (list is unhashable)
````

## Accessing Items

Value ko **key** se access karte hai (`[]` ke andar key do), ya `.get()` se — jo key na mile
to error ke bajaye `None` return karta hai.

````python
my_dict = {"name": "shahab", "age": 24}

# access by KEY (positional indexing does not work)
print(my_dict["name"])
print(my_dict["age"])

# .get() -> safer: returns None if the key is missing (no error)
print(my_dict.get("name"))
print(my_dict.get("salary"))   # None
````

## Adding key-value pair

Ek nayi key ko value assign karo, to wo pair dictionary me add ho jaata hai.

````python
person = {"name": "shahab", "age": 24}
person["gender"] = "male"     # new key -> added
person["weight"] = 70
print(person)
````

## Remove key-value pair

| Way | Kya remove karta hai |
|---|---|
| `pop(key)` | di gayi key |
| `popitem()` | last insert kiya gaya pair |
| `del d[key]` | ek specific key |
| `clear()` | sab kuch |

- **pop(key)** — di gayi key ko remove karta hai.
- **popitem()** — sabse last insert kiya gaya pair remove karta hai.
- **del d[key]** — ek specific key remove karta hai.
- **clear()** — poori dictionary khali kar deta hai.

````python
person = {"name": "shahab", "age": 24, "gender": "male", "weight": 70}

person.pop("name")     # remove a specific key
print(person)

person.popitem()       # remove the LAST inserted pair
print(person)

d = {"name": "shahab", "age": 24, "gender": "male", "weight": 70}
del d["name"]          # remove a specific key
print(d)

d.clear()              # empty the dict
print(d)
````

## Accessing items from a 2D (nested) dictionary

Keys ko chain karke andar tak jaate hai, jaise `student["subject"]["maths"]`.

````python
print(student)
print(student["subject"]["maths"])    # 50  -> go into 'subject', then 'maths'

# add / change a nested value
student["subject"]["maths"] = 75
print(student)

# delete a nested key
del student["subject"]["maths"]
print(student)
````

## Editing key-value pair

Kisi existing key ko nayi value assign karo, to uski value update ho jaati hai.

````python
print(student)
student["semester"] = 5              # change an existing value
print(student)
student["subject"]["science"] = 55   # change a nested value
print(student)
````

## Dictionary Operations

`in` keys par check karta hai (values par nahi), aur loop bhi keys par hi chalta hai.

````python
# membership -> checks KEYS, not values
print("shahab" in student)   # False -> 'shahab' is a value, not a key
print("name" in student)     # True

# iteration -> loops over keys; use d[key] to reach the value
d = {"name": "shahab", "age": 24, "gender": "male"}
for key in d:
    print(key, ":", d[key])
````

## Dictionary Functions

| Method | Kya return karta hai |
|---|---|
| `items()` | saare key-value pairs (tuples) |
| `keys()` | saari keys |
| `values()` | saari values |
| `update(d2)` | doosri dict merge (matching keys overwrite) |

- **items()** — saare key-value pairs (tuples ke form me).
- **keys()** — saari keys.
- **values()** — saari values.
- **update(d2)** — doosri dictionary ko merge kar deta hai (matching keys overwrite ho jaati hai).

`len`, `sorted`, `min`, `max` keys par kaam karte hai.

````python
d = {"name": "shahab", "age": 24, "gender": "male"}
print(len(d))          # 3
print(sorted(d))       # sorts the KEYS -> ['age', 'gender', 'name']
print(min(d), max(d))  # min/max of the keys (alphabetical)

print(d.items())       # all pairs as tuples
print(d.keys())        # all keys
print(d.values())      # all values

# update -> merge another dict in (key 4 gets overwritten)
d1 = {1: 2, 3: 4, 4: 5}
d2 = {4: 7, 6: 8}
d1.update(d2)
print(d1)              # {1: 2, 3: 4, 4: 7, 6: 8}
````

## Dictionary Comprehension

Dictionary comprehension ek concise tarika hai dictionaries banane ka — list comprehension
jaisa hi, bas yahan hum `key: value` pairs banate hai.

```
newdict = {key: value for var in iterable if condition}
```

````python
# first 10 numbers and their squares
print({i: i ** 2 for i in range(1, 11)})

# transform an existing dict -> km to miles (x 0.62)
distances = {"delhi": 1000, "mumbai": 2000, "banglore": 3000}
print({key: value * 0.62 for key, value in distances.items()})
````

````python
# build a dict from two lists using zip
days = ["sunday", "monday", "tuesday", "wednesday", "thursday", "friday", "saturday"]
temp = [30.5, 32.6, 31.8, 33.4, 29.8, 30.2, 29.9]
print({day: t for day, t in zip(days, temp)})
````

````python
# with an 'if' filter -> keep only in-stock products (value > 0)
products = {"phone": 10, "laptop": 0, "charger": 70, "tablet": 5}
print({key: value for key, value in products.items() if value > 0})
````

````python
# nested comprehension -> multiplication tables for 2, 3, 4 (a dict of dicts)
print({i: {j: i * j for j in range(1, 11)} for i in range(2, 5)})
````
