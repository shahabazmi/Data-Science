## What are lists?

A **list** ek aisa data type hai jisme aap **multiple items** ko ek hi naam ke under store
kar sakte ho. More technically, lists **dynamic arrays** ki tarah act karti hai — matlab aap
chalte-chalte (on the fly) aur items add kar sakte ho.

For example, socho aap movie reviews ke sentiments analyse kar rahe ho aur aapke paas 10,000
movies hai. Har ek movie ke liye alag variable banana practically impossible hai, aur bahut
zyada space bhi lega. In sab jhamele ke bajaye, hum ek list bana lenge aur saara data usme
store kar denge.

Note: list me aap kisi bhi type ke items rakh sakte ho — numbers, strings, ya doosri lists
bhi — sab ek saath (list heterogeneous hoti hai).

```python
l = [20, "jessica", 35.5, [30, 60, 70]]
```

## Arrays vs Lists

| Feature | Array | List |
|---|---|---|
| **Size** | fixed (pehle se batana padta hai) | dynamic (apne aap badhti hai) |
| **Data types** | homogeneous (ek hi type) | heterogeneous (mix allowed) |
| **Speed** | faster | thodi slow |
| **Memory** | kam | zyada |

- **Fixed vs Dynamic size** — arrays ko use karne se pehle batana padta hai ki hum kitne
  items store karne wale hai, whereas lists ke case me hum unlimited data store kar sakte
  hai — koi limit nahi. Dynamic array me hum ek fixed-size array se start karte hai aur items
  add karte jaate hai; jaise hi array ka size full ho jaata hai, hum ek naya array create
  karte hai — wo bhi previous array ka **double size** — phir previous array ke items ko usme
  copy karte hai. Jab ye bhi fill ho jaata hai, to yahi process repeat hoti hai.
- **Convenience → Heterogeneous** — arrays homogeneous hote hai, matlab sirf ek hi data type
  ke items store kar sakte hai; but lists me aisi koi restriction nahi, isliye lists
  heterogeneous hoti hai.
- **Speed of execution** — arrays ka program lists ke comparison me zyada fast execute hota hai.
- **Memory** — lists arrays ke comparison me zyada space occupy karti hai.

## How are lists stored in memory?

Pehle samajhte hai ki arrays memory me kaise store hote hai, for example `int arr[50]`. Ek
array apne saare elements ko ek continuous (side-by-side) memory block me store karta hai.
List thoda alag kaam karti hai — wo **references (addresses)** ka ek block store karti hai,
jahan har slot asli object ko *point* karta hai jo kahin aur memory me pada hota hai. Isi
"pointing" ki wajah se ek hi list alag-alag types ke objects rakh sakti hai.

## Characteristics of a list

- **Ordered** — items apni position/order maintain karte hai.
- **Changeable / Mutable** — Python me list ko create karne ke baad bhi change kar sakte hai.
- **Heterogeneous** — ek list me multiple different data types store kar sakte hai.
- **Can have duplicates** — same value list me kai baar aa sakti hai.
- **Dynamic** — jaise-jaise items add karte jaoge, list apne aap memory increase karti jaayegi.
- **Can be nested** — ek list ke andar doosri list daal sakte hai.
- **Items can be accessed** — list se koi bhi data index/slice se extract kar sakte hai.
- **Can contain any kind of object** in Python.

Neeche "ordered" wali baat ko ek chhote example se prove karte hai:

````python
# ORDERED -> order matters in a list
l1 = [1, 2, 3]
l2 = [3, 2, 1]
print(l1 == l2)   # False -> items same hai, par order alag hai
````

## Creating a list

Ek list banane ke kai tarike hai — empty, 1D, nested (2D/3D), mixed types, ya kisi doosre
sequence se type conversion karke. Neeche har type ek saath dekhte hai.

````python
# empty list
print([])
# 1D list -> homogeneous (saare items ek hi type ke)
print([1, 2, 3, 4, 5])
# 2D / nested list -> ek list jiske andar kam se kam ek aur list ho
print([1, 2, 3, [4, 5]])
# 3D list -> list ke andar list ke andar list
print([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
# heterogeneous list -> alag-alag data types
print([1, True, 5, 6, 5 + 6j, "hello"])
# type conversion se -> string ko characters ki list me todna
print(list("hello"))
````

## Accessing items from a list

List me se items nikalne ke do tarike hai:
1. **Indexing** — ek single item, uski position se.
2. **Slicing** — ek saath multiple items.

Positive indexing left se right chalti hai (0 se start), aur negative indexing right se left
(-1 se start).

````python
# 1:- INDEXING
l = [1, 2, 3, 4, 5]
print(l[0])            # 1

# 2D list ke andar access karna
l1 = [1, 2, 3, [4, 5]]
print(l1[3][0])        # 4  -> index 3 (andar wali list) par jao, phir uska index 0

# 3D list ke andar access karna
l3 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(l3[0][0][1])     # 2

# Negative indexing -> right se left, -1 se start
print(l[-3])           # 3

# 2D list me negative indexing
l2 = [1, 2, 3, [4, 5]]
print(l2[-1])          # [4, 5] -> last item khud ek list hai
print(l2[-1][-2])      # 4      -> ek level aur andar jao
````

### Slicing

Slicing me hum ek saath **ek se zyada items** nikaal sakte hai. Syntax hai
`l[start : end : step]` — dhyaan rahe ki `end` wala index **include nahi** hota (exclusive).

````python
# 2:- SLICING -> l[start : end : step]
l = [1, 2, 3, 4, 5, 6]
print(l[0:3])       # [1, 2, 3]
print(l[-3:])       # [4, 5, 6]   (negative)
print(l[0::2])      # [1, 3, 5]   (2-2 step jump)
print(l[-5:-2:2])   # [2, 4]      (negative + step)
print(l[::-1])      # poori list reverse
````

## Adding items to a list

| Method | Kaam |
|---|---|
| `append(x)` | ek single item end me add |
| `extend([...])` | ek saath multiple items add |
| `insert(i, x)` | position `i` par item add |

List me items add karne ke 3 tarike hai:

1. **append** — ek baar me ek single item ko list ke end me add karta hai.
2. **extend** — ek saath multiple items add karta hai (append se zyada powerful).
3. **insert** — kisi specific position par item add karta hai (`insert(index, item)`).

````python
# 1:- APPEND -> ek baar me sirf ek single item, end me
l = [1, 2, 3, 4, 5]
l.append(6)
print(l)
l.append(True)
print(l)
# agar list ko append karo to wo ek nested item ki tarah add hoti hai (merge nahi hoti)
a = [1, 2, 3, 4, 5]
a.append([6, 7, 8])
print(a)

# 2:- EXTEND -> ek saath multiple items (append se powerful)
s = [1, 2, 3, 4, 5]
s.extend([6, 7, 8])
print(s)
k = [1, 2, 3, 4, 5]
k.extend("delhi")     # string iterable hai -> har character alag add hota hai
print(k)

# 3:- INSERT -> specific position par: insert(index, item)
d = [1, 2, 3, 4, 5]
d.insert(1, 100)      # index 1 par 100 daalo
print(d)
````

## Editing items in a list

List me editing possible hai kyunki list **mutable** hoti hai — but string me nahi, kyunki
wo **immutable** hoti hai. Item ko hum index se ya slice se change kar sakte hai.

````python
# indexing se editing
l = [1, 2, 3, 4, 5]
l[-1] = 500
print(l)
# slicing se editing -> ek saath poora range replace
l[1:4] = [200, 300, 400]
print(l)
````

## Deleting items from a list

| Way | Kis se delete | Note |
|---|---|---|
| `del` | index / slice | poori list bhi delete kar sakta hai |
| `remove(x)` | value | pehla match |
| `pop(i)` | index | item **return** bhi karta hai; index na do to last |
| `clear()` | — | saare items hata deta hai (list rehti hai) |

Kyunki list **mutable** hai, isliye hum list se items delete kar sakte hai — ek-ek karke, ek
saath multiple, ya poori list ko bhi.

- **del** — index ya slice se delete (poori list bhi delete kar sakta hai).
- **remove** — value se delete (pehla match).
- **pop** — index se delete karta hai aur wo item **return** bhi karta hai; index na do to last
  item delete hota hai.
- **clear** — list ke saare items hata deta hai (list rehti hai, bas khali ho jaati hai).

````python
# 1:- del  (index ya slice se)
l = [1, 2, 3, 4, 5]
print(l)
del l[-1]        # last item delete
print(l)
del l[1:3]       # ek slice delete
print(l)
del l            # poore naam 'l' ko delete
print(l)         # NameError -> 'l' ab exist nahi karta
````

````python
# 2:- remove -> VALUE se delete (index ki zaroorat nahi)
l = [1, 2, 3, 4, 5]
l.remove(5)
print(l)
````

````python
# 3:- pop -> 2 tarike se
l = [1, 2, 3, 4, 5]
l.pop(0)      # (a) diye gaye index ka item delete
print(l)
l.pop()       # (b) index na do -> LAST item delete
print(l)
````

````python
# 4:- clear -> list khali kar deta hai (list object rehta hai)
l = [1, 2, 3, 4, 5]
l.clear()
print(l)      # []
````

## Operations on a list

1. **Arithmetic** — `+` (jodna / merge) aur `*` (repeat).
2. **Membership** — `in` / `not in`.
3. **Loop** — har item par ghoomna.

````python
# 1:- ARITHMETIC (+ aur *) -> sirf yahi do operators chalte hai
l1 = [1, 2, 3, 4]
l2 = [5, 6, 7, 8]
print(l1 + l2)   # concatenation / merge
print(l1 * 3)    # 3 baar repeat
````

````python
# 2:- MEMBERSHIP OPERATORS ( in / not in )
l1 = [1, 2, 3, 4, 5]
l2 = [1, 2, 3, 4, [5, 6]]
print(5 in l1)        # True
print(5 in l2)        # False -> l2 me 5 nahi hai, [5, 6] hai
print([5, 6] in l2)   # True
````

````python
# 3:- LOOPS -> mainly sequence data types ke saath
l1 = [1, 2, 3, 4, 5]
for i in l1:
    print(i)

# 2D list par loop -> har item khud ek list ho sakta hai
l2 = [1, 2, 3, 4, [5, 6]]
for i in l2:
    print(i)

# 3D list par loop
l3 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
for i in l3:
    print(i)
````

## List functions

Python me kuch functions (`len`, `min`, `max`, `sorted`) **universal** hote hai — universal
matlab ye saare data types par available hai, jaise list, set, tuple, dictionary. Kuch list
functions hum pehle padh chuke hai: `clear()`, `pop()`, `remove()`, `append()`, `extend()`,
`insert()`.

Yahan hum baaki functions dekhenge:

| Function | Kaam |
|---|---|
| `len` / `min` / `max` | length / sabse chhota / sabse bada |
| `sorted(l)` | nayi sorted list (original same) |
| `count(x)` | x kitni baar aaya |
| `index(x)` | x ka pehla position |
| `reverse()` | list ko permanently ulta karna |
| `sort()` | permanently sort (in place) |
| `copy()` | naye address par copy |

````python
# 1:- len / min / max / sorted
l = [2, 3, 7, 4]
print(len(l))                   # 4
print(min(l))                   # 2
print(max(l))                   # 7
# sorted() ek NAYI sorted list return karta hai -> original 'l' waisa ka waisa rehta hai
print(sorted(l))                # ascending (default)
print(sorted(l, reverse=True))  # descending
````

````python
# 2:- count -> koi value kitni baar aayi
l = [1, 2, 1, 3, 4, 1, 5]
print(l.count(1))   # 3
print(l.count(5))   # 1
print(l.count(3))   # 1
````

````python
# 3:- index -> kisi value ka PEHLA occurrence kis position par hai
l = [1, 2, 3, 4, 1, 5, 3]
print(l.index(2))   # 1
print(l.index(1))   # 0  -> multiple ho to sirf pehla index milta hai
````

````python
# 4:- reverse -> list ko IN PLACE ulta karta hai (permanent change)
l = [2, 1, 5, 7, 8]
l.reverse()
print(l)   # [8, 7, 5, 1, 2]
````

````python
# 5:- sort vs sorted
# sort()   -> list ko PERMANENTLY change karta hai (in place)
# sorted() -> ek NAYI list return karta hai, original waisa hi rehta hai
l = [2, 1, 5, 7, 8]
print(l)           # original
print(sorted(l))   # nayi sorted list
print(l)           # original abhi bhi same
l.sort()
print(l)           # ab permanently sorted
````

````python
# 6:- copy -> ek NAYE memory address par copy banata hai (shallow copy)
l = [2, 1, 5, 7, 8]
print(l)
print(id(l))       # l ka memory address
l1 = l.copy()
print(l1)
print(id(l1))      # alag address -> ye ek alag copy hai
````

## List comprehension

List comprehension lists banane ka ek **concise (chhota aur clean)** tarika deta hai.

```
newlist = [expression for item in iterable if condition == True]
```

**Advantages:**
- Loops ke comparison me zyada time-efficient aur space-efficient.
- Kam lines of code.
- Ek iterative statement ko ek chhote formule me badal deta hai.

````python
# 1 se 10 tak numbers ek list me daalo -> lamba tarika
l = []
for i in range(1, 11):
    l.append(i)
print(l)

# wahi kaam list comprehension se (bahut chhota)
l = [i for i in range(1, 11)]   # 'i' expression hai, range(...) iterable hai
print(l)
````

````python
# scalar multiplication on a vector
v = [2, 3, 4]
s = -3
print(v * s)   # [] -> list * negative number empty list deta hai (jo hume nahi chahiye)

# hume [-6, -9, -12] chahiye -> loop se
x = []
for i in v:
    x.append(i * s)
print(x)

# wahi kaam list comprehension se
print([s * i for i in v])   # expression s*i hai, iterable v hai
````

````python
# har element ka square
l = [1, 2, 3, 4, 5]
print([i ** 2 for i in l])
````

````python
# 1 se 50 tak jo numbers 5 se divisible hai
print([i for i in range(1, 51) if i % 5 == 0])
````

````python
# aise languages jo 'p' se start hote hai
languages = ["java", "python", "php", "javascript", "c"]
print([language for language in languages if language.startswith("p")])
````

````python
# nested 'if' list comprehension me
basket = ["apple", "guava", "cherry", "banana"]
my_fruits = ["apple", "kiwi", "grapes", "banana"]
# wahi fruits rakho jo basket me bhi hai AUR 'a' se start hote hai
print([fruit for fruit in my_fruits if fruit in basket if fruit.startswith("a")])
````

````python
# 3x3 matrix -> nested list comprehension
print([[i * j for i in range(1, 4)] for j in range(1, 4)])
````

````python
# cartesian-style product -> do lists par ek saath comprehension
l1 = [1, 2, 3, 4]
l2 = [5, 6, 7, 8]
print([i * j for i in l1 for j in l2])
````

## 2 ways to traverse a list

1. **Itemwise** — seedha items par loop chalao.
2. **Indexwise** — index par loop chalao, phir `l[i]` se item access karo.

````python
# itemwise -> seedha items par loop
l = [1, 2, 3, 4]
for i in l:
    print(i)

# indexwise -> index par loop, phir l[i] se item
l = [1, 2, 3, 4]
for i in range(0, len(l)):
    print(l[i])
````

## zip

`zip()` function ek **zip object** return karta hai, jo tuples ka ek iterator hota hai —
jahan har passed iterator ka pehla item ek saath pair hota hai, phir doosra item, aur aise hi
aage. Agar passed iterators ki length alag-alag ho, to jiski length **sabse kam** hai wahi
new iterator ki length decide karta hai.

````python
# do lists ke items ko indexwise add karna
l1 = [1, 2, 3, 4]
l2 = [-1, -2, -3, -4]
# indexwise pair banao, phir har pair ko add karo -> zip() ye aasaan bana deta hai
print(list(zip(l1, l2)))                 # [(1, -1), (2, -2), (3, -3), (4, -4)]
print([i + j for i, j in zip(l1, l2)])   # [0, 0, 0, 0]
````

## Disadvantages of Python lists

- **Slow** — arrays ke comparison me dheemi.
- **Risky usage** — kai types mix karne se subtle bugs aa sakte hai.
- **Eats up more memory** — zyada space leti hai.
