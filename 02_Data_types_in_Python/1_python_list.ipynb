## What are lists?
A **list** is a data type where you can store **multiple items under one name**.
More technically, lists behave like **dynamic arrays** — you can keep adding items *on the fly*.

**Real-world example:** suppose you are analysing the sentiment of movie reviews and you have
10,000 movies. Making a separate variable for each is impossible and wasteful — instead we
store them all in a single list.

> A list can hold **any type** of data — numbers, strings, booleans, even other lists — all
> together. Waise aksar hum ek jaise (similar) items ek list me rakhte hai, but Python isko
> force nahi karta.

```python
l = [20, "jessica", 35.5, [30, 60, 70]]   # mixed types, all in one list
```

## Arrays vs Lists
- **Fixed vs Dynamic size** — an array needs its size declared upfront, whereas a list can grow
  without limit. Internally a *dynamic array* starts with a fixed size; when it fills up, a new
  array of **double** the size is created, old items are copied over, and the process repeats.
  (Yehi reason hai lists "unlimited" feel karti hai.)
- **Homogeneous vs Heterogeneous** — arrays store only **one** data type (homogeneous), while
  lists can mix types (heterogeneous).
- **Speed** — arrays execute **faster** than lists.
- **Memory** — lists occupy **more** memory than arrays.

## How are lists stored in memory?
An array (e.g. `int arr[50]`) stores its elements in one **contiguous** block of memory.
A Python list instead stores a contiguous block of **references (addresses)** — each slot points
to the actual object living elsewhere in memory. Isi wajah se ek hi list alag-alag types ke
objects rakh sakti hai.

## Characteristics of a list
- **Ordered** — items keep their position.
- **Changeable / Mutable** — items can be changed after creation.
- **Heterogeneous** — can store multiple different data types.
- **Can have duplicates.**
- **Dynamic** — grows automatically as you add items.
- **Can be nested** — a list inside a list.
- **Items can be accessed** — by index / slice.
- **Can contain any kind of Python object.**

````python
# ORDERED -> order matters in a list
l1 = [1, 2, 3]
l2 = [3, 2, 1]
print(l1 == l2)   # False -> same items, but different order
````

## Creating a list

````python
# empty list
print([])
# 1D list -> homogeneous (all items same type)
print([1, 2, 3, 4, 5])
# 2D / nested list -> a list that contains at least one more list
print([1, 2, 3, [4, 5]])
# 3D list -> list inside a list inside a list
print([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
# heterogeneous list -> mixed data types
print([1, True, 5, 6, 5 + 6j, "hello"])
# using type conversion -> break a string into a list of characters
print(list("hello"))
````

## Accessing items from a list
1. Indexing
2. Slicing

````python
# 1:- INDEXING
# Positive indexing -> left to right, starts at 0
l = [1, 2, 3, 4, 5]
print(l[0])            # 1

# accessing inside a 2D list
l1 = [1, 2, 3, [4, 5]]
print(l1[3][0])        # 4  -> go to index 3 (the inner list), then its index 0

# accessing inside a 3D list
l3 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(l3[0][0][1])     # 2

# Negative indexing -> right to left, starts at -1
print(l[-3])           # 3

# negative indexing inside a 2D list
l2 = [1, 2, 3, [4, 5]]
print(l2[-1])          # [4, 5] -> last item is itself a list
print(l2[-1][-2])      # 4      -> go one level deeper
````

````python
# 2:- SLICING -> extract multiple items at once  ->  l[start : end : step]
l = [1, 2, 3, 4, 5, 6]
print(l[0:3])       # [1, 2, 3]
print(l[-3:])       # [4, 5, 6]   (negative)
print(l[0::2])      # [1, 3, 5]   (jump by 2)
print(l[-5:-2:2])   # [2, 4]      (negative + step)
print(l[::-1])      # reverse the list
````

## Adding items to a list
There are 3 ways to add items:
1. `append`  2. `extend`  3. `insert`

````python
# 1:- APPEND -> adds only a SINGLE item at the end (one at a time)
l = [1, 2, 3, 4, 5]
l.append(6)
print(l)
l.append(True)
print(l)
# appending a list adds it as ONE nested item (it is not merged in)
a = [1, 2, 3, 4, 5]
a.append([6, 7, 8])
print(a)

# 2:- EXTEND -> adds MULTIPLE items at once (more powerful than append)
s = [1, 2, 3, 4, 5]
s.extend([6, 7, 8])
print(s)
k = [1, 2, 3, 4, 5]
k.extend("delhi")     # a string is iterable -> each character is added separately
print(k)

# 3:- INSERT -> add at a specific position: insert(index, item)
d = [1, 2, 3, 4, 5]
d.insert(1, 100)      # put 100 at index 1
print(d)
````

## Editing items in a list
Editing is possible in a list (mutable), unlike a string (immutable).

````python
# editing with indexing
l = [1, 2, 3, 4, 5]
l[-1] = 500
print(l)
# editing with slicing -> replace a whole range at once
l[1:4] = [200, 300, 400]
print(l)
````

## Deleting items from a list
Since a list is mutable, we can delete items — one by one, several at once, or the whole list.

````python
# 1:- del  (delete by indexing or slicing)
l = [1, 2, 3, 4, 5]
print(l)
del l[-1]        # delete the last item
print(l)
del l[1:3]       # delete a slice
print(l)
del l            # delete the whole name 'l'
print(l)         # NameError -> 'l' ab exist nahi karta
````

````python
# 2:- remove -> delete by VALUE (no index needed)
l = [1, 2, 3, 4, 5]
l.remove(5)
print(l)
````

````python
# 3:- pop -> works in 2 ways
l = [1, 2, 3, 4, 5]
l.pop(0)      # (a) delete the item at a given index
print(l)
l.pop()       # (b) no index -> deletes the LAST item
print(l)
````

````python
# 4:- clear -> empties the list (keeps the list object, removes all items)
l = [1, 2, 3, 4, 5]
l.clear()
print(l)      # []
````

## Operations on a list
1. Arithmetic  2. Membership  3. Loop

````python
# 1:- ARITHMETIC (+ and *) -> only these two work on lists
l1 = [1, 2, 3, 4]
l2 = [5, 6, 7, 8]
print(l1 + l2)   # concatenation / merge
print(l1 * 3)    # repeat 3 times
````

````python
# 2:- MEMBERSHIP OPERATORS ( in / not in )
l1 = [1, 2, 3, 4, 5]
l2 = [1, 2, 3, 4, [5, 6]]
print(5 in l1)        # True
print(5 in l2)        # False -> l2 me 5 nahi hai, [5, 6] hai
print([5, 6] in l2)   # True
````

````python
# 3:- LOOPS -> mainly used with sequence data types
l1 = [1, 2, 3, 4, 5]
for i in l1:
    print(i)

# looping a 2D list -> each item can itself be a list
l2 = [1, 2, 3, 4, [5, 6]]
for i in l2:
    print(i)

# looping a 3D list
l3 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
for i in l3:
    print(i)
````

## List functions
Some functions (`len`, `min`, `max`, `sorted`) are **universal** — available on lists, sets,
tuples, dictionaries, etc. We have already used a few list methods: `clear()`, `pop()`,
`remove()`, `append()`, `extend()`, `insert()`.

Here we cover the rest:
1. `len` / `min` / `max` / `sorted`
2. `count`
3. `index`
4. `reverse`
5. `sort` (vs `sorted`)
6. `copy`

````python
# 1:- len / min / max / sorted
l = [2, 3, 7, 4]
print(len(l))                   # 4
print(min(l))                   # 2
print(max(l))                   # 7
# sorted() returns a NEW sorted list -> the original 'l' stays the same
print(sorted(l))                # ascending (default)
print(sorted(l, reverse=True))  # descending
````

````python
# 2:- count -> how many times a value appears
l = [1, 2, 1, 3, 4, 1, 5]
print(l.count(1))   # 3
print(l.count(5))   # 1
print(l.count(3))   # 1
````

````python
# 3:- index -> position of the FIRST occurrence of a value
l = [1, 2, 3, 4, 1, 5, 3]
print(l.index(2))   # 1
print(l.index(1))   # 0  -> agar multiple ho to sirf pehla index milta hai
````

````python
# 4:- reverse -> reverses the list IN PLACE (permanent change)
l = [2, 1, 5, 7, 8]
l.reverse()
print(l)   # [8, 7, 5, 1, 2]
````

````python
# 5:- sort vs sorted
# sort()   -> changes the list PERMANENTLY (in place)
# sorted() -> returns a NEW list, original stays same
l = [2, 1, 5, 7, 8]
print(l)           # original
print(sorted(l))   # new sorted list
print(l)           # original still unchanged
l.sort()
print(l)           # now permanently sorted
````

````python
# 6:- copy -> makes a copy at a NEW memory address (a shallow copy)
l = [2, 1, 5, 7, 8]
print(l)
print(id(l))       # memory address of l
l1 = l.copy()
print(l1)
print(id(l1))      # a DIFFERENT address -> it is a separate copy
````

## List comprehension
List comprehension gives a **concise** way to create lists.

```
newlist = [expression for item in iterable if condition == True]
```

**Advantages**
- More time- and space-efficient than loops.
- Fewer lines of code.
- Turns an iterative statement into a compact formula.

````python
# add numbers 1 to 10 into a list -> the long way
l = []
for i in range(1, 11):
    l.append(i)
print(l)

# the same thing with list comprehension (much shorter)
l = [i for i in range(1, 11)]   # 'i' is the expression, range(...) is the iterable
print(l)
````

````python
# scalar multiplication on a vector
v = [2, 3, 4]
s = -3
print(v * s)   # [] -> list * negative number gives an empty list (NOT what we want)

# the desired output is [-6, -9, -12] -> use a loop
x = []
for i in v:
    x.append(i * s)
print(x)

# same thing with list comprehension
print([s * i for i in v])   # expression is s*i, iterable is v
````

````python
# square each element
l = [1, 2, 3, 4, 5]
print([i ** 2 for i in l])
````

````python
# all numbers divisible by 5 in the range 1 to 50
print([i for i in range(1, 51) if i % 5 == 0])
````

````python
# find languages that start with the letter 'p'
languages = ["java", "python", "php", "javascript", "c"]
print([language for language in languages if language.startswith("p")])
````

````python
# nested 'if' in a list comprehension
basket = ["apple", "guava", "cherry", "banana"]
my_fruits = ["apple", "kiwi", "grapes", "banana"]
# keep fruits that are in the basket AND start with 'a'
print([fruit for fruit in my_fruits if fruit in basket if fruit.startswith("a")])
````

````python
# a 3x3 matrix using nested list comprehension
print([[i * j for i in range(1, 4)] for j in range(1, 4)])
````

````python
# cartesian-style product -> comprehension over two lists together
l1 = [1, 2, 3, 4]
l2 = [5, 6, 7, 8]
print([i * j for i in l1 for j in l2])
````

## Two ways to traverse a list
1. Item-wise
2. Index-wise

````python
# item-wise -> loop directly over the items
l = [1, 2, 3, 4]
for i in l:
    print(i)

# index-wise -> loop over indices, then access the item with l[i]
l = [1, 2, 3, 4]
for i in range(0, len(l)):
    print(l[i])
````

## zip
`zip()` returns a **zip object** (an iterator of tuples), where the first items of each iterable
are paired together, then the second items, and so on.

If the iterables have **different lengths**, the shortest one decides the length of the result.

````python
# WAP to add items of two lists index-wise
l1 = [1, 2, 3, 4]
l2 = [-1, -2, -3, -4]
# pair them index-wise, then add each pair -> zip() makes this easy
print(list(zip(l1, l2)))                 # [(1, -1), (2, -2), (3, -3), (4, -4)]
print([i + j for i, j in zip(l1, l2)])   # [0, 0, 0, 0]
````
