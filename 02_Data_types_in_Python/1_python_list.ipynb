## 🧱 What are lists?

**In simple words:** a list keeps **many values together under one name**, so you don't have to
make a separate variable for each thing.

🚂 **Analogy:** a list is like a *train*. Each **coach** holds one item, the coaches stay **in
order**, and you can **attach new coaches** whenever you want.

💡 Technically, a list is a **dynamic array** — it grows on its own as you add items.

**Why we need it —** imagine analysing 10,000 movie reviews. Making 10,000 variables is
impossible; one list holds them all neatly:

```python
reviews = ["good", "bad", "average", ...]   # 10,000 items, one name
```

> 🔑 A list can mix **any** data types together:
> ```python
> l = [20, "jessica", 35.5, [30, 60, 70]]   # int, str, float, even another list!
> ```

<a id="lists-vs-arrays"></a>
## ⚖️ Lists vs Arrays

| Feature | Array | List |
|---|---|---|
| **Size** | fixed (declare upfront) | dynamic (grows on its own) |
| **Data types** | homogeneous (one type only) | heterogeneous (mix freely) |
| **Speed** | ⚡ faster | slower |
| **Memory** | light | 🐘 heavier |

💡 **How "dynamic" actually works:** a dynamic array starts small; when it fills up, Python
quietly makes a **new array double the size**, copies everything over, and keeps going.
Isliye aapko kabhi list ka size batane ki zaroorat nahi padti.

<a id="memory"></a>
## 🧠 How lists live in memory

An **array** stores its values **side by side** in one continuous block.

A **Python list** stores a block of **addresses** 📍 — each slot just *points* to the real
object sitting somewhere else in memory.

> 🔑 That "pointing" trick is exactly **why one list can hold many different types** at once.

<a id="characteristics"></a>
## ✨ Characteristics of a list

| # | Property | Meaning |
|---|---|---|
| 1 | **Ordered** | items keep their position |
| 2 | **Mutable** | items can be changed after creation |
| 3 | **Heterogeneous** | can store different data types together |
| 4 | **Allows duplicates** | the same value can appear many times |
| 5 | **Dynamic** | grows automatically as you add items |
| 6 | **Nestable** | a list can sit inside another list |
| 7 | **Accessible** | any item can be reached by index / slice |
| 8 | **Holds any object** | numbers, strings, other lists… anything |

👇 Let's prove **"ordered"** with a tiny example:

````python
# ORDERED -> order matters in a list
l1 = [1, 2, 3]
l2 = [3, 2, 1]
print(l1 == l2)   # False -> same items, but different order
````

<a id="creating"></a>
## 🛠️ Creating a list

A list can take many shapes — empty, 1D, nested (2D/3D), mixed, or built from another
sequence. Neeche har type ek saath dekhte hai:

````python
# empty list
print([])
# 1D list -> homogeneous (all items same type)
print([1, 2, 3, 4, 5])
# 2D / nested list -> a list that contains at least one more list
print([1, 2, 3, [4, 5]])
# 3D list -> list inside a list inside a list
print([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
# heterogeneous list -> mixed data types
print([1, True, 5, 6, 5 + 6j, "hello"])
# using type conversion -> break a string into a list of characters
print(list("hello"))
````

<a id="accessing"></a>
## 🎯 Accessing items — Indexing & Slicing

Two ways to pull items out:

1. **Indexing** → grab **one** item by its position.
2. **Slicing** → grab **many** items at once.

🔢 **Index map** for `["a", "b", "c", "d"]`:

| item | a | b | c | d |
|---|---|---|---|---|
| **positive** | 0 | 1 | 2 | 3 |
| **negative** | -4 | -3 | -2 | -1 |

> 💡 Positive index counts **left → right (from 0)**, negative counts **right → left (from -1)**.

````python
# 1:- INDEXING
# Positive indexing -> left to right, starts at 0
l = [1, 2, 3, 4, 5]
print(l[0])            # 1

# accessing inside a 2D list
l1 = [1, 2, 3, [4, 5]]
print(l1[3][0])        # 4  -> go to index 3 (the inner list), then its index 0

# accessing inside a 3D list
l3 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(l3[0][0][1])     # 2

# Negative indexing -> right to left, starts at -1
print(l[-3])           # 3

# negative indexing inside a 2D list
l2 = [1, 2, 3, [4, 5]]
print(l2[-1])          # [4, 5] -> last item is itself a list
print(l2[-1][-2])      # 4      -> go one level deeper
````

**Slicing** grabs a range at once → `l[start : end : step]`
(the `end` index is **not** included).

````python
# 2:- SLICING -> extract multiple items at once  ->  l[start : end : step]
l = [1, 2, 3, 4, 5, 6]
print(l[0:3])       # [1, 2, 3]
print(l[-3:])       # [4, 5, 6]   (negative)
print(l[0::2])      # [1, 3, 5]   (jump by 2)
print(l[-5:-2:2])   # [2, 4]      (negative + step)
print(l[::-1])      # reverse the whole list
````

<a id="adding"></a>
## ➕ Adding items to a list

| Method | What it does | Adds |
|---|---|---|
| `append(x)` | attach **one** item at the end | a single item |
| `extend([...])` | attach **many** items at the end | each item separately |
| `insert(i, x)` | drop an item at position `i` | a single item |

🚂 In train terms: `append` = one coach at the back, `extend` = a whole rake of coaches at
the back, `insert` = squeezing a coach into the middle.

````python
# 1:- APPEND -> adds only a SINGLE item at the end (one at a time)
l = [1, 2, 3, 4, 5]
l.append(6)
print(l)
l.append(True)
print(l)
# appending a list adds it as ONE nested item (it is not merged in)
a = [1, 2, 3, 4, 5]
a.append([6, 7, 8])
print(a)

# 2:- EXTEND -> adds MULTIPLE items at once (more powerful than append)
s = [1, 2, 3, 4, 5]
s.extend([6, 7, 8])
print(s)
k = [1, 2, 3, 4, 5]
k.extend("delhi")     # a string is iterable -> each character is added separately
print(k)

# 3:- INSERT -> add at a specific position: insert(index, item)
d = [1, 2, 3, 4, 5]
d.insert(1, 100)      # put 100 at index 1
print(d)
````

<a id="editing"></a>
## ✏️ Editing items

Because lists are **mutable**, you can overwrite items — a single index or a whole slice.
(Strings can't do this, kyunki wo **immutable** hoti hai.)

````python
# editing with indexing
l = [1, 2, 3, 4, 5]
l[-1] = 500
print(l)
# editing with slicing -> replace a whole range at once
l[1:4] = [200, 300, 400]
print(l)
````

<a id="deleting"></a>
## 🗑️ Deleting items

| Way | Deletes by | Note |
|---|---|---|
| `del` | index / slice | can also delete the whole variable |
| `remove(x)` | **value** | removes the first match |
| `pop(i)` | index | **returns** the removed item; default = last |
| `clear()` | — | empties the list but keeps it |

⚠️ `del l` removes the **entire name** — using `l` afterwards gives a `NameError`
(dekhte hai neeche 👇).

````python
# 1:- del  (delete by indexing or slicing)
l = [1, 2, 3, 4, 5]
print(l)
del l[-1]        # delete the last item
print(l)
del l[1:3]       # delete a slice
print(l)
del l            # delete the whole name 'l'
print(l)         # NameError -> 'l' ab exist nahi karta
````

````python
# 2:- remove -> delete by VALUE (no index needed)
l = [1, 2, 3, 4, 5]
l.remove(5)
print(l)
````

````python
# 3:- pop -> works in 2 ways
l = [1, 2, 3, 4, 5]
l.pop(0)      # (a) delete the item at a given index
print(l)
l.pop()       # (b) no index -> deletes the LAST item
print(l)
````

````python
# 4:- clear -> empties the list (keeps the list object, removes all items)
l = [1, 2, 3, 4, 5]
l.clear()
print(l)      # []
````

<a id="operations"></a>
## 🔧 Operations on a list

1. **Arithmetic** → `+` (join) and `*` (repeat)
2. **Membership** → `in` / `not in`
3. **Loops** → walk through every item

````python
# 1:- ARITHMETIC (+ and *) -> only these two work on lists
l1 = [1, 2, 3, 4]
l2 = [5, 6, 7, 8]
print(l1 + l2)   # concatenation / merge
print(l1 * 3)    # repeat 3 times
````

````python
# 2:- MEMBERSHIP OPERATORS ( in / not in )
l1 = [1, 2, 3, 4, 5]
l2 = [1, 2, 3, 4, [5, 6]]
print(5 in l1)        # True
print(5 in l2)        # False -> l2 me 5 nahi hai, [5, 6] hai
print([5, 6] in l2)   # True
````

````python
# 3:- LOOPS -> mainly used with sequence data types
l1 = [1, 2, 3, 4, 5]
for i in l1:
    print(i)

# looping a 2D list -> each item can itself be a list
l2 = [1, 2, 3, 4, [5, 6]]
for i in l2:
    print(i)

# looping a 3D list
l3 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
for i in l3:
    print(i)
````

<a id="functions"></a>
## 📦 List functions

Some functions are **universal** (work on lists, tuples, sets, dicts): `len`, `min`, `max`,
`sorted`. We've already met `append`, `extend`, `insert`, `remove`, `pop`, `clear`.
Here are the remaining ones:

| Function | Purpose |
|---|---|
| `count(x)` | how many times `x` appears |
| `index(x)` | position of the **first** `x` |
| `reverse()` | flip the order **permanently** |
| `sort()` | sort **permanently** (in place) |
| `sorted(l)` | return a **new** sorted list (original untouched) |
| `copy()` | make a separate copy at a new address |

````python
# 1:- len / min / max / sorted
l = [2, 3, 7, 4]
print(len(l))                   # 4
print(min(l))                   # 2
print(max(l))                   # 7
# sorted() returns a NEW sorted list -> the original 'l' stays the same
print(sorted(l))                # ascending (default)
print(sorted(l, reverse=True))  # descending
````

````python
# 2:- count -> how many times a value appears
l = [1, 2, 1, 3, 4, 1, 5]
print(l.count(1))   # 3
print(l.count(5))   # 1
print(l.count(3))   # 1
````

````python
# 3:- index -> position of the FIRST occurrence of a value
l = [1, 2, 3, 4, 1, 5, 3]
print(l.index(2))   # 1
print(l.index(1))   # 0  -> agar multiple ho to sirf pehla index milta hai
````

````python
# 4:- reverse -> reverses the list IN PLACE (permanent change)
l = [2, 1, 5, 7, 8]
l.reverse()
print(l)   # [8, 7, 5, 1, 2]
````

````python
# 5:- sort vs sorted
# sort()   -> changes the list PERMANENTLY (in place)
# sorted() -> returns a NEW list, original stays same
l = [2, 1, 5, 7, 8]
print(l)           # original
print(sorted(l))   # new sorted list
print(l)           # original still unchanged
l.sort()
print(l)           # now permanently sorted
````

````python
# 6:- copy -> makes a copy at a NEW memory address (a shallow copy)
l = [2, 1, 5, 7, 8]
print(l)
print(id(l))       # memory address of l
l1 = l.copy()
print(l1)
print(id(l1))      # a DIFFERENT address -> it is a separate copy
````

## ⚡ List comprehension

A **one-line factory** for building lists — shorter and faster than a `for` loop.

```
newlist = [ expression   for item in iterable   if condition ]
              ↑ what to keep     ↑ the loop          ↑ optional filter
```

**Why use it?** fewer lines, faster to run, and it reads almost like a math formula.

````python
# add numbers 1 to 10 into a list -> the long way
l = []
for i in range(1, 11):
    l.append(i)
print(l)

# the same thing with list comprehension (much shorter)
l = [i for i in range(1, 11)]   # 'i' is the expression, range(...) is the iterable
print(l)
````

````python
# scalar multiplication on a vector
v = [2, 3, 4]
s = -3
print(v * s)   # [] -> list * negative number gives an empty list (NOT what we want)

# the desired output is [-6, -9, -12] -> use a loop
x = []
for i in v:
    x.append(i * s)
print(x)

# same thing with list comprehension
print([s * i for i in v])   # expression is s*i, iterable is v
````

````python
# square each element
l = [1, 2, 3, 4, 5]
print([i ** 2 for i in l])
````

````python
# all numbers divisible by 5 in the range 1 to 50
print([i for i in range(1, 51) if i % 5 == 0])
````

````python
# find languages that start with the letter 'p'
languages = ["java", "python", "php", "javascript", "c"]
print([language for language in languages if language.startswith("p")])
````

````python
# nested 'if' in a list comprehension
basket = ["apple", "guava", "cherry", "banana"]
my_fruits = ["apple", "kiwi", "grapes", "banana"]
# keep fruits that are in the basket AND start with 'a'
print([fruit for fruit in my_fruits if fruit in basket if fruit.startswith("a")])
````

````python
# a 3x3 matrix using nested list comprehension
print([[i * j for i in range(1, 4)] for j in range(1, 4)])
````

````python
# cartesian-style product -> comprehension over two lists together
l1 = [1, 2, 3, 4]
l2 = [5, 6, 7, 8]
print([i * j for i in l1 for j in l2])
````

<a id="traversing"></a>
## 🔁 Two ways to traverse a list

1. **Item-wise** → loop straight over the items (simple & clean).
2. **Index-wise** → loop over positions, then read `l[i]` (handy when you also need the index).

````python
# item-wise -> loop directly over the items
l = [1, 2, 3, 4]
for i in l:
    print(i)

# index-wise -> loop over indices, then access the item with l[i]
l = [1, 2, 3, 4]
for i in range(0, len(l)):
    print(l[i])
````

<a id="zip"></a>
## 🤐 zip()

`zip()` joins iterables **tooth by tooth**, like a zipper 🤐 — first items together, second
items together, and so on. It returns pairs (tuples).

> ⚠️ If the lists have **different lengths**, the **shortest** one decides where it stops.

````python
# WAP to add items of two lists index-wise
l1 = [1, 2, 3, 4]
l2 = [-1, -2, -3, -4]
# pair them index-wise, then add each pair -> zip() makes this easy
print(list(zip(l1, l2)))                 # [(1, -1), (2, -2), (3, -3), (4, -4)]
print([i + j for i, j in zip(l1, l2)])   # [0, 0, 0, 0]
````
