# Accessing items from a lists
## 1:- indexinng
## 2 :- slicing

```python
# 1:- indexing
# positive indexing(no.1 is always move from left to right and no.2 is always start with 0)
l=[1,2,3,4,5]
print(l[0])
# what if we extract positive 2D lists
l1=[1,2,3,[4,5]]
print(l1[3][0])
# what if we have positive 3D lists and extracting it
l3=[[[1,2],[3,4]],[[5,6],[7,8]]]
print(l3[0][0][1])
# negative indexing(move from right to left and always start with -1)
print(l[-3])
# what if we extract negative 2D list
l2=[1,2,3,[4,5]]
print(l2[-1]) # here we will get last list as a output but this is not true
print(l2[-1][-2])
```

```python
# 2:-Slicing( list k andar se 1 saath multiple items ko extract karna chahte hai )
# positive slicing
l=[1,2,3,4,5,6]
print(l[0:3])
# negative slicing
print(l[-3:])
# if we want to jump
print(l[0::2])  #(positive)
print(l[-5:-2:2]) #(negative)
# reversing a list
print(l[::-1])
```

```python
# 1:- append
# 1 baar me 1 items ko list k end me add kar sakte hai(append will try to add only
#single items at once).
l=[1,2,3,4,5]
l.append(6)
print(l)
l.append(True)
print(l)
# what will happen if we add the list with the help of append
a=[1,2,3,4,5]
a.append([6,7,8])
print(a)
# 2:- extend
# 1 saath multiple items ko add kar sakte hai(and extend is more powerful than append)
s=[1,2,3,4,5]
s.extend([6,7,8])
print(s)
k=[1,2,3,4,5]
k.extend("delhi") # here extend will convert strings into list
print(k)
# what if we want to add the items as per my desire for that we have
# 3:- insert(isme only two things provide karna hota 1st index position jahan
# pe hum items ko add karna chahte hai,and 2nd what i want to add.)
d=[1,2,3,4,5]
d.insert(1,100)
print(d)
```

# Editing items in a List
### in a list editing is possible but not in a string as it is immutable

```python
# editing with indexing
l=[1,2,3,4,5]
l[-1]=500
print(l)
# editing with slicing
l[1:4]=[200,300,400]
print(l)
```

# Deleting items from a List
### since list is mutable so that we can delete from the list.we can delete items one by one or we can delete multiple items as well or we can also delete the list

```python
# 1:- del
# (delete by indexing or slicing)
l=[1,2,3,4,5]
print(l)
# by indexing
del l[-1]
print(l)
# by slicing
del l[1:3]
print(l)
del(l) # we will talk later this list is not deleting at memory level
print(l)
```

```python
# 2:- remove
#(in this function we don't need to tell the index function,we can
# directly delete whatever we want to delete.)
l=[1,2,3,4,5]
l.remove(5)
print(l)
```

```python
# 3:- pop
# pop works in 2 ways 1.kissi particular index pe rakhe hue items ko delete kar
# sakta hai
l=[1,2,3,4,5]
l.pop(0)
print(l)
# 2nd ways if we not provide the the index then automatically pop will delete
# from the last,generally last item ko delete karne k lye use hota hai
l.pop()
print(l)
```

```python
# 4:- clear
# list ko delete nhi karta hai but uss list k andar k saare items ko delte krta hai.
l=[1,2,3,4,5]
l.clear()
print(l)
```

# operators on list
### 1:-Arithmatic

```python
# 1:- Arithmatic(+,*),only two operators works
l1=[1,2,3,4]
l2=[5,6,7,8]
# concatenation/merge
print(l1+l2)
print(l1*3)
```

```python
# 1:- len/min/max/sorted
l=[2,3,7,4]
print(len(l))
print(min(l))
print(max(l))
# not permanently sorted means l variable same as rahega
print(sorted(l))   # by default sorted in ascending order
print(sorted(l,reverse=True))   # in descending order
```

```python
# 2:- count
l=[1,2,1,3,4,1,5]
print(l.count(1))
print(l.count(5))
print(l.count(3))
```

```python
# 3:- index
# given items ka indexing batata hai
l=[1,2,3,4,1,5,3]
print(l.index(2))
print(l.index(1)) # multiple jagah pe hoga toh only 1st occurrence batata hai.
```

```python
# 4:- reverse
l=[2,1,5,7,8]
 # reverse ho gya permanently means list change ho gya
l.reverse()    # reverse ho gya but not permanently
```

```python
# 5:- sort vs sorted
# sort me permanently changes hota hai but sorted me nhi
l=[2,1,5,7,8]
print(l)
print(sorted(l))
print(l)
l.sort()
print(l)
```

```python
# 6:- copy (yeh shallow copy banata hai,memory me 1 new address pe 1 same list ban jata hai)
l=[2,1,5,7,8]
print(l)
print(id(l))
l1=l.copy()
print(l1)
print(id(l1))
```

# list comprehension
##list comprehension provides a concise way of creating lists.
###newlist=[expression for item in iterable if condition==True].

# Advantages of List comprehension
### . More time-efficient and space-efficient than loops.
### . Require fewer lines of code.
### . Transform iterative statement into a formula.

```python
# let us look how we can use list comprehension
# Add 1 to 10 numbers to a list.
l=[]  # created empty list
for i in range(1,11):
    l.append(i)
print(i)
# with the help of list comprehension we can make this programe much shorter
l=[i for i in range(1,11)]   # here 1st i is expression & iterable is range
print(l)
```

```python
# scalar multiplication on a vector
v=[2,3,4]
s= -3
print(v*s) # we will get empty list but we didn't get the desired output
# output should be [-6,-9,-12]
x=[]
for i in v:
    x.append(i*s)
print(x)
# with the help of list comprehension we can make it shorter
[s*i for i in v] # here expresssion is s*v and iterable is v
```

```python
# Add squares
l=[1,2,3,4,5]
[i**2 for i in l]
```

```python
# print all numbers divisible by 5 in the range of 1 to 50
[i for i in range(1,51) if i%5==0]
```

```python
# find languages which start with letter p
languages=["java","python","php","javascript","c"]
[language for language in languages if language.startswith("p")]
```

```python
# nested if with list comprehension
basket=["apple","guava","cherry","banana"]
my_fruits=["apple","kiwi","grapes","banana"]
# add new list from my_fruits and items if the fruits exists in basket and also starts with "a"
[fruit for fruit in my_fruits if fruit in basket if fruit.startswith("a")]
```

```python
# print(3,3),matrix using list comprehension -> nested list comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# a cartesian products -> list comprehension on 2 lists together
l1=[1,2,3,4]
l2=[5,6,7,8]
[i*j for i in l1 for j in l2]
```

# 2 ways to traverse a list
### 1:- itemwise
### 2:- indexwise

```python
# itemwise
l=[1,2,3,4]
for i in l:
    print(i)
# indexwise
l=[1,2,3,4]
for i in range(0,len(l)):
    print(i)
```

# zip
### the zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together and then the second.
### item in each passed iterator are paired together.
### if the passed iterators have different lengths,the iterator with the least items decides the length of the new iterator.

```python
# wap to add items of 2 lists indexwise
l1=[1,2,3,4]
l2=[-1,-2,-3,-4]
# two lists se humko 3rd list banana hai(means l1 ka first item ko l2 se add
# karna hai, phir l1 ka 2nd item ko l2 k 2nd item se add krenge and so on) basically indexwise
# item ko add ko karna hai or yahan pe zip() function help karega
list(zip(l1,l2))
[i+j for i,j in zip (l1,l2)]
```
