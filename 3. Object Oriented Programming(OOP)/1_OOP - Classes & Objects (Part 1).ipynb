## 3️⃣ What is a Class?

A **class** is a **blueprint** — a set of rules that describes how its objects will look and behave.

🧠 **Analogy:** the blueprint of a house is not a house itself — it's the *plan*. From one blueprint you can build many houses. Similarly, from one class you can create many objects.

Let's connect this to what we already know.

````python
l = [1, 2, 3, 4]
print(type(l))   # <class 'list'>
````

When we ask for the `type`, Python says **`<class 'list'>`**. So `list` is a **class**! And `l` (our variable) is an **object** of that class.

👉 So all the data types you already learned — `list`, `set`, `dict`, `int`, `float`, `tuple` — are actually **built-in classes** (someone already wrote them). Whenever you make a variable of that type, you are creating an **object** of that class.

**More real-life examples:**
- A **mentorship course** is a *class*; each **student** who joins is an *object*.
- **Car** is a *class*; **Maruti Suzuki** is an *object*.

## 4️⃣ What Goes Inside a Class?

A class describes **two** kinds of things:

| Inside a class | Meaning | Car example |
|----------------|---------|-------------|
| **Attributes** (data/properties) | the *information* | color, top speed, mileage |
| **Methods** (functions/behaviour) | the *actions* | calculate average speed |

**Basic structure of a class:**
```python
class Car:
    color = "blue"        # data (attribute)
    model = "sports"      # data (attribute)

    def cal_avg_speed(km, time):   # behaviour (method)
        # some code
```

## 5️⃣ What is an Object?

An **object** is an **instance** (a real example) created from a class.

- Class `Car` → object `wagonR` → `wagonR = Car()`
- Class `Sports` → object `gillidanda` → `gillidanda = Sports()`
- Class `Animals` → object `langoor` → `langoor = Animals()`

**Syntax to create an object:**
```python
object_name = ClassName()
```

You have actually been doing this already! `list()` and `str()` create objects.

````python
# Making a list the "long" way shows the class-object pattern:
l = list()   # l is the object, list is the class
print(l)

# similarly for strings:
s = str()    # s is the object, str is the class
print(repr(s))
````

> 💡 You normally write `l = [1, 2, 3]` (called an *object literal*) instead of `list()`, because Python gives common types a shortcut to make them easier to create.

### 🏷️ Naming a class — PascalCase

Class names use **PascalCase**: every word starts with a **capital letter** and words are joined together, e.g. `HelloWorld`, `BankAccount`, `Atm`.

## 6️⃣ Building an ATM Machine 🏧

Let's build a real mini-project: the software for an **ATM machine**. This will teach us classes, objects, the constructor, and `self` — all at once.

**What data (attributes) does an ATM need?**
1. **Balance** — money is withdrawn from, deposited to, and checked against the balance.
2. **PIN** — without a PIN you cannot do anything on an ATM.

**What actions (methods) does it need?**
- create a PIN, change the PIN, check the balance, withdraw money.

So our `Atm` class will have **2 attributes** (`pin`, `balance`) and several **methods**.

### 🔑 The constructor `__init__`

Before creating the two variables, we meet a special function called the **constructor**, written as `def __init__(self):`.

**Why is it special?** Normally you must *call* a function to run its code. But the **constructor runs automatically** the moment you create an object — you never call it yourself.

Let's prove it: the message inside `__init__` prints just by creating the object.

````python
# The constructor runs AUTOMATICALLY when the object is created
class Atm:
    def __init__(self):
        self.pin = ""
        self.balance = 0
        print("I (the constructor) ran automatically!")

    def menu(self):
        user_input = input("""
        Hi, how can I help you?
        1. Press 1 to create a pin
        2. Press 2 to change pin
        3. Press 3 to check balance
        4. Press 4 to withdraw
        5. Press anything else to exit
        """)

obj = Atm()   # notice: we never called __init__, yet it ran
````

### 🏗️ The full ATM class

Now the complete version. Notice how the constructor calls `self.menu()` automatically, and how each method uses `self` to reach the class's data and other methods.

> 📝 **Cleaned up:** the original had some rude Hindi messages for wrong input — I replaced them with polite English messages so the notes stay classroom-friendly.

````python
class Atm:
    # a non-parameterized constructor (it takes no extra input)
    def __init__(self):
        self.pin = ""
        self.balance = 0
        self.menu()          # the constructor calls the menu automatically

    def menu(self):
        user_input = input("""
        Hi, how can I help you?
        1. Press 1 to create a pin
        2. Press 2 to change pin
        3. Press 3 to check balance
        4. Press 4 to withdraw
        5. Press anything else to exit
        """)
        if user_input == "1":
            self.create_pin()
        elif user_input == "2":
            self.change_pin()
        elif user_input == "3":
            self.check_balance()
        elif user_input == "4":
            self.withdraw()
        else:
            exit()

    def create_pin(self):
        user_pin = input("enter your pin")
        self.pin = user_pin
        user_balance = int(input("enter your balance"))
        self.balance = user_balance
        print("pin created successfully")
        self.menu()

    def change_pin(self):
        old_pin = input("enter your old pin")
        if old_pin == self.pin:
            new_pin = input("enter new pin")
            self.pin = new_pin
            print("pin changed successfully")
            self.menu()
        else:
            print("old pin is incorrect, cannot change the pin")
            self.menu()

    def check_balance(self):
        user_pin = input("enter your pin")
        if user_pin == self.pin:
            print("your balance is", self.balance)
        else:
            print("incorrect pin!")

    def withdraw(self):
        user_pin = input("enter the pin")
        if user_pin == self.pin:
            amount = int(input("enter the amount"))
            # careful: don't let the user withdraw more than they have
            if amount <= self.balance:
                self.balance = self.balance - amount
                print("withdraw successful, balance is", self.balance)
            else:
                print("insufficient balance!")
        else:
            print("incorrect pin!")
            self.menu()
````

### ▶️ Running the ATM

Creating the object starts everything (because the constructor runs the menu). This is **interactive** — run it and type your choices.

````python
obj = Atm()
````

## 7️⃣ Method vs Function

This is a small but important vocabulary point:

| Term | Meaning |
|------|---------|
| **Method** | a function that is written **inside a class** |
| **Function** | a normal function that lives **outside** any class (independent) |

So the *same* piece of code is called a **method** if it's inside a class, and a **function** if it's on its own.

````python
l = [1, 2, 3]
len(l)        # len is a FUNCTION (it lives independently)
l.append(4)   # append is a METHOD (it lives inside the list class)
````

## 8️⃣ Magic Methods (Dunder Methods)

A **magic method** (also called a **dunder** method, short for "double underscore") is a special method whose name looks like `__name__` — with double underscores on both sides.

- Whenever you see something like `__init__`, `__str__`, `__add__` → it's a magic method.
- Each magic method has a special "superpower" and runs automatically in certain situations.
- The **constructor `__init__` is itself a magic method!**

We will soon use magic methods to build our **own data types**.

````python
# A tiny demo: __init__ (a magic method) runs automatically
class Temp:
    def __init__(self):
        print("hello")

obj = Temp()   # prints "hello" without us calling __init__
````

## 9️⃣ Constructor — Deep Dive

The **constructor** (`__init__`) is a method whose code runs **automatically** the instant an object is created.

### 🤔 Why is the constructor so useful?

In our ATM, methods like `withdraw` or `check_balance` only run **when the user chooses them** — the *user* is in control.

But some code must run **no matter what**, without waiting for the user. For example:
- connecting your app to a **database**,
- connecting to the **internet** or a backend service.

You cannot *rely on the user* to press a button for these. So you put such **configuration/setup code in the constructor** — it runs automatically, guaranteed, the moment the object is born. 🎯

## 🔟 The `self` Keyword ⭐

If you look back at the ATM code, `self` appears everywhere:
1. Every method has `self` as its **first parameter**.
2. Every attribute is written as `self.pin`, `self.balance`.
3. Every method call is written as `self.menu()`, `self.withdraw()`.

### 🏅 The golden rule of OOP

> **Whatever is inside a class (its data and its methods) can only be accessed by an *object* of that class.**

Even one method **cannot directly talk to** another method in the same class. So how did the constructor call `menu()`? The answer is **`self`**.

**`self` is the current object itself.** When a method needs to reach the class's data or call another method, it goes *through the object* — and inside the class, that object is `self`. Let's prove that `self` really is the object.

````python
class Atm:
    def __init__(self):
        print(id(self))   # print the memory address of self
        self.pin = ""
        self.balance = 0

    def menu(self):
        pass

obj = Atm()
print(id(obj))   # print the memory address of the object
````

Both `id(self)` and `id(obj)` are the **same number** — proving that **`self` and the object are the exact same thing** in memory. 🎯

> 📌 **Note:** the name `self` is just a convention. You *could* name it anything — but everyone uses `self`, so you should too.

## 1️⃣1️⃣ Building Our Own Data Type — a Fraction! 🧮

Here's the real payoff of OOP. Python has **no built-in fraction type**. Watch what happens with normal division:

````python
3/4 * 1/2   # we get a decimal, but we want the answer to stay a FRACTION
````

We want a proper **Fraction** data type that displays like `3/4` and supports `+`, `-`, `*`, `/` — all returning fractions. We build it with a class and **magic methods**:

| Magic method | Superpower (runs when you...) |
|--------------|-------------------------------|
| `__init__` | create the object (constructor) |
| `__str__` | `print()` the object |
| `__add__` | use `+` on two objects |
| `__sub__` | use `-` |
| `__mul__` | use `*` |
| `__truediv__` | use `/` |

> 🛠️ **Cleaned up:** the original had a stray `u` character that would break the class — I removed it.

````python
class Fraction:
    # a PARAMETERIZED constructor (it expects two inputs when the object is created)
    def __init__(self, x, y):
        self.num = x       # numerator (top)
        self.den = y       # denominator (bottom)

    # __str__ controls what print() shows
    def __str__(self):
        return "{}/{}".format(self.num, self.den)

    # __add__ runs when we use +  (adds two fractions)
    def __add__(self, other):
        new_num = self.num * other.den + other.num * self.den
        new_den = self.den * other.den
        return "{}/{}".format(new_num, new_den)

    # __sub__ runs when we use -
    def __sub__(self, other):
        new_num = self.num * other.den - other.num * self.den
        new_den = self.den * other.den
        return "{}/{}".format(new_num, new_den)

    # __mul__ runs when we use *
    def __mul__(self, other):
        new_num = self.num * other.num
        new_den = self.den * other.den
        return "{}/{}".format(new_num, new_den)

    # __truediv__ runs when we use /
    def __truediv__(self, other):
        new_num = self.num * other.den
        new_den = self.den * other.num
        return "{}/{}".format(new_num, new_den)

    # we can also add normal (non-magic) methods
    def convert_to_decimal(self):
        return self.num / self.den
````

````python
# Create two fraction objects
frac1 = Fraction(3, 4)
frac2 = Fraction(1, 2)
````

````python
# __str__ makes print() show them nicely as fractions
print(frac1)
print(frac2)
````

````python
# Now our magic methods make +, -, *, / all work on fractions!
print(frac1 + frac2)   # 3/4 + 1/2
print(frac1 - frac2)   # 3/4 - 1/2
print(frac1 * frac2)   # 3/4 * 1/2
print(frac1 / frac2)   # 3/4 / 1/2
````

````python
# our own normal method: convert the fraction to a decimal
frac1.convert_to_decimal()   # 3/4 = 0.75
````

🎉 **You just built your very own data type!** This is the true power of OOP — you are no longer limited to Python's built-in types; you can create your own with your own behaviour.
