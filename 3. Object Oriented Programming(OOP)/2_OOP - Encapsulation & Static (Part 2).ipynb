# 🅾️ Part 0 — Warm-up Project: Coordinate Geometry

Let's practise everything from Part 1 by solving a real geometry problem with classes. We need OOP classes that can:
1. Create and view **2D coordinates** (points).
2. Find the **distance between two points**.
3. Find a point's **distance from the origin**.
4. Check if a point **lies on a line**.
5. Find the **shortest distance** between a point and a line.

We build **two classes**: `Point` and `Line`.

````python
class Point:
    def __init__(self, x, y):
        self.x_cod = x
        self.y_cod = y

    # (1) __str__ decides how the point is shown when printed
    def __str__(self):
        return "<{},{}>".format(self.x_cod, self.y_cod)

    # (2) distance between THIS point and another point (Euclidean distance)
    def euclidean_distance(self, other):
        return ((self.x_cod - other.x_cod) ** 2 + (self.y_cod - other.y_cod) ** 2) ** 0.5

    # (3) distance from the origin (0, 0) — just reuse the method above
    def distance_from_origin(self):
        return self.euclidean_distance(Point(0, 0))


class Line:
    def __init__(self, A, B, C):     # a line is Ax + By + C = 0
        self.A = A
        self.B = B
        self.C = C

    def __str__(self):
        return "{}x + {}y + {} = 0".format(self.A, self.B, self.C)

    # (4) check whether a point lies on this line
    def point_on_line(line, point):
        if line.A * point.x_cod + line.B * point.y_cod + line.C == 0:
            return "lies on the line"
        else:
            return "does not lie on the line"

    # (5) shortest (perpendicular) distance from a point to this line
    def shortest_distance(line, point):
        return abs(line.A * point.x_cod + line.B * point.y_cod + line.C) / (line.A ** 2 + line.B ** 2) ** 0.5
````

**(1) Create and view points** — the `__str__` method makes them print nicely.

````python
p1 = Point(0, 0)
p2 = Point(-2, -1)
print(p1)
print(p2)
````

**(2) Distance between two points** — `(0,0)` to `(1,1)` is √2.

````python
p1 = Point(0, 0)
p2 = Point(1, 1)
p1.euclidean_distance(p2)
````

**(3) Distance from origin** — the point `(3,4)` is 5 units from `(0,0)`.

````python
p = Point(3, 4)
p.distance_from_origin()
````

**(4) Does a point lie on a line?**

````python
l1 = Line(1, 1, -2)   # the line x + y - 2 = 0
p1 = Point(1, 2)
print(l1)
print(p1)
l1.point_on_line(p1)   # 1 + 2 - 2 = 1, not 0 -> does not lie
````

**(5) Shortest distance from a point to a line.** *(The original had a typo — it used an undefined `p5`; I fixed it to `p1`.)*

````python
l1 = Line(1, 1, -2)
p1 = Point(1, 10)
print(l1)
print(p1)
l1.shortest_distance(p1)   # |1 + 10 - 2| / sqrt(1 + 1) = 9 / 1.414
````

# 1️⃣ How Objects Access Attributes

Once you have an object, you reach its **attributes** (data) and **methods** (actions) using the **dot `.`**.

````python
class Person:
    def __init__(self, name_input, country_input):
        self.name = name_input
        self.country = country_input

    def greet(self):
        # decide the greeting based on the country
        if self.country == "India":
            print("Namaste", self.name)
        else:
            print("Hello", self.name)
````

````python
# accessing ATTRIBUTES with the dot
p = Person("nitish", "India")
print(p.name)
print(p.country)
````

````python
# accessing a METHOD (note the brackets)
p.greet()
````

### Accessing something that doesn't exist → error

If you ask for an attribute the object doesn't have, Python raises an `AttributeError`.
```python
p.gender   # ❌ AttributeError: 'Person' object has no attribute 'gender'
```

### But you CAN add a new attribute from outside

Python lets you attach a brand-new attribute to an object from outside the class, just by assigning it.

````python
p.gender = "male"   # add a new attribute from outside the class
p.gender
````

# 2️⃣ Reference Variables

This is a subtle but very important idea:

1. A **reference variable holds an object** (actually, its memory address).
2. You can create objects **without** a reference variable.
3. One object can have **multiple** reference variables.
4. Assigning a new variable to an existing object does **not** create a new object.

> 💡 When you write `p = Person()`, `p` is **not** the object. `Person()` creates the object in memory, and `p` just **stores the address** of that object.

````python
class Person:
    def __init__(self):
        self.name = "nitish"
        self.country = "india"

# you can even create an object WITHOUT storing it in a variable:
Person()   # this makes an object; the output shows its memory location
````

````python
# one object, TWO reference variables pointing to it
p = Person()
q = p               # q now points to the SAME object as p
print(id(p))
print(id(q))        # same address -> same object
````

### ⚠️ Changing one reference changes them all

Because `p` and `q` point to the **same** object, editing through one is visible through the other. Be very careful with this!

````python
print(p.name)
print(q.name)       # both show "nitish"

q.name = "ankit"    # change it via q
print(p.name)
print(q.name)       # BOTH now show "ankit" (same object!)
````

# 3️⃣ Pass by Reference

You can pass an object **into** a function, and even **return** an object from a function.

````python
class Person:
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender

# a normal function (outside the class) that takes an object
def greet(person):
    print("hi my name is", person.name, "and i am a", person.gender)

p = Person("nitish", "male")
greet(p)            # pass the object into the function
````

````python
# a function can also RETURN an object
class Person:
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender

def greet(person):
    print("hi my name is", person.name, "and i am a", person.gender)
    p1 = Person("ankit", "male")
    return p1        # return a new object

p = Person("nitish", "male")
x = greet(p)         # x now holds the returned object
print(x.name)
print(x.gender)
````

### The object's address is the SAME inside and outside the function

When you pass an object to a function, Python does **not** copy it — it passes the **same address**. Proof: `id()` is identical inside and outside.

````python
class Person:
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender

def greet(person):
    print(id(person))    # address inside the function

p = Person("nitish", "male")
print(id(p))             # address outside
greet(p)                 # ...they are the same!
````

### Editing an object inside a function affects it outside too

Since it's the same object, a change made inside the function is visible outside.

````python
class Person:
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender

def greet(person):
    person.name = "ankit"    # change the name inside the function
    print(person.name)

p = Person("nitish", "male")
greet(p)
print(p.name)                # "ankit" — the change stuck (same object)
````

# 4️⃣ Object Mutability

Objects in Python are **mutable** by default — you can change them in place. Proof: after a function edits an object and returns it, the memory address stays the **same** (a new object was NOT created).

````python
class Person:
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender

def greet(person):
    person.name = "ankit"
    return person

p = Person("nitish", "male")
print(id(p))
p1 = greet(p)
print(id(p1))   # SAME address -> the object was changed in place (mutable)
````

# 5️⃣ Instance Variables

An **instance variable** is a variable whose value is **different for every object**.

🧠 Example: `p1.name` is `"nitish"` and `p2.name` is `"steeve"`. It's the *same* variable name (`name`), but each object has its **own** value. Instance variables are created **inside the constructor** with `self.`.

````python
class Person:
    def __init__(self, name_input, country_input):
        self.country = country_input
        self.name = name_input       # instance variable

p1 = Person("nitish", "india")
p2 = Person("steeve", "usa")

print(p1.name)   # each object has its OWN name
print(p2.name)
````

# 6️⃣ Encapsulation ⭐

**Encapsulation** means **bundling data together and protecting it** so it cannot be misused from outside the class.

### 📖 A little story

Imagine **you** are a senior developer who wrote a nice `Atm` class. A **careless junior developer** joins your project. They only use your class through an object — they can't see your code, but they *can* touch your object's data. And they do something silly...

Let's watch the problem unfold, then fix it step by step.

### Step 1 — The problem: outsiders can break your data

The junior developer sets `balance` to a **string**. No error appears... yet. But later, when `withdraw` tries to do `balance - amount`, the code **crashes** — because you can't subtract a number from a string.

````python
class Atm:
    def __init__(self):
        self.pin = ""
        self.balance = 0          # PUBLIC — anyone can change it

    def withdraw(self, amount):
        if amount <= self.balance:     # crashes if balance is not a number
            self.balance = self.balance - amount
            print("withdraw successful, balance is", self.balance)
        else:
            print("insufficient balance!")

obj = Atm()
obj.balance = "hehehe"    # 😈 the junior sets balance to a string — no error here
obj.withdraw(5000)        # 💥 but this crashes!
````

```
TypeError: '<=' not supported between instances of 'int' and 'str'
```
The bug is dangerous: **someone outside the class changed your internal data** and broke your logic.

### Step 2 — Make the data private with `__`

To protect a variable, put **two underscores `__`** in front of its name. This makes it **private** — it won't show up in auto-suggestions and can't be reached the normal way. You can make **methods** private the same way.

But here's the twist — even private variables can be *tricked*... let's see how Python actually protects them.

````python
class Atm:
    def __init__(self):
        self.pin = ""
        self.__balance = 0       # PRIVATE now (note the __)

obj = Atm()

# The junior tries to change it the normal way:
obj.__balance = "hehehe"        # this does NOT touch the real private variable!
print(obj.__balance)            # prints the fake one they just created
print(obj._Atm__balance)        # the REAL private value is safe and unchanged
````

### 🔍 What just happened? — Name Mangling

When you make a variable private with `__balance`, Python secretly **renames** it to `_Atm__balance` (class name + variable name). This is called **name mangling**.

- Your whole class code uses the real name `_Atm__balance`, so it stays safe.
- When the junior writes `obj.__balance = "hehehe"`, they accidentally create a **brand-new, unrelated** variable — the real one is untouched. 🛡️

> ⚠️ **Nothing in Python is *truly* private.** A determined person could still write `obj._Atm__balance = "hehehe"` and break it. Python's philosophy is: *"we are all adults here"* — privacy is a polite convention, not a locked door.

### Step 3 — Getters and Setters

Sometimes you *do* need to let outsiders **read** or **change** a private value — but in a **controlled, safe** way. For that we write two methods:

| Method | Job |
|--------|-----|
| **Getter** (`get_balance`) | safely **returns** the private value |
| **Setter** (`set_balance`) | safely **changes** the private value |

````python
class Atm:
    def __init__(self):
        self.pin = ""
        self.__balance = 0

    # getter — lets the outside READ the private value
    def get_balance(self):
        return self.__balance

    # setter — lets the outside CHANGE the private value
    def set_balance(self, new_value):
        self.__balance = new_value

obj = Atm()
print(obj.get_balance())     # read it safely -> 0
obj.set_balance(1000)        # change it safely
print(obj.get_balance())     # -> 1000
````

### Step 4 — The real power: validation inside the setter

A plain setter still lets someone set balance to `"hehehe"`. The **magic** is that a setter is **your** method, so you can put **checks (validation)** inside it. Now bad values are rejected!

````python
class Atm:
    def __init__(self):
        self.pin = ""
        self.__balance = 0

    def get_balance(self):
        return self.__balance

    def set_balance(self, new_value):
        if type(new_value) == int:        # only allow numbers
            self.__balance = new_value
        else:
            print("invalid value! balance must be a number")

obj = Atm()
print(obj.get_balance())     # 0
obj.set_balance("hehehe")    # rejected by the validation
````

### 📦 Definition of Encapsulation

> **Encapsulation** = bundling **data** together with the **methods (getter + setter)** that control access to it, and making the data **private** so it can only be touched in safe, controlled ways.

**Two habits of a good OOP programmer:**
1. Make your attributes **private** (`__`) by default.
2. Expose them only through **getters and setters** (with validation) when needed.

# 7️⃣ Collection of Objects

You can store objects inside a **list**, **dictionary**, or **tuple** — and then loop over them like any other collection.

````python
class Person:
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender

p1 = Person("nitish", "male")
p2 = Person("ankit", "male")
p3 = Person("ankita", "female")

# a LIST of objects
l = [p1, p2, p3]
for i in l:
    print(i.name, i.gender)
````

````python
# a DICTIONARY of objects
d = {"p1": p1, "p2": p2, "p3": p3}
for key in d:
    print(d[key].name, d[key].gender)
````

# 8️⃣ Instance vs Static Variables ⭐

**Scenario:** the manager wants every customer to get a **unique customer id** (1, 2, 3, ...). Let's try to do it with an instance variable — and see why it fails.

````python
class Atm:
    def __init__(self):
        self.pin = ""
        self.__balance = 0
        self.cid = 0        # instance variable
        self.cid += 1       # try to give a unique id

customer1 = Atm()
customer2 = Atm()
customer3 = Atm()

print(customer1.cid)
print(customer2.cid)
print(customer3.cid)
````

### ❌ Why did all three get id `1`?

Every time an object is created, the constructor runs **fresh**: it sets `cid = 0`, then `+1`, giving `1` — **every single time**. Each object has its **own separate** `cid`, so they can't count together.

The problem: **an instance variable cannot act as a shared counter.** We need a variable that is **shared by all objects** — a **static variable**.

### Instance vs Static

| | Instance variable | Static variable |
|--|-------------------|-----------------|
| Belongs to | the **object** | the **class** |
| Value is | **different** for each object | **same** for all objects |
| Written with | `self.name` (in constructor) | `ClassName.name` (in the class, outside methods) |
| Example | a customer's **balance**, a student's **CGPA** | the bank's **IFSC code**, the **college name** |

> 📌 **Rule of thumb:** `self.` in front → instance variable. `ClassName.` in front → static variable.

### ✅ The fix: a static variable as a shared counter

We declare `counter` **inside the class but outside all methods** — that makes it **static** (shared). Each new object reads the current counter for its id, then increases it for the next object.

````python
class Atm:
    counter = 0                      # STATIC variable (shared by all objects)

    def __init__(self):
        self.pin = ""
        self.__balance = 0
        self.cid = Atm.counter       # give this object the current counter value
        Atm.counter = Atm.counter + 1  # increase the shared counter for the next object

obj = Atm()          # cid 0, counter -> 1
customer1 = Atm()    # cid 1, counter -> 2
customer2 = Atm()    # cid 2, counter -> 3
customer3 = Atm()    # cid 3, counter -> 4

print(customer1.cid)
print(customer2.cid)
print(customer3.cid)
````

````python
# the static variable is shared and keeps the running total
Atm.counter   # 4 objects were created (obj + 3 customers)
````

Now each customer gets a **unique** id, because they all share the **same** `counter`. That is exactly what static variables are for. 🎯
