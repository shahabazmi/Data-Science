## ⭐ Problem — Maximum Sum Subarray

**Question:** Given a list of numbers (which can be positive *or* negative), find the **contiguous** part (a "subarray") whose numbers add up to the **largest** total, and return that total.

- **Contiguous** means the items must be **next to each other** — you cannot skip items.

**Our example list:**
```python
L = [-2, 4, 7, -1, 6, -11, 14, 3, -1, -6]
```

🧠 **Real-life idea:** imagine these numbers are the **daily profit/loss** of a shop. A negative number is a loss day, a positive number is a profit day. We want to find the **best continuous streak of days** that earned the most money in total.

Let's peek at the answer first, then learn *two* ways to find it.

### 👀 The answer for our list

The best continuous streak is `[4, 7, -1, 6, -11, 14, 3]`, which adds up to **22**.

Let's check a few streaks by hand to build intuition:

| Subarray | Sum |
|----------|-----|
| `[4, 7]` | 11 |
| `[4, 7, -1, 6]` | 16 |
| `[14, 3]` | 17 |
| `[4, 7, -1, 6, -11, 14, 3]` | **22** ✅ |

So the maximum sum is **22**. Notice it even includes a big loss (`-11`) in the middle — because the profits around it are large enough to be worth it. Now let's write code to find it automatically.

## 🐢 Method 1 — Brute Force (the simple, slow way)

**The idea:** simply **try every possible subarray**, calculate each one's sum, and keep the biggest.

- The outer loop `i` picks the **starting** point.
- The inner loop `j` picks the **ending** point.
- We add up the numbers from `i` to `j` and compare with the best sum so far.

This is easy to understand but slow, because it checks a huge number of subarrays.

````python
# Method 1: Brute force — try every possible subarray
L = [-2, 4, 7, -1, 6, -11, 14, 3, -1, -6]

max_sum = L[0]                     # start by assuming the first element is the best

for i in range(len(L)):            # i = starting point of the subarray
    for j in range(i, len(L)):     # j = ending point of the subarray
        current_sum = 0
        for k in range(i, j + 1):  # add up the numbers from i to j
            current_sum = current_sum + L[k]
        if current_sum > max_sum:  # found a bigger sum? remember it
            max_sum = current_sum

print("Maximum subarray sum is:", max_sum)
````

> ⏱️ **Speed:** this uses three nested loops, so it is **slow** for big lists (in Big-O terms, `O(n³)`). It works, but we can do much better.

## ⚡ Method 2 — Kadane's Algorithm (the smart, fast way)

**Kadane's Algorithm** is the famous, super-fast solution. It goes through the list **only once**.

**The idea (in plain words):**
As we walk through the numbers, we keep a **running total** (`current_sum`). At each number we ask one simple question:

> "Is it better to **add this number to my current streak**, or to **start a fresh streak from this number**?"

- If the running total has gone **negative**, it is only dragging us down — so we **throw it away** and start fresh from the current number.
- Along the way, we always remember the **best total** we have ever seen (`max_sum`).

🧠 **Shop analogy:** if your current profit streak has become negative, there is no point carrying that loss forward — better to start counting fresh from today.

````python
# Method 2: Kadane's Algorithm — go through the list only ONCE (fast!)
L = [-2, 4, 7, -1, 6, -11, 14, 3, -1, -6]

max_sum = L[0]         # best total we have ever seen
current_sum = L[0]     # total of the current streak

for i in range(1, len(L)):
    # either extend the current streak, or start fresh from L[i]
    current_sum = max(L[i], current_sum + L[i])
    # remember the best total we have seen so far
    max_sum = max(max_sum, current_sum)

print("Maximum subarray sum is:", max_sum)
````

### 🔬 Step-by-step: how Kadane's algorithm walks through the list

Watch how `current_sum` restarts whenever starting fresh is better, and how `max_sum` only ever goes **up** (never down):

| Number | `current_sum = max(num, current_sum + num)` | `max_sum` |
|--------|---------------------------------------------|-----------|
| -2 | -2 (start) | -2 |
| 4 | max(4, -2+4=2) → **4** (fresh start is better) | 4 |
| 7 | max(7, 4+7=11) → 11 | 11 |
| -1 | max(-1, 11-1=10) → 10 | 11 |
| 6 | max(6, 10+6=16) → 16 | 16 |
| -11 | max(-11, 16-11=5) → 5 | 16 |
| 14 | max(14, 5+14=19) → 19 | 19 |
| 3 | max(3, 19+3=22) → 22 | **22** ✅ |
| -1 | max(-1, 22-1=21) → 21 | 22 |
| -6 | max(-6, 21-6=15) → 15 | 22 |

> 💡 The final `max_sum` is **22** — matching our hand calculation. The key lesson is the **method**: at each number, decide "**extend** my streak or **restart** from here", and always keep the best total seen. Try printing `current_sum` and `max_sum` inside the loop to watch it live!

> ⏱️ **Speed:** Kadane's goes through the list just **once** (`O(n)`), so it is dramatically faster than brute force. This is why it is a favourite interview answer.

## 🎯 Summary

| | Brute Force | Kadane's Algorithm |
|--|-------------|--------------------|
| **Idea** | try every subarray | one pass: "extend or restart?" |
| **Speed** | slow (`O(n³)`) | fast (`O(n)`) |
| **When to use** | small lists / learning | the real, interview-ready answer |

**The one big lesson of Kadane's algorithm:** if your running total ever becomes negative, **drop it and start fresh** — carrying a loss forward never helps.
