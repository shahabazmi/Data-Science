## 1️⃣ What Are Lists?

A **list** is a data type where you can store **multiple items under one name**. More technically, a list acts like a **dynamic array** — meaning you can keep adding more items to it whenever you want.

🧠 **Real-life example:** imagine you are analysing the reviews of **10,000 movies**. It is impossible to make 10,000 separate variables (one for each movie) — that would waste huge amounts of space and effort. Instead, we make **one list** and store all the similar items together.

> 📌 Lists are perfect for storing a **group of related items** together.

**Example of a list:**
```python
l = [20, "jessica", 35.5, [30, 60, 70]]
```
Notice it can hold a number, a text, a decimal, and even *another list* — all at once!

## 2️⃣ Arrays vs Lists

You may have heard of **arrays** in languages like C. Lists are similar but more flexible. Here is the difference:

| Feature | Array | List (Python) |
|---------|-------|---------------|
| **Size** | Fixed — you must decide the size in advance | Dynamic — grows automatically, no limit |
| **Data types** | Homogeneous — only **one** type allowed | Heterogeneous — **any mix** of types allowed |
| **Speed** | Faster to run | Slightly slower |
| **Memory** | Uses less memory | Uses more memory |

**How does a dynamic array (list) grow?** It starts with a fixed size. As you add items, when it becomes full, Python creates a **new array of double the size**, copies the old items over, and continues. This repeats whenever it fills up — which is why a list can grow "endlessly" without you worrying about size.

## 3️⃣ Characteristics of a List

A Python list has these key properties:

| Property | Meaning |
|----------|---------|
| **Ordered** | Items stay in the order you put them |
| **Changeable (mutable)** | You *can* change a list after creating it |
| **Heterogeneous** | Can store different data types together |
| **Allows duplicates** | The same value can appear more than once |
| **Dynamic** | Grows automatically as you add items |
| **Can be nested** | A list can contain other lists |
| **Items can be accessed** | You can pull out any item by its position |
| **Holds any object** | Can contain any kind of Python object |

**"Ordered" in action:** because order matters, two lists with the same items in a *different order* are **not** equal.

````python
# ORDERED — order matters, so these two lists are NOT equal
l1 = [1, 2, 3]   # same items...
l2 = [3, 2, 1]   # ...but a different order
l1 == l2         # therefore not equal -> False
````

## 4️⃣ Creating a List

A list is written inside **square brackets `[ ]`**, with items separated by commas. Here are the many ways to create one:
- **Empty list** → `[]`
- **1D list** → a simple flat list
- **2D / nested list** → a list *inside* a list
- **3D list** → a list inside a list inside a list
- **Heterogeneous** → mixing different data types
- **Type conversion** → turning something (like a string) into a list

````python
# Different ways to create a list
# empty list
print([])
# 1D list (homogeneous — all items are the same type)
print([1, 2, 3, 4, 5])
# 2D / nested list (a list that contains at least one more list)
print([1, 2, 3, [4, 5]])
# 3D list (a list, inside a list, inside a list)
print([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
# heterogeneous list (a mix of different data types)
print([1, True, 5, 6, 5 + 6j, "hello"])
# using type conversion (turn a string into a list of characters)
print(list("hello"))
````

## 5️⃣ Accessing Items — Indexing & Slicing

There are two ways to pull items out of a list:
1. **Indexing** → get **one** item by its position
2. **Slicing** → get **many** items at once

### 🔢 Indexing

Just like strings, list positions start at **0**.
- **Positive indexing** → counts left to right, starting at `0`.
- **Negative indexing** → counts right to left, starting at `-1`.

For **nested lists**, we use multiple brackets: `l[3][0]` means "go to item 3, then inside it go to item 0".

````python
# INDEXING
# positive indexing (left to right, starts at 0)
l = [1, 2, 3, 4, 5]
print(l[0])                              # first item -> 1

# indexing into a 2D list
l1 = [1, 2, 3, [4, 5]]
print(l1[3][0])                          # item 3 is [4,5], then its item 0 -> 4

# indexing into a 3D list
l3 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(l3[0][0][1])                       # step in three times -> 2

# negative indexing (right to left, starts at -1)
print(l[-3])                             # third from the end -> 3

# negative indexing into a 2D list
l2 = [1, 2, 3, [4, 5]]
print(l2[-1])                            # last item is the whole list [4, 5]
print(l2[-1][-2])                        # inside [4,5], second from end -> 4
````

### ✂️ Slicing

Slicing grabs **many items at once** using `l[start:stop:step]`. It works exactly like string slicing — including the famous `l[::-1]` to **reverse** a list.

````python
# SLICING — extract multiple items at once
l = [1, 2, 3, 4, 5, 6]
print(l[0:3])       # items 0, 1, 2 -> [1, 2, 3]
print(l[-3:])       # last three items -> [4, 5, 6]
print(l[0::2])      # every 2nd item (positive) -> [1, 3, 5]
print(l[-5:-2:2])   # jumping with negative indexes -> [2, 4]
print(l[::-1])      # reverse the whole list -> [6, 5, 4, 3, 2, 1]
````

## 6️⃣ Adding Items to a List

There are **3 ways** to add items:

| Method | What it does |
|--------|--------------|
| **`append()`** | adds **one** item at the **end** |
| **`extend()`** | adds **multiple** items at the end |
| **`insert()`** | adds an item at a **specific position** |

````python
# 1. APPEND — adds ONE item at the end
l = [1, 2, 3, 4, 5]
l.append(6)
print(l)
l.append(True)
print(l)
# if you append a list, the WHOLE list is added as a single item
a = [1, 2, 3, 4, 5]
a.append([6, 7, 8])
print(a)

# 2. EXTEND — adds MULTIPLE items at the end (more powerful than append)
s = [1, 2, 3, 4, 5]
s.extend([6, 7, 8])
print(s)
k = [1, 2, 3, 4, 5]
k.extend("delhi")   # a string gets split into its characters
print(k)

# 3. INSERT — adds an item at a SPECIFIC position: insert(index, item)
d = [1, 2, 3, 4, 5]
d.insert(1, 100)    # put 100 at index 1
print(d)
````

## 7️⃣ Editing Items in a List

Unlike strings (which are immutable), a list **can be changed** after it is created. You can edit **one item** using its index, or **several items** using slicing.

````python
# EDITING a list (lists are mutable / changeable)
# edit one item using its index
l = [1, 2, 3, 4, 5]
l[-1] = 500
print(l)
# edit several items at once using slicing
l[1:4] = [200, 300, 400]
print(l)
````

## 8️⃣ Deleting Items from a List

Because lists are mutable, we can delete items in several ways:

| Method | What it does |
|--------|--------------|
| **`del`** | delete by index or slice (or the whole list) |
| **`remove()`** | delete by **value** (you give the item, not its index) |
| **`pop()`** | delete by index; **default is the last** item |
| **`clear()`** | empties the list (keeps the list, removes all items) |

````python
# 1. DEL — delete by indexing or slicing
l = [1, 2, 3, 4, 5]
print(l)
del l[-1]      # delete the last item by index
print(l)
del l[1:3]     # delete a slice
print(l)
del l          # delete the whole list variable
print(l)       # ❌ ERROR now: name 'l' is not defined (it was deleted)
````

````python
# 2. REMOVE — delete by VALUE (you don't need the index)
l = [1, 2, 3, 4, 5]
l.remove(5)    # remove the value 5
print(l)
````

````python
# 3. POP — delete by index; if no index is given, removes the LAST item
l = [1, 2, 3, 4, 5]
l.pop(0)       # remove the item at index 0
print(l)
l.pop()        # no index -> remove the last item
print(l)
````

````python
# 4. CLEAR — keeps the list but removes ALL items inside it
l = [1, 2, 3, 4, 5]
l.clear()
print(l)       # empty list -> []
````

## 9️⃣ Operators on Lists

Three kinds of operators work with lists:
1. **Arithmetic** (`+`, `*`)
2. **Membership** (`in`, `not in`)
3. **Loops**

### ➊ Arithmetic — only `+` and `*`
- `+` **joins** (concatenates) two lists into one.
- `*` **repeats** a list.

````python
# ARITHMETIC on lists: only + and * work
l1 = [1, 2, 3, 4]
l2 = [5, 6, 7, 8]
print(l1 + l2)   # concatenation (merge two lists)
print(l1 * 3)    # repeat the list 3 times
````

### ➋ Membership — `in` / `not in`

Checks whether an item exists in the list. Notice that `5` is *not* directly in `l2` (it is hidden inside the nested list `[5, 6]`), so `5 in l2` is `False`, but `[5, 6] in l2` is `True`.

````python
# MEMBERSHIP operators
l1 = [1, 2, 3, 4, 5]
l2 = [1, 2, 3, 4, [5, 6]]
print(5 in l1)        # True
print(5 in l2)        # False (5 is hidden inside the nested list, not directly present)
print([5, 6] in l2)   # True (the nested list [5,6] is a direct item)
````

### ➌ Loops on lists

A `for` loop goes through a list one item at a time. With **nested lists**, each inner list is treated as a single item.

````python
# LOOPS on lists
l1 = [1, 2, 3, 4, 5]
for i in l1:
    print(i)

# looping over a 2D list — the nested list is one item
l2 = [1, 2, 3, 4, [5, 6]]
for i in l2:
    print(i)

# looping over a 3D list
l3 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
for i in l3:
    print(i)
````

## 🔟 List Functions

Some functions (`len`, `min`, `max`, `sorted`) are **universal** — they work on all collection types (lists, sets, tuples, dictionaries). You already learned some list methods (`clear`, `pop`, `remove`, `append`, `extend`, `insert`). Now let's cover the rest:

| Function | What it does |
|----------|--------------|
| `len` / `min` / `max` / `sorted` | length / smallest / largest / a sorted copy |
| `count()` | how many times a value appears |
| `index()` | the position of the first occurrence |
| `reverse()` | reverses the list **permanently** |
| `sort()` vs `sorted()` | `sort` changes the list; `sorted` returns a copy |
| `copy()` | makes a separate copy of the list |

````python
# len / min / max / sorted
l = [2, 3, 7, 4]
print(len(l))                    # number of items -> 4
print(min(l))                    # smallest -> 2
print(max(l))                    # largest -> 7
# sorted() returns a NEW sorted list; the original 'l' stays the same
print(sorted(l))                 # ascending (default)
print(sorted(l, reverse=True))   # descending
````

````python
# COUNT — how many times a value appears
l = [1, 2, 1, 3, 4, 1, 5]
print(l.count(1))   # 1 appears 3 times
print(l.count(5))   # 5 appears 1 time
print(l.count(3))   # 3 appears 1 time
````

````python
# INDEX — the position of a value (only the FIRST occurrence)
l = [1, 2, 3, 4, 1, 5, 3]
print(l.index(2))   # 2 is at index 1
print(l.index(1))   # 1 appears twice, but index() gives only the first -> 0
````

**`reverse()`** reverses the list **in place** (permanently). *(The original notes didn't print the result — here we add a `print` so you can see it.)*

````python
# REVERSE — reverses the list permanently (in place)
l = [2, 1, 5, 7, 8]
l.reverse()
print(l)   # the list itself is now reversed
````

### ⭐ `sort()` vs `sorted()` — an important difference

- **`sort()`** changes the original list **permanently** (in place).
- **`sorted()`** does **not** change the original — it returns a **new** sorted list.

````python
# sort vs sorted
l = [2, 1, 5, 7, 8]
print(l)            # original
print(sorted(l))    # sorted() returns a NEW sorted list...
print(l)            # ...and the original is unchanged
l.sort()            # sort() changes the list permanently
print(l)            # now the original itself is sorted
````

### 📑 `copy()`

`copy()` makes a **separate copy** of the list at a **new memory address**. We can check the memory address with `id()` — notice the two IDs are different, proving they are separate lists.

````python
# COPY — makes a separate copy at a new memory address
l = [2, 1, 5, 7, 8]
print(l)
print(id(l))       # memory address of the original
l1 = l.copy()
print(l1)
print(id(l1))      # a DIFFERENT memory address -> it is a separate copy
````

## 1️⃣1️⃣ List Comprehension ⭐

**List comprehension** is a **short, elegant way to create a list** in a single line, instead of writing a full `for` loop.

**The formula:**
```python
newlist = [expression for item in iterable if condition == True]
```
- **expression** → what to put in the list
- **item / iterable** → what we loop over
- **if condition** → (optional) only include items that pass a test

**✅ Advantages:**
- More time-efficient and space-efficient than loops
- Fewer lines of code
- Turns a whole loop into a neat one-line formula

Let's see it grow from a normal loop into a comprehension.

````python
# Example: add numbers 1 to 10 into a list
# The long way, using a loop:
l = []
for i in range(1, 11):
    l.append(i)
print(l)

# The short way, using list comprehension:
l = [i for i in range(1, 11)]   # 'i' is the expression, range(1,11) is the iterable
print(l)
````

### Example — Scalar Multiplication of a Vector

Careful: multiplying a list by a number **repeats** it (and a *negative* number gives an empty list). To actually multiply each element, we need a loop or a comprehension.

````python
# scalar multiplication on a vector
v = [2, 3, 4]
s = -3
print(v * s)   # ⚠️ this gives [] — list * number REPEATS, it does not multiply elements
# we want [-6, -9, -12], so we must multiply each item:
x = []
for i in v:
    x.append(i * s)
print(x)

# the same thing, shorter, with list comprehension:
print([s * i for i in v])   # expression is s*i, iterable is v
````

````python
# Add squares of each number
l = [1, 2, 3, 4, 5]
[i ** 2 for i in l]
````

````python
# Print all numbers divisible by 5 in the range 1 to 50 (using an 'if' condition)
[i for i in range(1, 51) if i % 5 == 0]
````

````python
# Find languages that start with the letter 'p'
languages = ["java", "python", "php", "javascript", "c"]
[language for language in languages if language.startswith("p")]
````

````python
# Two conditions: fruits that are in the basket AND start with 'a'
basket = ["apple", "guava", "cherry", "banana"]
my_fruits = ["apple", "kiwi", "grapes", "banana"]
[fruit for fruit in my_fruits if fruit in basket if fruit.startswith("a")]
````

### Advanced — Nested List Comprehension

You can even build a **matrix** (a list of lists) or a **cartesian product** using nested comprehensions. These are advanced, so read them slowly.

````python
# Build a 3x3 matrix using nested list comprehension
[[i * j for i in range(1, 4)] for j in range(1, 4)]
````

````python
# Cartesian product — combine every item of l1 with every item of l2
l1 = [1, 2, 3, 4]
l2 = [5, 6, 7, 8]
[i * j for i in l1 for j in l2]
````

## 1️⃣2️⃣ Two Ways to Traverse (Go Through) a List

1. **Item-wise** → the loop gives you each **item** directly.
2. **Index-wise** → the loop gives you each **position number**, and you use it to reach the item.

````python
# Two ways to traverse a list
# 1. item-wise (you get the items)
l = [1, 2, 3, 4]
for i in l:
    print(i)

# 2. index-wise (you get the positions 0, 1, 2, 3)
l = [1, 2, 3, 4]
for i in range(0, len(l)):
    print(i)
````

## 1️⃣3️⃣ The `zip()` Function

`zip()` **pairs up** items from two (or more) lists — the first item of each, then the second of each, and so on.

🧠 **Analogy:** think of a **zipper** on a jacket — it joins the left teeth and the right teeth, one pair at a time.

> 📌 If the lists have **different lengths**, `zip` stops at the **shorter** one.

Below we use `zip` to add two lists **index-wise** (first + first, second + second, ...).

````python
# Add the items of two lists index-wise, using zip()
l1 = [1, 2, 3, 4]
l2 = [-1, -2, -3, -4]

print(list(zip(l1, l2)))            # see the pairs: [(1,-1), (2,-2), (3,-3), (4,-4)]
print([i + j for i, j in zip(l1, l2)])   # add each pair -> [0, 0, 0, 0]
````

## 1️⃣4️⃣ Disadvantages of Python Lists

Lists are powerful, but they are not perfect. Keep these in mind:

| Disadvantage | Why |
|--------------|-----|
| **Slow** | Slower than arrays for heavy number-crunching |
| **Risky usage** | Because they can hold *anything*, bugs can sneak in |
| **Eats up more memory** | Lists use more memory than arrays |

> 💡 This is exactly why, later in Data Science, we use **NumPy arrays** for large numerical data — they are faster and use less memory.
