# 🅰️ Part A — Tuples

## What is a Tuple?

A **tuple** is very similar to a list. The **one big difference**: once a tuple is created, you **cannot change** its elements — a list can be changed, a tuple cannot.

> 🧠 **In short: a tuple is an *immutable* list.** Once created, it can never be changed in any way.

### Characteristics of a tuple

| Property | Meaning |
|----------|---------|
| **Ordered** | Order matters — `(1,2,3)` is not the same as `(3,2,1)` |
| **Unchangeable (immutable)** | Cannot be edited after creation |
| **Allows duplicates** | The same value can appear more than once |

## Creating Tuples

Tuples are written inside **round brackets `( )`**.

> ⚠️ **Very important gotcha:** to make a tuple with a **single item**, you MUST put a comma at the end: `("hello",)`. Without the comma, `("hello")` is just a normal string, **not** a tuple!

````python
# Creating tuples in different ways
# empty tuple
t1 = ()
print(t1)

# ⚠️ single item WITHOUT a comma -> this is NOT a tuple, it is a string!
t2 = ("hello")
print(type(t2))    # <class 'str'>

# correct way to make a single-item tuple: add a comma at the end
t3 = ("hello",)
print(t3)
print(type(t3))    # <class 'tuple'>

# homogeneous tuple (all same type)
t4 = (1, 2, 3, 4)
print(t4)

# heterogeneous tuple (mixed types)
t5 = (1, 2, 3, True, [1, 2, 3])
print(t5)

# 2D tuple (a tuple inside a tuple)
t6 = (1, 2, 3, 4, (5, 6, 7))
print(t6)

# using type conversion (string -> tuple of characters)
t7 = tuple("hello")
print(t7)
````

## Accessing Items — Indexing & Slicing

Accessing items from a tuple works **exactly like a list** — using indexing (one item) and slicing (many items).

````python
# Indexing and slicing a tuple
t1 = (1, 2, 3, 4)
print(t1[0])       # first item -> 1
print(t1[-1])      # last item -> 4

# slicing
print(t1[0:4])     # (1, 2, 3, 4)
print(t1[0:4:2])   # every 2nd item -> (1, 3)
print(t1[-3:-1])   # (2, 3)
print(t1[::-1])    # reversed -> (4, 3, 2, 1)

# accessing a tuple inside a tuple
t2 = (1, 2, 3, 4, (5, 6))
print(t2[-1])      # last item is the tuple (5, 6)
print(t2[-1][0])   # inside it, first item -> 5
````

## Editing, Adding & Deleting

Because tuples are **immutable**, most changes are simply **not allowed**.

**Editing → not allowed.** Trying to change an item gives an error (just like with strings).

````python
# Editing a tuple -> ERROR (tuples are immutable)
t1 = (1, 2, 3, 4)
t1[0] = 100   # ❌ TypeError: 'tuple' object does not support item assignment
````

**Adding → not possible.** Tuples have no `append`, `add`, or `insert` methods. The only option is to build a **new** tuple (e.g. by joining with `+`).

````python
# Tuples are immutable, so there is NO append/add/insert.
t1 = (1, 2, 3, 4)
print(t1)
# t1.append(5)   would fail: 'tuple' object has no attribute 'append' 
````

**Deleting → only the whole tuple.** You can delete the entire tuple with `del`, but you cannot delete a part of it.

````python
# Deleting a whole tuple with del
t1 = (1, 2, 3, 4)
del t1
# print(t1)   would now error: name 't1' is not defined (the whole tuple was deleted)
````

## Operations on Tuples

Just like lists: `+` joins, `*` repeats, `in` checks membership, and a `for` loop goes through each item.

````python
# Operations on tuples
t1 = (1, 2, 3, 4)
t2 = (5, 6, 7, 8)
print(t1 + t2)     # join two tuples
print(t1 * 2)      # repeat the tuple

print(1 in t1)     # membership -> True

# iteration
for i in t1:
    print(i)
````

## Tuple Functions

Same universal functions as lists: `len`, `min`, `max`, `sum`, `sorted`, plus `count` and `index`.

````python
# Tuple functions
t = (1, 2, 3, 4)
print(len(t))                    # 4
print(min(t))                    # 1
print(max(t))                    # 4
print(sum(t))                    # 10
print(sorted(t))                 # [1, 2, 3, 4] (returns a list)
print(sorted(t, reverse=True))   # [4, 3, 2, 1]
print(t.count(3))                # how many times 3 appears -> 1
print(t.index(3))                # position of 3 -> 2
````

````python
# index() raises an error if the value is not found
t = (1, 2, 3, 4)
t.index(6)   # ❌ ValueError: tuple.index(x): x not in tuple
````

## Lists vs Tuples ⭐ (Common Interview Question)

| Feature | List | Tuple |
|---------|------|-------|
| **Syntax** | square brackets `[ ]` | round brackets `( )` |
| **Mutability** | mutable (can change) | immutable (cannot change) |
| **Speed** | slower | **faster** (immutable types are faster) |
| **Memory** | uses more memory | uses **less** memory |
| **Built-in functions** | more methods available | fewer methods |
| **Error-prone** | more error-prone to debug | safer |

Let's *prove* some of these differences with code.

````python
# SPEED test — looping over a huge list vs a huge tuple
import time
l = list(range(10000000))
t = tuple(range(10000000))

start = time.time()
for i in l:
    i * 5
end = time.time()
print("time taken by list", end - start)

start = time.time()
for i in t:
    i * 5
end = time.time()
print("time taken by tuple", end - start)
````

````python
# MEMORY test — a tuple uses slightly less memory than a list
import sys
l = list(range(10000000))
t = tuple(range(10000000))
print("size of list", sys.getsizeof(l))
print("size of tuple", sys.getsizeof(t))
````

### Why lists are more "error-prone"

When you copy a list with `b = a`, both names point to the **same** list in memory. So changing one changes the other! With a tuple, since it is immutable, reassigning creates a brand-new tuple — the original stays safe.

````python
# LISTS — b = a makes both point to the SAME list
a = [1, 2, 3]
b = a
a.append(4)
print(a)   # [1, 2, 3, 4]
print(b)   # [1, 2, 3, 4] -> b changed too! (both share the same memory)

# TUPLES — reassigning makes a new tuple; the original is untouched
a = (1, 2, 3)
b = a
a = (4, 5, 6)
print(a)   # (4, 5, 6)
print(b)   # (1, 2, 3) -> b is safe
````

## Special Tuple Syntax ⭐

Tuples have some neat tricks that make Python code short and elegant.

### Tuple unpacking — assign many variables at once

You can "unpack" a tuple straight into variables. But the **number of variables must match** the number of items, or you get an error.

````python
# TUPLE UNPACKING — put each item into its own variable
a, b, c = (1, 2, 3)
print(a, b, c)

# ⚠️ if the count does not match, you get an error
c, d = (4, 5, 6)   # ❌ ValueError: too many values to unpack (3 values, 2 variables)
````

### Swapping values — the elegant Python way

No temporary variable needed! Python swaps two values in a single line.

````python
# SWAPPING two values using tuple syntax
a = 1
b = 2
a, b = b, a       # swap in one line
print(a, b)
````

### The `*` operator — collect the "rest"

If you only care about a few values, grab them into named variables and dump the **rest** into a variable using `*`.

````python
# Grab the first two, collect the rest with *
a, b, *others = (1, 2, 3, 4)
print(a, b)       # 1 2
print(others)     # [3, 4] -> the leftover items as a list
````

### Zipping tuples

`zip()` pairs up items from two tuples, exactly like it did with lists.

````python
# ZIP two tuples together
a = (1, 2, 3, 4)
b = (5, 6, 7, 8)
list(zip(a, b))
````

# 🅱️ Part B — Sets

## What is a Set?

A **set** is a collection of items with two special rules:
1. Every element is **unique** — **no duplicates** allowed.
2. Each element must be **immutable** (so you can't put a list inside a set).

However, the **set itself is mutable** — you *can* add or remove items from it.

Sets are also great for **mathematical operations** like union, intersection, and difference.

### Characteristics

| Property | Meaning |
|----------|---------|
| **Unordered** | No fixed order — `{1,2,3}` equals `{3,2,1}` |
| **Mutable** | You can add or remove items |
| **No duplicates** | Repeated values are automatically removed |
| **Only immutable items** | Can't hold a list or another set (no 2D set) |

## Creating a Set

Sets use **curly braces `{ }`**.

> ⚠️ **Big gotcha:** empty `{}` creates a **dictionary**, NOT a set! To make an empty set, you must use `set()`.

````python
# ⚠️ empty {} is a DICTIONARY, not a set
s = {}
print(type(s))     # <class 'dict'>

# to make an empty SET, use set()
p = set()
print(type(p))     # <class 'set'>
````

````python
# Creating sets in different ways
# 1D set
s1 = {1, 2, 3}
print(s1)

# heterogeneous set. Note: True equals 1, so the duplicate is dropped
s3 = {1, 2, 3, True, "hello"}
print(s3)

# a tuple CAN be inside a set (tuples are immutable)
s4 = {1, 2, 3, (4, 5)}
print(s4)

# type conversion from a list
s5 = set([1, 2, 3])
print(s5)

# duplicates are removed automatically
s6 = {1, 1, 2, 3, 4, 3}
print(s6)

# a set CANNOT hold a list: {1,2,[3,4]} would raise an error (lists are mutable)

# unordered: order does not matter, so these are equal
print({1, 2, 3} == {3, 2, 1})
````

## Accessing & Editing Items

Because a set is **unordered**, there is no concept of position. So **indexing and slicing do not work**, and for the same reason you **cannot edit** an item by index. You can only add or remove whole items.

## Adding Items

| Method | What it does |
|--------|--------------|
| `add()` | adds **one** item (you can't choose the position — sets are unordered) |
| `update()` | adds **multiple** items (pass them inside a list) |

````python
# Adding items to a set
s1 = {1, 2, 3, 4}
s1.add(5)              # add one item
print(s1)

s1.update([5, 6, 7])   # add multiple items (duplicates like 5 are ignored)
print(s1)
````

## Deleting Items

We have several options:

| Method | What it does | If item is missing |
|--------|--------------|--------------------|
| `del` | deletes the **whole set** | — |
| `discard()` | removes an item | **no error** |
| `remove()` | removes an item | **raises an error** |
| `pop()` | removes a **random** item (set is unordered) | — |
| `clear()` | empties the set | — |

````python
# discard — removes an item; NO error if it is missing
s = {1, 2, 3, 4}
s.discard(4)
print(s)          # {1, 2, 3}
s.discard(100)    # 100 not present -> no error
print(s)          # {1, 2, 3}

# remove — like discard, BUT raises an error if the item is missing
s = {1, 2, 3, 4}
s.remove(3)
print(s)          # {1, 2, 4}

# pop — removes a random item (because sets are unordered)
s = {1, 2, 3, 4}
s.pop()
print(s)          # one item removed, e.g. {2, 3, 4}

# clear — empties the set completely
s = {1, 2, 3, 4}
s.clear()
print(s)          # set()
````

## Set Operations ⭐ (the maths part)

This is where sets shine. Think of two overlapping circles (a Venn diagram):

| Operation | Symbol | Meaning |
|-----------|--------|---------|
| **Union** | `\|` | all items from **both** sets |
| **Intersection** | `&` | only the **common** items |
| **Difference** | `-` | items in the first set but **not** the second |
| **Symmetric Difference** | `^` | items in either set but **not both** (removes the common ones) |

````python
# SET OPERATIONS
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8}

print(s1 | s2)    # union -> everything from both
print(s1 & s2)    # intersection -> only common items {4, 5}
print(s1 - s2)    # difference -> in s1 but not s2 {1, 2, 3}
print(s2 - s1)    # in s2 but not s1 {6, 7, 8}
print(s1 ^ s2)    # symmetric difference -> everything except the common ones

# membership
print(1 in s1)        # True
print(1 not in s2)    # True

# iteration
for i in s1:
    print(i)
````

## Set Functions

Besides the universal `len/max/min/sorted`, sets have **method versions** of each operation. The methods ending in **`_update`** change the set **permanently**, while the plain versions return a **new** set.

````python
# len / max / min / sorted
s = {1, 2, 3, 4, 5}
print(len(s))                    # 5
print(max(s))                    # 5
print(min(s))                    # 1
print(sorted(s))                 # [1, 2, 3, 4, 5]
print(sorted(s, reverse=True))   # [5, 4, 3, 2, 1]
````

````python
# union / update
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8}
print(s1.union(s2))   # returns a NEW set (same as s1 | s2)
s1.update(s2)         # update = a union that CHANGES s1 permanently
print(s1)
print(s2)             # s2 is unchanged
````

````python
# intersection / intersection_update
s2 = {1, 2, 3, 4, 5}
s3 = {4, 5, 6, 7, 8}
print(s2.intersection(s3))    # returns a new set -> {4, 5}
s2.intersection_update(s3)    # changes s2 permanently
print(s2)
print(s3)
````

````python
# difference / difference_update
s4 = {1, 2, 3, 4, 5}
s5 = {4, 5, 6, 7, 8}
print(s4.difference(s5))     # returns a new set -> {1, 2, 3}
s4.difference_update(s5)     # changes s4 permanently
print(s4)
print(s5)
````

````python
# symmetric_difference / symmetric_difference_update
s6 = {1, 2, 3, 4, 5}
s7 = {4, 5, 6, 7, 8}
print(s6.symmetric_difference(s7))   # new set -> {1, 2, 3, 6, 7, 8}
s6.symmetric_difference_update(s7)   # changes s6 permanently
print(s6)
print(s7)
````

````python
# isdisjoint / issubset / issuperset / copy
# isdisjoint -> True if the two sets have NO items in common
s1 = {1, 2, 3, 4}
s2 = {7, 8, 5, 6}
print(s1.isdisjoint(s2))   # True

# issubset -> is s4 completely inside s3?
s3 = {1, 2, 3, 4, 5}
s4 = {2, 3, 4}
print(s4.issubset(s3))     # True

# issuperset -> does s5 completely contain s6?
s5 = {1, 2, 3, 4, 5}
s6 = {2, 3, 4}
print(s5.issuperset(s6))   # True

# copy -> a separate copy at a new memory address
s7 = {1, 2, 3}
s8 = s7.copy()
print(s7)
print(s8)
````

## Frozenset

A **frozenset** is just an **immutable version** of a set. Once created, it **cannot be changed** (no add/remove). Because it is immutable, a frozenset **can** be placed inside another set — allowing a kind of "2D set".

````python
# Create a frozenset (an immutable set)
fs = frozenset([1, 2, 3])
print(fs)
````

````python
# A frozenset can even be nested inside another frozenset (since it is immutable)
fs = frozenset([1, 2, frozenset([3, 4])])
print(fs)
````

## Set Comprehension

Just like list comprehension, but with curly braces `{ }` — it builds a **set** in one line.

````python
# Set comprehension examples
print({i for i in range(1, 11)})            # a set of numbers 1 to 10
print({i for i in range(1, 11) if i > 5})   # only numbers greater than 5
print({i ** 2 for i in range(1, 11)})       # squares
````

# 🅲 Part C — Dictionaries

## What is a Dictionary?

A **dictionary** stores data as **key → value pairs**. Instead of just single values, each item has a **key** (a label) and a **value**.

🧠 **Real-life example:** think of a real dictionary — you look up a *word* (the key) to find its *meaning* (the value). Or a student record: `"name" → "shahab"`, `"age" → 24`.

In some other languages, this is called a **map** or an **associative array**.

```python
d = {"name": "shahab", "age": 24, "gender": "male"}
```

### Characteristics

| Property | Meaning |
|----------|---------|
| **Mutable** | Can be changed after creation |
| **Indexing has no meaning** | You access by **key**, not by position |
| **Keys can't be duplicated** | Each key must be unique |
| **Keys must be immutable** | A key can be a string/number/tuple, but **not** a list |

> 📌 **Remember:** lists, tuples, and sets store **single items**. A dictionary stores items in **pairs** (key + value).

## Creating a Dictionary

Dictionaries use **curly braces `{ }`** with `key: value` pairs.

````python
# Creating dictionaries
# empty dict
d = {}
print(d)

# 1D dictionary
d1 = {"name": "shahab", "gender": "male"}
print(d1)

# a 2D dictionary (a dictionary inside a dictionary)
student = {"name": "shahab",
           "college": "snu",
           "semester": 4,
           "subject": {
               "maths": 50,
               "science": 60,
               "english": 70
           }}
print(student)

# duplicate keys -> the LAST value wins
ds = {"name": "shahab", "name": "rahul"}
print(ds)

# a tuple CAN be a key (immutable), but a list CANNOT
ms = {"name": "shahab", (1, 2, 3): 2}
print(ms)
# {"name": "shahab", [1, 2, 3]: 2} would raise an error (list keys are not allowed)
````

## Accessing Items

You access a value by its **key** (not by index). There are two ways:
- **`d["key"]`** → gives the value, but **errors** if the key is missing.
- **`d.get("key")`** → gives the value, and returns **`None`** (no error) if the key is missing.

````python
# Accessing dictionary values by key
my_dict = {"name": "shahab", "age": 24}
# my_dict[0]  -> indexing does NOT work; dictionaries use keys, not positions
print(my_dict["name"])       # shahab
print(my_dict["age"])        # 24

# .get() is safer — it returns None instead of an error if the key is missing
print(my_dict.get("name"))   # shahab
print(my_dict.get("age"))    # 24
````

## Adding a Key-Value Pair

To add a new pair, just assign a value to a new key: `d["newkey"] = value`.

````python
# Adding new key-value pairs
d = {"name": "shahab", "age": 24}
d["gender"] = "male"    # add a new pair
d["weight"] = 70        # add another
print(d)
````

## Removing a Key-Value Pair

| Method | What it does |
|--------|--------------|
| `pop("key")` | removes a **specific** key |
| `popitem()` | removes the **last** inserted pair |
| `del d["key"]` | removes a specific key (like `pop`) |
| `clear()` | removes **all** pairs (empties the dict) |

````python
# Removing key-value pairs
d = {"name": "shahab", "age": 24, "gender": "male", "weight": 70}
d.pop("name")     # remove a specific key
print(d)

d.popitem()       # remove the LAST pair
print(d)

d2 = {"name": "shahab", "age": 24, "gender": "male", "weight": 70}
del d2["name"]    # remove a specific key using del
print(d2)

d2.clear()        # empty the whole dictionary
print(d2)
````

## Accessing & Editing a 2D (Nested) Dictionary

For a dictionary inside a dictionary, we chain the keys: `student["subject"]["maths"]` means "go to `subject`, then get `maths`".

````python
# Working with a nested (2D) dictionary
student = {"name": "shahab", "college": "snu", "semester": 4,
           "subject": {"maths": 50, "science": 60, "english": 70}}

print(student["subject"]["maths"])     # access a nested value -> 50

student["subject"]["maths"] = 75       # edit a nested value
print(student)

del student["subject"]["maths"]        # delete a nested value
print(student)
````

## Editing a Key-Value Pair

To edit, just assign a new value to an existing key.

````python
# Editing values in a dictionary
student = {"name": "shahab", "college": "snu", "semester": 4,
           "subject": {"maths": 50, "science": 60, "english": 70}}

student["semester"] = 5              # edit a top-level value
print(student)

student["subject"]["science"] = 55   # edit a nested value
print(student)
````

## Dictionary Operations

- **Membership** (`in`) works on **keys**, not values.
- **Looping** over a dictionary gives you the **keys**; use `d[key]` to get each value.

````python
# Membership and iteration
d = {"name": "shahab", "age": 24, "gender": "male"}

print("shahab" in d)   # membership checks KEYS -> "shahab" is a value, so False
print("name" in d)     # "name" is a key -> True

# looping gives the keys; use d[i] to get the value
for i in d:
    print(i, d[i])
````

## Dictionary Functions

| Function | What it does |
|----------|--------------|
| `len(d)` | number of pairs |
| `sorted(d)` | sorts the **keys** |
| `d.items()` | all key-value pairs (as tuples) |
| `d.keys()` | all the keys |
| `d.values()` | all the values |
| `d.update(d2)` | merges another dictionary into this one |

````python
# Dictionary functions
d = {"name": "shahab", "age": 24, "gender": "male"}
print(len(d))         # 3
print(sorted(d))      # sorts the KEYS -> ['age', 'gender', 'name']

print(d.items())      # all pairs as tuples
print(d.keys())       # all keys
print(d.values())     # all values

# update -> merge another dict in (matching keys get overwritten)
d1 = {1: 2, 3: 4, 4: 5}
d2 = {4: 7, 6: 8}
d1.update(d2)         # key 4 is updated to 7, key 6 is added
print(d1)
````

## Dictionary Comprehension

Just like list/set comprehension, but it builds **key: value** pairs.

**The formula:**
```python
{key: value for variable in iterable}
```

````python
# Print the first 10 numbers and their squares
{i: i ** 2 for i in range(1, 11)}
````

````python
# Build a new dict from an existing one (convert km to miles by * 0.62)
distances = {"delhi": 1000, "mumbai": 2000, "banglore": 3000}
{key: value * 0.62 for (key, value) in distances.items()}
````

````python
# Using zip — pair each day with its temperature
days = ["sunday", "monday", "tuesday", "wednesday", "thursday", "friday", "saturday"]
temp = [30.5, 32.6, 31.8, 33.4, 29.8, 30.2, 29.9]
{i: j for (i, j) in zip(days, temp)}
````

````python
# Using an if condition — keep only products that are in stock (value > 0)
products = {"phone": 10, "laptop": 0, "charger": 70, "tablet": 5}
{key: value for (key, value) in products.items() if value > 0}
````

````python
# Nested comprehension — multiplication tables for numbers 2 to 4
{i: {j: i * j for j in range(1, 11)} for i in range(2, 5)}
````

## 🎯 Summary — Quick Revision

Amazing work! 🎉 You now know all four core Python collections. Here is the big picture:

| | List | Tuple | Set | Dictionary |
|--|------|-------|-----|------------|
| **Brackets** | `[ ]` | `( )` | `{ }` | `{key: value}` |
| **Ordered?** | Yes | Yes | No | insertion order kept |
| **Changeable?** | Yes | **No** | Yes | Yes |
| **Duplicates?** | Yes | Yes | **No** | keys: No |
| **Best for** | a changeable list of items | fixed data that won't change | unique items + maths ops | labelled data (key→value) |

**Key takeaways:**
- **Tuple** = an immutable list; faster, safer, uses less memory. Watch the single-item comma `("x",)`.
- **Set** = unique items, great for union/intersection/difference. Empty set is `set()`, not `{}`.
- **Dictionary** = key → value pairs; access by key, not position. Keys must be unique and immutable.
- All three support **comprehensions** for building them in one clean line.
